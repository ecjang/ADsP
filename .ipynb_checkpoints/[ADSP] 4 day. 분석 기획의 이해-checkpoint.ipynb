<h1><center>1절. 빅데이터 상식</center></h1>

분석 기획은 어떤 [What 목표]를 달성하기 [Why 위해]  어떤 [How 방법]을 수행할지에 대한 일련의 계획을 수립하는 단계이다. 어떤 일이든 그 목적과 목표를 정하는 것은 중요한 일이다. 

## 1.  분석 기획의 종류


### 1_1. 분석 종류

분석 종류는 [What 분석 대상]과 [How 분석 방법]에 따라 4가지로 나눈다.

* [Optimazation 최적화] : 분석 대상과 방법 모두 알 때 

* [Solusiton 솔루션] : 분석 대상을 알지만 방법을 모를 때

* [Insight 통찰] : 분석 대상을 모르지만 방법을 알 때

* [Discovery  발견] : 분석 대상과 방법 모두 모를 때

### 1_2. 분석 시점

분석 시점별로 과제단위의 [과제 중심점 연구]와 마스터플랜 단위의 [장기적 마스터플랜] 방식으로 나눈다.

* [과제 단위] : 당면한 과제 분석 주제 해결.

* [마스터플랜] : 지속적 분석 문화 내제화.

### 1_3. 요구 역량

분석가는 3가지 기본 역량(분석기술, 도메인 전문성, 의사소통)과 추가적으로 프로젝트 관리 역량, 러더쉽이 필요하다.

## 2. 고려사항


### 2_1. 유즈 케이스

활용 가능한 유즈 케이스의 탐색이 필요 : "바퀴를 재발명하지 마라" 기존에 잘 구현되어 있는 유사 분석 시나리오 및 솔루션을 활용하는 것이 중요.



### 2_2. 변화관리

충분하고 지속적인 교육 및 활용방안 등의 [Change Management 변화관리]가 필요.

<h1><center>2절 분석 방법론</center></h1>

## 1. 분석 방법론 개요
데이터 분석의 성공적인 내제화를 위해 체계화한 절차와 방법이 정리된 데이터 분석 방법론이 필요하다.

* 방법론의 생성 과정 : 암묵지 -> 형식화 -> 형식지 -> 체계화 -> 방법론 -> 내제화



### 1_1. 방법론 모델

* [Waterfall Model 폭포수 모델] : 한 단계가 완료 후 다음 단계로 진행하는 방식으로 문제가 있을 경우 피드백을 통해 해결한다.

* [Spiral Model 나선형 모델] : 반복을 통해 점진적으로 개발하는 방식. 처음 시도하는 프로젝트에 용이하지만 관리체계가 없으면 복잡도가 상승.

* [Prototype Model 프로토타입 모델] : 중요한 일부분만 개발하여 시험 및 평가하여 개선해가는 방식.



### 1_2. 방법론 구성

일반적인 방법론은 계층적 모델로 구성되어 있다.

* [Phase 단계] : 최상위 계층으로 단계별 산출물이 생성됨.

* [Task 태스크] : 단계의 단위 활동, 품질 검토 항목.

* [Step 스탭] : 각 테스크를 구분하는 작업들.

## 2.KDD 분석 방법론
1996년 Fayyad가 개발한 방법론으로 [KDD Kwonledge Discovery in Database 지식 발굴 데이터베이스]이다. 통계적 패턴이나 지식을 찾는 것이 목적으로 다양한 분야에 응용할 수 있다.

* 분석 절차 : 데이터셋 [Selection 선택] -> 노이즈와 이상 값을 정제하는 [Preprocessing 전처리] 작업 -> 목표에 맞는 변수를 찾고 데이터 차원을 축소하는 등의 데이터 [Transformation 변환] -> [Data Mining 데이터 마이닝] 작업 실행, 최적의 결과물 산출 -> 데이터 마이닝 [Evaluation 평가]

## 3. CRISP-DM 분석 방법론
1996년 유럽연합의 주요 5개 업체들이 주도하여 개발한 방식으로 전체 이름은 [CRISP-DM CRoss Insdustry Standard Processing for Data Mining 산업 공통 표준 진행 방법-데이터 마이닝]이다. KDD와 마찬가지로 계층적으로 구성되어 있다. 총 6단계로 구성되어 있으며 각 단계는 단방향이 아닌 여러 방향으로 피드백을 통하여 완성도를 높인다.

* 구성 : 여러 개의 [Phase 단계 ]로 구성되어 있고 각 단계는 [Generic Tasks 일반화 테스크] (단일 프로세스를 완전하게 수행하는 단위)를 포함한다. 각 [일반화 태스크]는 수행 레벨에 따라 [Specialized Tasks 세분화 테스트]로 나뉘고, [세분화 테스크]는 구체적인 실행하는 [Process Instances 프로세스 실행]으로 구성된다.


### 3_1. [Business Understanding 업무 이해]

프로젝트의 목적과 요구사항을 이해하고, 도메인을 문제 정의로 보고 초기 프로젝트의 계획을 수립한다.

업무 목적 파악, 데이터 마이닝 목표 설정, 프로젝트 계획 수립

### 3_2. [Data Understanding 데이터 이해]

데이터 품질에 대한 문제를 식별, 분석을 위한 데이터 수집, 데이터 속성을 이해한다.

초기 데이터 수집, 데이터 기술 분석, 데이터 탐색 및 품질 확인

### 3_3. [Data Preparation 데이터 준비]

수집한 데이터에서 적합한 데이터를 선택 및 정제하여 분석 작업 진행. 많은 시간이 소요.

데이터셋 선택 및 정제, 데이터 통합, 데이터 포맷팅

### 3_4. [Modeling 모델링]

모델링 기법을 선택하여 파라미터를 최적화하며 테스트용과 검증용으로 나눈다.

모델링 기법 선택, 모델 테스트 계획 설계 모델 작성, 모델 평가.

### 3_5. [Evaluation 평가]

모델링 결과가 목적에 부합한 지 평가하여 최종적으로 수용할지 판단한다.

분석 결과 평가, 과정 평가, 적용성 평가.

### 3_6. [Deplyment 전개]

완성된 모델을 실무에 적용하고 유지 보수 계획을 수립한다. 

전개 계획 수립, 모니터링과 유지보수 계획, 프로젝트, 종료 보고서, 프로젝트 리뷰

## 4. 빅데이터 분석 방법론
분석 기획, 데이터 준비, 시스템 구현, 평가 및 전개로 5가지 [단계]로 구성되어 있으며 위의 CRISP-DM 분석 방법론보다 더 상세한 과정을 거친다.

* [Phase 단계] : 각 단계별 산출물이 생성되며 [Baseline 기준선]으로 관리.

* [Task 태스크] : 각 단계별 품질 검토 항목.

* [Step 스탭] : [WBS Work Breakdown Structure 업무 분업 구조, 작업 분해도]의 [Work Package 워크 패키지]에 해당하여 입력자료, 처리 및 도구, 출력자료로 구성된 [Unit Process 단위 프로세스]이다.



### 4_1. [Planing 분석 기획]

비즈니스에 대한 충분한 이해와 도메인에 대한 문제점을 파악하고 프로젝트 범위를 명확화 한다.
비즈니스 이해 및 범위 설정, 프로젝트 정의 및 계획 설정, 프로젝트 위험 계획 수립
[SOW Statemnt of Work 프로젝트 범위 정의서], 프로젝트 정의서. 위험 관리 계획서

### 4_2. [Preparing 데이터 준비]

데이터 속성, 오너, 관련 시스템 담장자 등을 정의한 데이터 정의서를 작성한다. 
필요 데이터 정의, 데이터 스토어, 데이터 정합성 점검
데이터 정의서, 데이터 획득 계획서, 데이터 매핑 정의서, 데이터 정합성 보고서

### 4_3. [Analyzing 데이터 분석]

비즈니스 룰을 확인, 전사적 데이터 스토어에서 데이터를 추출, 구조화된 형태로 편성한다.
분석용 데이터 준비, 텍스트 분석, 탐색적 분석, 모델링, 모델링 평가
분석용 데이터셋, 데이터 스토어, 데이터 시각화 보고서, 훈련 및 평가용 데이터, 모델링 평가 보고서

* 텍스트 분석 : 웹페이지 데이터, 로그, 텍스트 등 비정형 및 반정형 텍스트 데이터를 분석

* 탐색적 분석 : [EDA Exploraroty Data Analysis] 정합성 검토, 데이터 요약, 특성 파악

* 데이터 시각화 : 향후 시스템 구현을 위한 인터페이스나 프로토타입으로 활용 가능

* [Modeling 모델링] : 가설 설정을 통하여 통계 모델을 만들거나 기계학습으로 분류, 예측, 군집 등의 기능을 수행하는 모델을 만드는 과정. 훈련용과 테스트용 데이터로 나누어 [Overfitting 과적합]을 방지하고 [일반화 Generalization]에 이용.

### 4_4. [Developing 시스템 구현]

모델링 결과를 가동 중인 시스템에 적용하기 때문에 운영 시스템에 대한 분석도 같이 수행한다. 시스템 설계서를 바탕으로 [BI Business Intelligence 비즈니스 인텔리전스] 패키지를 활용하거나 새로운 프로그램을 통해 구축한다.
설계 및 구현, 시스템 설계 테스트 및 운영
시스템 분석 및 설계서, 구현 시스템, 시스템 테스트 결과 보고서, 운영자 및 사용자 매뉴얼, 운영 계획서

### 4_5. [Deploying 평가 및 전개]

모델의 [Life of Cycle 생명주기]를 설정하고 주기적으로 유지보수, 재구축 계획을 수립한다. 
모델 발전 계획 수립, 프로젝트 평가 및 보고
모델 발전 보고서, 프로젝트 최종 보고서
