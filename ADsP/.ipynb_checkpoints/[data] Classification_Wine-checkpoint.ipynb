# 분류 분석 사례


Party 패키지를 활용한 분류 분석 실습을 진행한다.
- 패키지 설치가 안 되어 있을 경우 다음 명령어 실행 
- `install.packages("party")`
- `install.packages("HDclassif")`

## 1. Wine Dataset

이 데이터는 이탈리아의 같은 지역에서 재배되었지만 세 개의 다른 재배지에서 파생된 포도주에 대한 화학 분석 결과이다. 이 분석은 세 가지 종류의 와인 각각에 포함된 13명의 성분들의 양을 결정했다.


1. V1 : 알코올
1. V2 : 말산
1. V3 : 재
1. V4 : 화산재
1. V5 : 마그네슘
1. V6 : 총 페놀양
1. V7 : 플라보노이드
1. V8 : 비플라바노이드 페놀
1. V9 : 프로안토시아닌스
1. V10 : 색상 강도
1. V11 : 빛깔
1. V12 : 희석된 와인의 OD280/OD315
1. V13 : 프로라인


```r
library(HDclassif)
data(wine)
head(wine)
```

## 2. Target Variable

숫자형태로 되어 있는 Class 변수를 factor로 변환하고 각 facotor별 개수를 확인한다.
- 1이 59개, 2가 71개, 3이 48개 있음을 확인

```r
wine$class <- factor(wine$class)
table(wine$class)
```

## 3. Data Splite

훈련용과 검증요 데이터로 분리한다. 해당 예제에서는 7:3 비율로 분리한다. 
- **sample()** : 입력변수를 특정 비율로 나누는 기능
- prob 옵션 : 표본 추출되는 벡터의 요소를 얻기 위한 가중치 벡터

```r
set.seed(2020)
ind <- sample(2, nrow(wine), replace=T, prob=c(0.7, 0.3))
```

```r
ind
```

ind에서 값이 1인 것을 훈련용 데이터에, ind값이 0인 것을 검증용 데이터에 대입한다.

```r
tr_wine <- wine[ind==1,]
val_wine <- wine[ind==2,]
```

## 4. Party Packages

Party 패키지를 로드한다. 없을 경우 아래 명령어로 설치한다.
- `install.packages("party")`

```r
library(party)
```

## 5. Classification Model 

각 객체의 leaf를 확인한다. 각 모델의 결과나 각 객체가 분류된 leaf의 위치 등 얻고자 하는 모델의 결과를 확인하는 작업을 진행한다.
- class를 타겟으로 하고 나머지 변수를 투입변수로 하여 모델을 개발.
- 결과값이 엄청나게 길게 나오는데 그래프로 출력하면 편함

```r
md1_party <- ctree(class ~ . , data = tr_wine)
str(md1_party)
```

```r
# tree 객체 확인
md1_party@tree
```

```r
md1_party@where
```

```r
# 검증용 데이터로 검증
md1_party@get_where(newdata = val_wine)
```

```r
# 예측
predict(md1_party, newdata = val_wine)
```

## 6. Visualization

생성된 모델을 시각화하여 의사결정 나무를 확인해본다.


```r
plot(md1_party, terminal_panel=node_barplot)
```

**[결과 분석]**
1. **뿌리 노드**가 V7이 되고, **노드**는 V7, V11, V1, V6, Node 3,4,6,8,9이며 **잎**은 Node 3,4,6,8,9
1. Node 8은 Train_Set에서 V7>1.39, V1>13.03, V6<=2.45인 8개가 속해있다는 의미
1. 그 8개 중에 class가 1인 것이 4개, 2가 3개, 3이 1개
1. 전체 잎 들 중에서 가장 분류가 안된 잎으로 판별됨

```r
# 다른 그래프 양식으로 출력
plot(md1_party, type="simple")
```

## 7. Performance

분류 결과를 Predict를 이용해 확인 가능하며 이를 실제값과 비교하여 매트릭스로 저장하여 **[Accuracy 정확성], [Precision 정밀도], [Detect Rate 탐지율]**을 구할 수 있다.
- [Detect Rate 탐지율] : 타겟변수에 대해서는 **[Sensititivity 민감도]**를, 그 외 변수에 대해서는 **[Specificity 특이도]**라는 용어를 사용해 모델을 평가
- diag : 대각행렬을 추출하는 기능

```r
predict(md1_party)
```

```r
res_tr <- table(predict(md1_party), tr_wine$class)
res_tr/sum(res_tr)
```

```r
res_tr
```

```r
sprintf("정확도 : %f", sum(diag(res_tr)) / sum(res_tr))

sprintf("정밀도_1행 : %f", res_tr[1,1]/sum(res_tr[1,]))
sprintf("정밀도_2행 : %f", res_tr[2,2]/sum(res_tr[2,]))
sprintf("정밀도_3행 : %f", res_tr[3,3]/sum(res_tr[3,]))

sprintf("탐지율_1열 : %f", res_tr[1,1]/sum(res_tr[,1]))
sprintf("탐지율_2열 : %f", res_tr[2,2]/sum(res_tr[,2]))
sprintf("탐지율_3열 : %f", res_tr[3,3]/sum(res_tr[,3]))
```

## 8. Performance : Validation Set

이번에는 동일한 과정을 검증용 데이터로 테스트 해본다.

```r
res_val <- table(predict(md1_party, newdata=val_wine), val_wine$class)
res_val/sum(res_val)
```
