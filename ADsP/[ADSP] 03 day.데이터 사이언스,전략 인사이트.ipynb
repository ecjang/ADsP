<h1><center>1절. 빅데이터 분석과 전략 인사이트</center></h1>

## 1. 빅데이터 열풍과 회의론
빅데이터의 인기과 발전과 다르게 회의적으로 바라보는 시각도 있다. 이런 회의론들은 빅데이터의 가치를 제대로 보기도 전에 활용 자체를 사전에 차단해 버릴 수 있으므로 빅데이터에 대한 올바른 이해가 필요하다.

## 2. 빅데이터 회의론의 원인 및 진단
빅데이터는 기존의 데이터에서 가치, 즉 통찰을 이끌어내 성과를 창출하는 것이 관건이며 단순히 많은 데이터를 수집하는 것이 목적이 아니다. 분석을 통해 가치를 만드는 것에 집중해야 한다.

* 과거 CRM의 부정적인 학습효과 : 초기 CRM에 빅데이터 분석이 도입만 하면 무조건 매출이 증가할 것이라는 공포 마케팅식으로 홍보가 되었다. 그래서 거액을 투자하여 분석 설비와 설루션을 도입했지만 어떻게 가치를 추출할지 제대로 알지 못하였다.

* 굳이 빅데이터 분석이 필요 없는 경우 : 우수고객, 이탈 고객, 구매패턴 분석 등.

* 국내 빅데이터 업체들인 CRM 분석 성과를 빅데이터 분석으로 과대 포장 및 홍보.

## 3. 싸이월드
싸이월드는 2004년 세계 최대의 소셜 네트워크 서비스이다. 엄청난 인기와 데이터를 보유했지만 페이스북처럼 성장하지 못했다. 


### 3_1. 싸이월드의 퇴보 원인
```
데이터 분석 환경이 존재했지만 의사결정에 활용되지 않음.
분석이 웹로그 분석 같은 1차원적인 분석이며 비즈니스 핵심 가치 분석이 이루어지지 않음.
SNS 서비스였지만 회원들의 활동 분석을 위한 분석 환경이 미비.
트렌드 변화가 사업 모델에 미치는 영향을 적시에 파악하지 못함.
```

### 3_2. 하라스 엔터테인먼트
러브먼은 하라스 엔터테인먼트의 회장이다. 하라스 엔터테인먼트는 카지노, 호텔, 레스토랑 등을 포함하는 종합 엔터테인먼트 기업으로 만년 3위였지만 데이터 분석경영을 통해 업계 1위인 시저스를 인수하여 독보적인 1위로 성장하였다.

러브먼의 분석 기반 경영 도입이 힘든 이유

1. 기존의 관행을 유지.
2. 직관적인 결정을 우대.
3. 분석경영의 전문가 부재.
4. 아이디어를 낸 사람한테 집중하는 경향.


## 4. Big이 핵심이 아니다
시간이 지날수록 빅데이터에 대한 관심이 증가하고 있으며 데이터 기반의 분석과 통찰이 중요하다는 인식을 키웠다. 그러면서 자연히 빅데이터에 대한 고민을 하게 되었다.

* 기대 : 자동화가 이루어지고 점차적으로 전략적 가치를 이끌어 낼 수 있다.

* 조슈아 보거 박사 : 직관적 의사결정보다 데이터에 의한 의사결정이 중요해진다.

* 쟁점 : 크기가 아니라 어떤 통찰을 얻을 수 있는 데이터를 수집하는 것이 중요.

* 가치 : 전략과 비즈니스 핵심 가치에 집중하고, 시장과 고객 변화에 대응할 수 있어야 한다.

## 5. 전략적 통찰이 없는 분석의 함점
경영 방식이 제대로 반영되지 못한 분석은 의미가 없으며 경쟁사와 차별화할 수 없다. 전략적 통찰력을 가지고 분석하고 핵심적인 비즈니스 이슈에 집중하고 차별적인 전략을 위한 분석이 이루어져야 한다.

### 5_1. 아메리칸 항공

데이터 분석을 통해 수익관리와 가격 최적화를 실시하여 3년 만에 14억 달러 수익을 창출하였지만, 시간이 지나 타 경쟁사들도 비숫한 분석 능력을 갖춤으로써 경쟁우위가 사라짐.

### 5_2. 사우스웨스트 항공

보잉 737 한 가지 기종, 최적화 모델, 좌석 가격 등을 분석하여 엄청난 수익을 창출. 



## 6. 일차원적인 분석 
성과관리, 신용점수 산정, 재고 관리, 가격 책정 등 일차원적인 분석으로 상당한 효과를 얻을 수 있다. 하지만 큰 환경변화에 제대로 대응하기 어렵고, 급변하는 트렌드에 맞추어 전술적으로 사용하기 힘들다.

* 일차원적인 분석으로 분석 경험을 쌓고 작은 성과를 이루면서 점점 전략적으로 발전해야 한다.

* 차별화를 위해선 전략적 통찰을 추출할 수 있는 가치기반 분석 단계로 나아가야 한다.

<h1><center>2절. 전략 인사이트 도출을 위한 필요역량</center></h1>

## 1. 데이터 사이언스의 의미와 역할
데이터 사이언스는 데이터로부터 의미 있는 정보를 추출하는 학문으로써 데이터 공학, 수학, 통계학, 컴퓨터 공학, 시각화, 해커의 사고방식, 해당 분야의 전문 지식 등이 필요하다. 그리고 비즈니스의 핵심 이슈에 대한 답을 제시하고 사업의 성과를 이끌어 나갈 수 있어야 하므로 소통력이 필요하다.



## 2. 데이터 사이언스의 구성요소
데이터 사이언스가 갖추어야 할 역량 중 하나는 바로 강력한 호기심이다. 문제 해결의 핵심적인 질문을 찾고 검증 가능한 가성을 세우는 능력을 의미한다. 

* 분석적 영역 : 수학, 확률 모델, 분석학, 패턴인식 등

* 데이터 처리 : 시그널 프로세싱, 데이터 엔지니어링, 데이터 웨어하우스 등

* 비즈니스 컨설팅 : 프레젠테이션, 스토리텔링

### 2_1. 데이터 사이언스의 요구 역량
크게 **[Hard Skill 하드 스킬]**과 **[Soft Skill 소프트 스킬]**로 나뉜다. 분석 기술보다 중요한 것이 소프트 스킬이 더 중요하다. 전략적인 통찰을 위한 분석을 위해서는 분석 능력이나 지식 외에도 스토리텔링, 창의력, 직관력, 비판적 시각이 필요하다. 

* Hard Skill : 빅데이터 지식, 분석 기술

* Soft SKill : 통찰력 있는 분석, 설득력 있는 전달, 다분야 커뮤니케이션

## 3. 인문학의 부활
기존의 틀을 벗어나 문제를 바라보고 해결하는 능력, 문제 핵심 파악, 고객의 요구사항 이해 등 인문학에서 배울 수 있는 능력이 필요하다.

* [Convergence 컨버전스] 단순 세계에서 [Devergence 디버전스] 복잡한 세계로 변화하면서 다양성, 연결성, 창조성이 중요 키워드가 됨.

* 비즈니스의 중심이 제품 생산에서 서비스로 이동, 뛰어난 서비스를 제공하는 것이 관건.

* 경제와 산업의 논리가 생산에서 시장 창조로 바뀜. 공급자 중심에서 무형 자산의 경쟁으로 바뀌어 현지 사회와 문와에 대한 지식이 필요.

## 4. 인문학적 사고의 특성화 역할
데이터 사이언티스트를 위한 6가지 핵심 질문이 있다. 과거, 현재, 미래 시점으로 분류하여 정리할 수 있다.

* 정보 :  리포팅(과거) -> 경고(현재) -> 추측(미래)
```
예: 과거 무슨 일이 일어났는가? -> 현재 무슨 일이 일어나고 있는가? -> 미래 무슨 일이 일어날 것인가?
```
* 통찰력 : 모델링(과거) -> 권고(현재) -> 예측, 최적화(미래)
```
예: 과거 어떻게 일어났나? -> 현재 차선 행동은 무엇인가? -> 미래 최악이나 최선의 상황은 무엇인가?
```

<h1><center>3절. 빅데이터 그리고 데이터 사이언스의 미래</center></h1>

## 1. 가치 패러다임의 변화
데이터 사이언스는 빅데이터를 빠르고 정확하게 처리하여 하이퍼 연결을 효과적으로 관리하는 역할을 한다.

* [Digitalization 디지털화] : 아날로그 데이터를 디지털화하여 가치 창출.

* [Connection 연결] : 디지털화된 대상들을 얼마나 효과적, 효율적으로 연결하느냐 가 중요.

* [Agency 에이전시] : 복잡한 연결을 잘 관리하여 효과적으로 제공하도록 관리.
