<h1><center>3절. 분석 과제 발굴</center></h1>

분석 과제 발굴을 위한 방법은 크게 [Top Down Approach 하향식 접근]과 [Bottom Up Approach 상향식 접근] 2가지가 있다.

* [하향식 접근] : 문제가 주어지고 해법을 찾는 방법

* [상향식 접근] : 문제 정의가 어려운 경우 문제 정의를 위한 방법

* 5가지 고려 사항
    - [Size 크기]
    - [Complexity 복잡성]
    - [Speed 속도]
    - [Analysis Complexity 분석 복잡성]
    - [Accurancy, Prediction 정확도 및 정밀도]




## 1. 하향식 접근 방법 : 최적화에서 솔루션로
* 4단계 과정 : [Discovery 탐색] -> [Definition 정의] -> [Search 탐색] -> [Feasibility 타당성 평가]



### 1_1. [Discovery 문제 탐색]

비즈니스 기반 문제 탐색, 외부 사례 기반 문제 탐색 2가지가 있다.

* [비즈니스 모델 캔버스]  : 비즈니스를 업무, 제품, 고객, 규제/감사, 지원 인프라 5가지 분야로 간략화

    1. [Operation 업무] : 제품 및 서비스를 생산하기 위해 운영하는 내부 프로세스 및 자원 관련 주제 도출

    1. [Product 제품] : 생산 및 제공하는 제품과 서비스를 개선하기 위한 주제 도출

    1. [Customer 고객] : 제품 및 서비스를 제공받는 고객, 이를 제공하는 채널 관점에서 주제 도출

    1. [Regulation 규제] : 제품 생산 및 전달하는 과정 중 발생하는 규제 및 보안 관점에서 주제 도출

    1. [Resource 지원 인프라] : 분석 시스템 영역 및 운영 관리하는 인력 관점에서 주제 도출


* [거시적 관점의 메가 트렌트] : STEEP. 사회, 기술, 경제, 환경, 정치 영역의 주요 사항을 고려

* [경쟁자 확대 관점] : SCN. 대체제, 경쟁자, 신규 진입자 관점으로 확대하여 분석

* [시장 니즈 탐색 관점] : CCI, 고객 채널, 영향자들에 대한 폭넓은 관점으로 분석

* [역량 재해석 관점] : CP. 내부 역량, 파트너와 네트워크 영역으로 활용 가능한 역량을 토대로 분석

외부 참조 모델 기반 분석은 유사한 사례 벤치마킹을 통해 분석 기회를 발굴하는 것을 의미한다. 분석 테마 호보 [Pool 풀]을 통해 [Quick & Easy] 방식으로 아이디어를 얻고, [Brain Stomming 브레인스토밍] 방식으로 빨게 도출한다. 분석 기회들을 구체적인 과제로 만들기 위해 [분석 유즈 케이스]로 작성해두는 것이 필요하다.



### 1_2. [Definiation 문제 정의]

식별된 비즈니스 문제를 [Why 왜]에 해당하는 데이터의 분석 문제로 전환하는 단계이다. 최종 사용자 관점에서 이루어져야 하며, 용이한 데이터  정의 및 기법 선택을 위해 명확히 문제를 정의해야 한다.

예: 고객 이탈 증대 -> 고객 이탈에 영향을 미치는 요인을 식별하고 이탈 가능성을 예측

예: 예상치 못한 설비 장애로 판매량 감소 -> 설비의 장애 신호를 감지하여, 장애 요인을 식별하고 장애 발생 시점 및 가능성을 예측



### 1_3. [Solution Search 솔루션 탐색]

정의된 데이터 분제를 해결하기 위한 방안을 찾는다. 분석 역량을 보유하지 않을 경우 교육이나 전문 인력 채용으로 역량을 확보, 혹은 전문 업체에 의뢰하여 해결할 수 있다.



### 1_4. [Feasibility Study 타당성 검토]

* 경제적 타당성 : 비용 대비 편익 분석 관점 접근. 비용 절감, 추가적 매출, 수익 등 경제적 가치로 산출.

* 데이터 및 기술적 타당성 : 분석 역량 타당성 평가를 위해 비즈니스 지식과 기술적 지식이 요구.





## 2. 상향식 접근 방법 : 발견에서 통찰로
다양한 원천 데이터를 대상으로 분석하여 가치 있는 모든 문제를 도출하는 과정이다.

예: 의약품 제조업체는 특허가 만료되는 2천여 건의 의약품 데이터를 통해 효과적인 조합 방법을 찾아 다양한 방법으로 데이터를 분석하였다.



### 2_1. 기존 하향식 접근법의 한계

주어진 문제 해결에는 적합하지만 새로운 문제 탐색에는 문제점이 발생하여 이를 해결하고자 [Design Think 디자인 사고] 과정이 새로 도입되었고, 데이터를 보다 객관적으로 이해하려고 노력했다.

* 통상적인 관점에서 사물을 인식하는 Why를 강조하지만, 이는 이미 문제를 알고 있다는 가정이 전재

* 답을 미리 내는 것이 아닌 사물을 있는 그대로 관찰하고 인식하려는 What에 대한 관점을 강조

* 그래서 d.school에서는 첫 단계로 [Emphathize 감정이입]을 강조

* 일반적으로 상향식 접근법은 [Unsupervised Learning 비지도 학습] 밥법(장바구니 분석, 군집분석, 기술 통계 및 프로파일링)으로 수행. O, X로 구분 짓는 것은 [Classification 분류]에 해당하고 [Supervised Learning 지도 학습]인반면, 유관성과 유사성으로 구분하는 것은 [Clustering 군집]이다



 

### 2_2. 시행착오를 통한 문제 해결 

[Prototyping 프로토타이핑] 방법론은 일단 분석을 해보고 그 결과를 확인해가면서 반복적으로 개선해 나가는 방법이다.

* 문제에 대한 인식 : 불분명한 문제나 새로운 유형의 문제일 경우 프로토타이핑을 이용해 문제를 이해

* 필요 데이터 존재 여부의 불확실성 : 문제 해결을 위한 데이터 수집, 대체 여부 등 리스크를 사전에 방지

* 데이터 사용 목적의 가변성 : 보유한 데이터라도 재검토하여 사용 목적과 범위를 확대 가능

* 예: 이동통신사의 사용자 위치 정보는 단순한 위치정보 수집 용도지만 현재 다양한 분야에 사용





3. 분석 과제 정의
분석 과제 정의서 : 소스 데이터, 분석 방법 데이터 입수 및 분석 난이도, 수행 주기, 상세 분석 등을 정의한다. 






## 3. 분석 과제 정의
분석 과제 정의서 : 소스 데이터, 분석 방법 데이터 입수 및 분석 난이도, 수행 주기, 상세 분석 등을 정의한다.

<h1><center>4절. 분석 프로젝트 관리 방안</center></h1>

분석 기획은 프로젝트를 통해 가치가 증명돼야 한다. 다른 프로젝트와 마찬가지로 범위, 일정, 리스크, 의사소통 등 영역별로 관리해야 하며 5가지 주요 속성을 고려한 추가적인 관리가 필요하다.

* 5가지 주요 특성 : 크기, 데이터 복잡도, 속도, 분석 복잡도, 정밀도 및 정확도

* [Accurancy 정확도] : 모델과 실제값의 차이, [Precision 정밀도] : 반복 시 편차 수준

* 분석 활용면에서는 정확도가 중요하며, 안전성 측면에서는 정밀도가 중요





## 1. 분석 프로젝트 특성
분석가는 원천 데이터를 분석하고 최종적으로 결과를 비즈니스에 적용하기 때문에 프로젝트를 어떻게 관리할지 중요 포인트를 이해하는 것이 중요하다.

* [Agile 어자일] 프로젝트 관리 방식 : 대부분 반복적인 분석 및 교정이 수행되므로 해당 관리 방법이 필요. 지속적인 반복과 개선을 통해 목표에 부합하는 프로젝트로 관리







## 2. 분석 프로젝트 관리 방안
프로젝트 관리 지침의 관리 체계 : 통합, 이해 관리자, 범위, 자원, 시간, 원가, 리스크, 품질, 조달, 의사소통
