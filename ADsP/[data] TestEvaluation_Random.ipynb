# 회귀 모델 적합성 검증

참고 : 데이터 분석 준전문가 한 권으로 끝내기

Summary 함수를 통해 결정계수, F통계량, 잔차의 표준오차 등 주요 통계량의 정보를 출력할 수 있다.

## 1. Regression Analysis

```r
# 난수 생성기의 시드 설정. 
set.seed(2) 
```

```r
u <- runif(10, 0, 11)
v <- runif(10, 11, 20)
w <- runif(10, 1, 30)

u;v;w
```

```r
y = 3 + 0.1*u + 2*v -3*rnorm(10, 0, 0.1)
y
```

- **runif(10, 0, 11)** : 평균이 0, 표준편차가 11인 연속균등분포로 부터 난수 10개를 추출하는 기능
- **rnorm(10,0,0.2)** : 평균이 0, 표준편차가 2인 정규분포로부터 난수 10개를 추출

## 2. Covert to DataFrame

```r
dfrm <- data.frame(u,v,w,y)
dfrm
```

## 3. Regression Analysis

```r
lm(y~u+v+w, data=dfrm)
```

- **Intercept** : 상수


그러므로 회귀식은 아래와 같다. 
$$ y = 2.6011 + 0.1002\times{u} + 2.0276\times{v} + 0.0004\times{w} $$
