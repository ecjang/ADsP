## 1. 트라우마를 부정하라

1. 알려지지 않은 '심리학 제3의 거장'
1. 인간이 변할 수 있는 이유는 무엇인가
1. 트라우마란 존재하지 않는다
1. 인간은 분로를 지어낸다
1. 과거에 지배받지 않는 삶
1. 소크라테스, 그리고 아들러
1. 당신은'이대로' 좋습니까?
1. 나의 불행은 스스로 '선택'한 것
1. 인간은 끊임없이 '변하지 않겠다'고 결심한다
1. 나의 인생은 '지금 여기'에서 결정된다.

## 2. 모든 고민은 인간관계에서 비롯된다

1. 왜 자기 자신을 싫어하는가
1. 모든 고민은 '인간관계에서 비롯된 고민'이다
1. 열등감은 주관적인 감정이다
1. 변명으로서의 열등 콤플렉스
1. 자랑하는 사람은 열등감을 느끼는 사람
1. 인생은 타인과의 경쟁이 아니다
1. 내 얼굴을 주의 깊게 보는 사람은 나뿐이다
1. 권력투쟁에서 복수로
1. 잘못을 인정하는 것은 패배가 아니다
1. '인생의 과제'를 어떻게 극복할 것인가
1. 붉은 실과 단단한 쇠사슬
1. '인생의 거짓말'을 외면하지 말라
1. '소유의 심리학'에서 '사용의 심리학'으로

## 3. 타인의 과제를 버려라

1. 인정욕구를 부정하라
1. '그 사람'의 기대를 만족시키지 위해 살지 말라
1. '과제를 분리'하라
1. 타인의 과제를 버려라
1. 인간관계의 고민을 단숨에 해결하는 방법
1. '고르디우스의 매듭'을 끊어라
1. 인정욕구는 부자유를 강조한다
1. 진정한 자유는 무엇인가

## 4. 세계의 중심은 어디에 있는가

1. 개인 심리학과 전체론
1. 인간관계의 목표는 공동체 감각을 향한 것
1. 왜 나이외에는 관심을 두지 않는가
1. 나는 이 세계의 중심이 아니다
1. 더 큰 공동체의 목소리를 들어라
1. 칭찬도 하지 말고, 야단도 치지 말라
1. '용기 부여'하는 과정
1. 스스로 가치있는 사람이라고 생각하려면
1. 여기에 존재하는 것만으로 가치가 있다
1. 인간은 '나'를 구분할 수 없다

## 5. 지금 여기를 진지하게 살아간다

1. 과도한 자의식이 브레이크를 건다
1. 자기 긍정인 아닌 자기 신뢰를 하라
1. 신용과 신뢰는 어떻게 다른가
1. 일의 본질은 타인에게 공유하는 것
1. 젊은 사람은 어른보다 앞서간다
1. 일이 전부라는 인생의 거짓말
1. 인간은 지금 이 순간부터 행복해질 수 있다
1. '특별한 사람'이 되고 싶은 사람 앞에 놓인 두 갈래 길
1. 평범해질 용기
1. 인생이란 찰나의 연속이다
1. 춤을 추듯이 살라
1. 지금 여기에 강렬한 스포트라이트를 비춰라
1. 인생 최대의 거짓말
1. 무의미한 인생에 인생을 부여하라
