# 2. 모든 고민은 인간관계에서 비롯된다.

아무리 어려워 보이는 관계일지라도 마주하는 것을 회피하고 뒤로 미뤄서는 안돼. 가장 해서는 안되는 것이 이 상황, '이대로' 멈춰 서 있는 것이라네.

----

지금 내가 할 수 있는 일이란 일단 **'지금의 나'를 받아들이고, 결과가 어떻게 나오든 간에 앞으로 나아갈 용기**를 갖게 하는 것이라네. 이러한 접근 방식을 **'용기 부여'**라고 하지. p.78

자네를 둘러싼 타인, 사회, 공동체가 있고 이러한 것들에게서 소외받고 있다고 느끼기 때문에 고독한 거지. **우리는 고독을 느끼는 데에도 타인을 필요로 한다네.** 즉, 인간은 사회라는 맥락 속에서 비로소 '개인'이 되는 걸세. p.81

분명히 155cm라는 키는 객관적으로 측정된 숫자라서 언뜻보면 열등하게 느껴지지. 하지만 문제는 그 키에 내가 어떤 의미를 부여하느냐, 어떤 가치를 주느냐 하는 점이지. 우리는 괴롭히는 **열등감은 '객관적 사실'이 아니라 '주관적 해석'이네.** p.88

다이아몬드, 혹은 돈. 우리는 여기에 어떤 가치를 매기고 1캐럿에 얼마라거나 물가가 얼마라거나 가치를 매기네. 하지만 관점을 바꾸면 다아이몬드는 한낱 돌덩이에 불과하지 p.90

우선 인간은 무기력한 상태로 태어났네. 그리고 무기력한 상태에서 벗어나려는 보편적인 욕구를 가지고 있지. 아들러는 이것을 **'우월성 추구'**라고 했네. 간단히 말하면 '향상되기를 바라는 것', '이상적인 상태를 추구하는 것'이지. p.91

주의하게. 현재 우리 사회에는 콤플렉스와 열등감이 같은 말처럼 사용되고 있지. 콤플렉스란 복잡하게 얽힌 도착적인 심리상태를 의미하는 것이며 열등감과 관계 없네. 열등감 자체는 나쁜 것이 아니지만 **열등 콤플렉스는 열등감을 변명거리로 삼기 시작한 상태**를 의미하네. p.94

열등감 자체를 첨예화시켜 특이한 우울감에 빠지는 패턴이라네. 구체적으로는 '불행 자랑'이라고 하지. 상처를 입었을 때 적지 않은 사람들이 이런 태도를 취하며 '특별한 존재'가 되려고 하네. p.101,102

아들러가 "오늘날 연약함은 매우 강한 권력을 지닌다."고 했을 정도야. p.103

건전한 열등감이란 타인과 비교해서 생기는 것이 아니라 이상적인 나와 비교해서 생기는 것이라네. p.105

자네에게 타인은 어떤 존재일까? 단순한 라이벌이 아닐세. 언제부터인가 자네를 제외한 모든 사람을, 더 나아가서는 세계를 '적'으로 느끼게 된다네. p.110

참는다는 것은 아직 권력투쟁에 사로잡혀 있다는 증거일세. 상대가 싸움을 걸어오면, 그것이 권력투쟁임을 알아차렸더라면 서둘러 싸움에서 물러나게. 상대의 도발에 넘어가지 말게. 우리가 할 수 있는 것은 그뿐이네. p.121

왜 자네는 타인을 '적'으로 보고 '친구'로 보지 못하는 것일까? 아들러 심리학은 인간의 행동과 심리, 양 측면에서 아주 분명한 목표를 제시했지. **행동의 목표로는 '자립할 것', '사회와 조화를 이루며 살아갈 것'이라는 두가지를, 심리적 목표로는 '내게는 능력이 있다'는 의식과, '사람들을 내 친구로 만든다'라는 의식**을 가질것을 제시했네.
