# 4. 세계의 중심은 어디에 있는가

자네도 나도 세계의 중심이 아니야. 내 발로 인간관계의 과제에 다가가지 않으면 안 되네. '내가 이 사람에게 무엇을 줄 수 있을까?'를 생각해야지.

----

개인 심리학은 영어로 Individual Psychology라고 하네. 그리고 이 개인을 뜻하는 Individual의 어원을 살펴보면 '분할할 수 없다'는 의미가 담겨 있네. 더이상 나눌 수 없다는 의미지. 그렇다면 무엇을 나눌 수 없는 걸까? 아들러는 신체와 정신, 이성과 감정, 이식과 무이식 등 이원론적 가치관에 반대했네.  p.201

의식과 무의식이 있는 것은 사실이네. 하지만 예를 들어 화가 나서 타인에게 소리를 지른다면, 그것은 '전체로서의 나'가 소리지르는 것을 선택한 것일세. 결코 감정이라는 독립된 개체가 내 의향과 관계없이 악을 썻다고는 생각할 수 없네. 여기서 '나'와 '감정'을 따로 떼어놓고 '감정이 나를 그렇게 만들었다.' 또는 '감정에 치우쳤다'라고 생각하면 쉽게 인생의 거짓말에 빠지게 되지. p.203

친구들과 같은 공간에 있다면 그곳에서 '있을 곳'을 찾게 돼. 나아가서 친구들(공동체)을 위해 공헌하는 것도 고려하게 되겠지. 이렇게 **타인을 친구로 여기고 거기서 내가 있을 곳은 여기라고 느낄 수 있는 것이 '공동체 감각'일세.** p.206

내 인생의 주인공은 '나'라는 것은 이상할 것이 없어. 하지만 '나'는 이 세상의 중심이 아니지. 나는 인생의 주인공이자 어디까지나 공동체의 일원이자 전체의 일부란 말이야. 자기밖에 관심이 없는 사람은 본인이 세계의 중심에 있다고 생각하지. 이런 사람은 타인이란 '나를 위해 뭔가를 해줄 사람'에 불과해. 스스로 '인생의 주인공'을 넘어 '세계의 주인공'이라고 믿네. p.212,213

**소속감이란 태어나면서부터 주어지는 것이 아니라, 스스로 획득하는 것일세.** p.216

칭찬의 배후에는 상하관계, 즉 수직관계를 보여주는 것지. 인간이 남을 칭찬하는 것은 '자기보다 능력이 뒤떨어지는 상대를 조종하기 위한 것'이라네. 거기에는 감사의 마음도 존경의 마음도 없지 p.226

강제하지도 않고, 과제를 분리한 상태에서 자력으로 해결할 수 있도록 지원하는 거야. '말을 물가에 대려가게 할 수 있지만 물을 마시게 할 수 없다'라는 말과 딱 들어맞는 일이지. **칭찬하지도 않고 야단치지도 않네. 이러한 수평관계에 근거한 지원을 아들러 심리학에서는 '용기 부여'라고 부르지.** p.231

그때 아들러의 대답은 이러했네. "누군가 시작하지 않으면 안됩니다. 다른 사람이 협력하지 않더라도 그것은 당신과 관계없습니다. 내 조언은 이래요. **당신부터 시작하세요. 다른 사람이 협력하든 안 하든 상관하지 말고**". p.243

연장자를 공경하는 것, 회사에서 직책에 따른 구분 있는 것은 당연해. 누구와도 친하게 지내라, 허물없이 행동하라는 것이 아닐세. **의식상에서 대등할 것, 그리고 주장할 것은 당당하게 주장하는 것이 중요하다는 말이지** p.247
