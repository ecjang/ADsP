# 5. '지금, 여기'를 진지하게 살아간다

우리 인생도 길잡이 별이 필요한네. 그 별은 이방향으로 쭉 가다 보면 행복이 기다리고 있을 거라는 믿을 주는 절대적인 이상향이라네.

----

결국 공동체 감각이 필요라지. 구체적으로는 '자기에 대한 집착(Self Interest)'을 **타인에 대한 관심(Social Interest)**으로 돌리고 공동체 감각을 기르는 것. 이에 필요한 것이 **'자기수용'과 '타자신뢰', '타자공헌'**이라네. p.259

'자기긍정'이란 내가 하지도 못할 일에 할 수 있다고 스스로 주문을 거는 것일세. 이는 자신한테 거짓말을 하는 것이고 우월 콤플렉스에 빠질 수 있지. 한편 '자기수용'은 '하지 못하는 나'를 그대로 받아들이고 할 수 있는 데까지 앞으로 나아가는 것일세. p.260

신이여, 바라옵건데 제가 바꾸지 못하는 일을 받아들이는 차분함과 바꿀 수 있는 용기와 그 차이를 구분하는 지혜를 주옵소서. 제5 도살장이라는 소설이었어요. p.261,262

'당신이 빛을 갚는다는 전제하에 빌려주는 거에요.', '당신이 갚을 수 있을 정도만 빌려줄께요'라는 식으로. 이건 신뢰가 아니라 신용이지. p.263

'배신당한 상황'에만 주목하고 있어. 그럴 때 받을 상처에만 주목하고 있어. 그런데 신뢰하는 것을 두려워하면 결국은 누구와도 싶은 관계를 맺을 수 없다네. p.268

타자공헌이 의미하는 것은 자기희생이 아니라네. 오히려 아들러는 타인을 위해 자기 인생을 희생하는 사람을 보고 '사회에 지나치게 적응한 사람'이라며 경종을 울리기도 했지. 즉 나를 버리고 타인에게 최선을 다하는 것이 아니라, 오히려 '나'의 가치를 실감하기 위한 행위이지. p.272

우선 '자기 자신을 수용'한다. 그러면 두려움 없이 '타인을 신뢰'할 수 있다. 타인을 무조건 신뢰하고 친구라고 여기면 '공동체를 위해 공헌'할 수 있다. 그리고 그 공헌으로 인해 '나는 누군가한테 도움이 되고 있다'라고 실삼하게 되면 자신을 받아들일 수 있다. p.276

신경질적인 사람은 걸핏하면 모두, 늘, 전부라는 말을 입에 담는다네. "모두 나를 싫어해", "늘 나만 손해봐", "전부 틀렸어"라는 식으로. 만약 자네가 이런 성급하게 일반화시키는 말을 입에 담는다면 주의해야 하네. p.280

공헌감을 얻기 위한 수단이 '남들로부터 인정 받는 것'이라면 결국 남의 의도대로 밖에 살 수 없지. **인정욕구를 통해 얻은 공헌감은 자유가 없지. 우리는 자유를 선택하면서 동시에 행복을 추구하는 존재하네.** p.290

평범함을 거부하는 것은 아마 평범하다는 것을 무능하다고 착각하기 때문이겠지. **평범한 것은 무능한 것이 아니라네.** 일부러 자신의 우월성을 과시할 필요가 없는 것 뿐이야. p.297

목적지에 도달하려는 인생은 **'키네시스(Kinesis)적 인생'**이라고 할 수 있네. 그에 반래 춤을 추는 인생은 **'에네르게이아(Energeia)적  인생'**이라고 할 수 있을 걸세. 에네르게이아란 지금 하고 있는 것이 그대로 이루어진 상태가 된 운동을 가리키네. 달리 말하면 과정 자체를 결과로 보는 운동이야. p.304, 305

**천재지변을 당했을 때 원인론에 입각해서 "왜 이런 일이 일어난 걸까?"라고 과거를 돌아보며 따져봤자 무슨 의미가 있겠나?** 우리는 곤경에 처했을 때야말로 앞으로 보며 "이제부터 무엇을 할 수 있을까"를 생각해야 하네. p.315

자네가 어떠한 찰나를 보내더라도, 설령 자네를 싫어하는 사람이 있다고 하더라도 '타인에게 공헌한다'라는 길잡이 별만 놓치지 않는다면 헤맬 일도 없고 뭘 해도 상관없어. 나를 싫어하는 사람에게 미움받으며 자유롭게 살면 되네 p.318
