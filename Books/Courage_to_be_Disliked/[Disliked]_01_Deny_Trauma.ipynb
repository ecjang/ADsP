# 1. 트라우마를 부정하라

자네가 불행한 것은 과거의 환경 탓이 아니네. 그렇다고 능력이 부족해서도 아니고, 자네에게는 그저 '용기'가 부족한 것뿐이야.

----

오스트리아 출신의 정신과 의사 **[Alfred Adler 알프레드 아들러]**가 20세기 초엽에 창설한 심리학으로 그의 이름을 따서 '아들러 심리학'이라고 부르지. p.27

과거의 원인에 주목해서 사건을 설명하려고 하면 모든 이야기는 **결정론**에 도달하게 되네. 즉 현재와 미래는 과거에 의해 결정되고 진다고 말이지. p.33

그 친구는 '밖으로 나갈 수 없다'라는 목적이 먼저고, 그 목적을 위해 불안감을 조성한 것이라네. 아들러 심리학에서는 이걸 **목적론**이라고 한다네. p.34

아들러는 트라우마 이론을 부정하면서 이렇게 말했네. "어떠한 경험도 그 자체는 성공의 원인도 실패의 원인도 아니다. 우리는 경험을 통해서 받은 충격(=트라우마)으로 고통받는 것이 아니라 경험 안에서 목적을 찾아낸다. **경험에 의해 결정되는 것이 아니라, 경험에 부여한 의미에 따라 자신을 결정하는 것이다**". p.36

간단해. 자네는 화가 나서 큰소리를 친 것이 아니라, 큰소리를 내기 위해 화를 낸 것이지. p.41

**요컨데 분노란 언제든 넣었다 빼서 쓸 수 있는 '도구'라네.** 엄마는 화를 참지 못해서 화를 낸 것이 아니야. 그저 딸을 위압하기 위해, 그렇게 해서 자기 주장을 밀어 붙이기 위해 분노라는 감정을 이용한 것일세. p.43

"중요한 것은 무엇이 주어졌느냐가 아니라 주어진 것을 어떻게 활용하느냐이다." 자네가 다른 누군가가 되고 싶다고 생각하는 것은 '무엇이 주어졌는가'에만 주목하기 때문일세. 그러지 말고 '주어진 것을 어떻게 활용할 것인가'에 주목하게나. p.53

모든 범죄자는 범행을 저지를 만한 '**마땅한 이유**'가 있지. 가령 금전에 대한 문제로 살인을 저질렀다고 하세. 이것도 당사자에게는 '마땅한 이유'이자 '선'의 수행이라네. 물론 도덕적 의미의 선이 라니라 **'자기 자신에게 득이 된다'는 의미에서의 선이지만.** p.56

아니, 자네는 바꾸지 못하는 것이 아니야. **인간은 언제든 어떤 환경이든 변할 수 있어.** 자네가 변하지 않는 것은 스스로 변하지 않겠다고 결심했기 때문이네. p.62

세계와 자신에 대해 **어떤 의미를 부여하느냐에 따라 세계와 관계를 맺는 법, 그리고 행동도 변할 수 밖에 없지.** 여기서 "변할 수 밖에 없다"는 점에 주목하길 바라네. 자네는 '자네'인체로 그저 생활양식을 고르기만 하면 되는 걸세. 잔인할지도 모르지만 간단하지. p.67
