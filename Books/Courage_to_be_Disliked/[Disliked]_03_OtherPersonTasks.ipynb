# 3. 타인의 과제를 버려라

행복해지려면 '미움받을 용기'도 있어야 하네 그런 용기가 생겼을 때, 자네의 인간관계는 한 순간에 달라질 걸세. 

----

아들러는 이런한 상벌교육을 맹렬히 비판했네. 상벌교육으로 생기는 결과는 '칭찬하는 사람이 없으면 행동을 하지 않는다.', '벌주는 사람이 없으면 부적절한 행동을 한다.'등과 같은 잘못된 생활양식일세. 딱 봐도 이상한 예기지. 우리는 타인의 기대를 만족시키지 위해 사는 것이 아닐세. p.153 

유대교에 이런 말이 있어. "내가 나를 위해 내 인생을 살지 않으면 대체 누가 나를 위해 살아준단 말인가?" 자네는 자네만의 인생을 살고 있어. p.154

신이 존재하기 때문에 선한 행동을 한다는 것은 신이 존재하지 않으면 모든 악행은 허용된다는 것과 등을 맞대고 있는 사상이라네. p.156

부모가 자식한테 '공부해'라고 명령하는 것은 타인의 과제에, 비유하자면 흙투성이의 발을 들이미는 행위일세. 그러면 충돌을 피할 수 없게 되지. 우리는 **'이것은 누구의 과제인가'라는 관점에서 자신의 과제와 타인의 과제를 분리할 필요가 있네**. p.160

어디까지가 내 과제이고, 어디서부터가 타인의 과제인지 냉정하게 선을 긋는 걸세. 그리고 **누구도 과제에 개입시키지 말고, 나도 타인의 과제에 개입하지 않는다.*8 이것이 바로 대인관계의 고민을 해결할 수 있는 아들러 심리학의 핵심이지. p.171

자네는 이것을 이해해야 하네. 타인의 기대를 충족시키면서 살면, 그리고 타인에게 내 인생을 맡기면, 자신뿐만 아니라 타인에게도 거짓말을 하게 되지. 그리고 더 나아가 이런 삶을 살게 되겠지. 과제를 분리하는 것은 자기중심적인 사고가 아니야. **타인의 과제에 개입하는 것이야말로 자기중심적인 발상**이지. 부모가 자식한테 공부할 것을 강요하고, 진로와 배우자를 간섭하나는 것이 자기중심적인 게 아니면 뭔가? p.182

아들러 심리학에서는 "모든 고미은 인간관계에서 비롯된 고민이다"라고 주자아지. 즉, 우리는 인간관계에서 해방되길 바라고, 인간관계로부터 자유로워지기를 바라네. 단적으로 말해 **"자유란 타인에게 미움을 받는 것"일세.** 자네는 아마 '조직에서의 해방'이라고 생각했겠지. 하지만 실제로 조직을 뛰쳐나와도 진정한 자유를 얻을 수 없네. **남이 나에게 어떤 평가를 내리든 개의치 않고, 남이 나를 싫어해도 두려워하지 않고, 인정받지 못한다는 대가를 치르지 않는 한 자신의 뜻대로 살 수 없어.** 자유롭게 살 수 없지. p.186, 187
