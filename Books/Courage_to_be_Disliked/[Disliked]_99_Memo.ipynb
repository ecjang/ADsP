## 1. 트라우마를 부정하라

아들러는 트라우마 이론을 부정하면서 이렇게 말했네. "어떠한 경험도 그 자체는 성공의 원인도 실패의 원인도 아니다. 우리는 경험을 통해서 받은 충격(=트라우마)으로 고통받는 것이 아니라 경험 안에서 목적을 찾아낸다. **경험에 의해 결정되는 것이 아니라, 경험에 부여한 의미에 따라 자신을 결정하는 것이다**".

**요컨데 분노란 언제든 넣었다 빼서 쓸 수 있는 '도구'라네.** 

 **인간은 언제든 어떤 환경이든 변할 수 있어.** 자네가 변하지 않는 것은 스스로 변하지 않겠다고 결심했기 때문이네.

## 2. 모든 고민은 인간관계에서 비롯된다.

**우리는 고독을 느끼는 데에도 타인을 필요로 한다네.**

다아이몬드는 한낮 돌덩이에 불과하지.

열등감 자체는 나쁜 것이 아니지만 **열등 콤플렉스는 열등감을 변명거리로 삼기 시작한 상태**를 의미하네.

오늘날 연약함은 매우 강한 권력을 지닌다.

건전한 열등감이란 타인과 비교해서 생기는 것이 아니라 이상적인 나와 비교해서 생기는 것이라네.

왜 자네는 타인을 '적'으로 보고 '친구'로 보지 못하는 것일까? 아들러 심리학은 인간의 행동과 심리, 양 측면에서 아주 분명한 목표를 제시했지. **행동의 목표로는 '자립할 것', '사회와 조화를 이루며 살아갈 것'이라는 두가지를, 심리적 목표로는 '내게는 능력이 있다'는 의식과, '사람들을 내 친구로 만든다'라는 의식**을 가질것을 제시했네.

## 3. 타인의 과제를 버려라

신이 존재하기 때문에 선한 행동을 한다는 것은 신이 존재하지 않으면 모든 악행은 허용된다는 것과 등을 맞대고 있는 사상이라네.

 **타인의 과제에 개입하는 것이야말로 자기중심적인 발상**이지. 부모가 자식한테 공부할 것을 강요하고, 진로와 배우자를 간섭하나는 것이 자기중심적인 게 아니면 뭔가?
 
자네는 아마 '조직에서의 해방'이라고 생각했겠지. 하지만 실제로 조직을 뛰쳐나와도 진정한 자유를 얻을 수 없네. **남이 나에게 어떤 평가를 내리든 게이치 않고, 남이 나를 싫어해도 두려워하지 않고, 인정받지 못한다는 대가를 치르지 않는 한 자신의 뜻대로 살 수 없어.** 자유롭게 살 수 없지

## 4. 세계의 중심은 어디에 있는가

**'나'와 '감정'을 따로 떼어놓고 '감정이 나를 그렇게 만들었다.' 또는 '감정에 치우쳤다'라고 생각하면 쉽게 인생의 거짓말에 빠지게 되지.**

자기밖에 관심이 없는 사람은 본인이 세계의 중심에 있다고 생각하지. 이런 사람은 타인이란 '나를 위해 뭔가를 해줄 사람'에 불과해. 스스로 '인생의 주인공'을 넘어 '세계의 주인공'이라고 믿네.

인간이 남을 칭찬하는 것은 '자기보다 능력이 뒤떨어지는 상대를 조종하기 위한 것'이라네. 거기에는 감사의 마음도 존경의 마음도 없지.

**칭찬하지도 않고 야단치지도 않네. 이러한 수평관계에 근거한 지원을 아들러 심리학에서는 '용기 부여'라고 부르지.**

**당신부터 시작하세요. 다른 사람이 협력하든 안 하든 상관하지 말고**

**의식상에서 대등할 것, 그리고 주장할 것은 당당하게 주장하는 것이 중요하다는 말이지**

## 5. '지금, 여기'를 진지하게 살아간다

공동체 감각을 기르는 것. 이에 필요한 것이 **'자기수용'과 '타자신뢰', '타자공헌'**이라네.

'당신이 갚을 수 있을 정도만 빌려줄께요'라는 식으로. 이건 신뢰가 아니라 신용이지. 

신뢰하는 것을 두려워하면 결국은 누구와도 싶은 관계를 맺을 수 없다네.

즉 나를 버리고 타인에게 최선을 다하는 것이 아니라, **오히려 '나'의 가치를 실감하기 위한 행위이지.**

우선 '자기 자신을 수용'한다. 그러면 두려움 없이 '타인을 신뢰'할 수 있다. 타인을 무조건 신뢰하고 친구라고 여기면 '공동체를 위해 공헌'할 수 있다. 그리고 그 공헌으로 인해 '나는 누군가한테 도움이 되고 있다'라고 실삼하게 되면 자신을 받아들일 수 있다.

**인정욕구를 통해 얻은 공헌감은 자유가 없지. 우리는 자유를 선택하면서 동시에 행복을 추구하는 존재하네.** p.290

평범함을 거부하는 것은 아마 평범하다는 것을 무능하다고 착각하기 때문이겠지. **평범한 것은 무능한 것이 아니라네.** 일부러 자신의 우월성을 과시할 필요가 없는 것 뿐이야.

목적지에 도달하려는 인생은 **'키네시스(Kinesis)적 인생'**이라고 할 수 있네. 그에 반래 춤을 추는 인생은 **'에네르게이아(Energeia)적  인생'**이라고 할 수 있을 걸세.

**천재지변을 당했을 때 원인론에 입각해서 "왜 이런 일이 일어난 걸까?"라고 과거를 돌아보며 따져봤자 무슨 의미가 있겠나?** 우리는 곤경에 처했을 때야말로 앞으로 보며 "이제부터 무엇을 할 수 있을까"를 생각해야 하네.

설령 자네를 싫어하는 사람이 있다고 하더라도 '타인에게 공헌한다'라는 길잡이 별만 놓치지 않는다면 헤맬 일도 없고 뭘 해도 상관없어.
