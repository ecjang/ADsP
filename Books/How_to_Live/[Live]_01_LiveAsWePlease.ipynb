## 1_1.마음 가는 대로 살자

진정한 프로가 되는 것, 이것이 삶의 행복과 인생의 성공을 절반 결정한다. 그런 점에서 행복한 삶을 원한다면 일이 아니라 놀이를 앞자리에 두어야 한다. p.19

인디밴드 크라잉넛은 '진정한 프로'라고 할 수 있다. 그들은 하고 싶은 일을 제멋대로 하면서 돈도 번다. 그래서 자기네들은 행복하다고 침을 튀기며 자랑한다. p.22

어떻게 살 것인가? 크라잉 넛은 자기네 생각을 이야기했다. "좋아한다면 부딪쳐! 까짓거 부딪쳐!" 훌륭한 대답이다. 그들은 자기네가 좋아하는 펑크록 음악을 들고 세상과 부딪쳤 나름 성공했다. 물론 그렇게 산다고 다 좋은 것은 아니며 성공이라고 할 만한 결실을 맺지 못할 수도 있다. 하지만 좋아하는 일이 없거나 있어도 포기한다면 그 인생은 성공했다고 할 수 없다. p.23

나는 크라잉넛 맴버가 나보다 더 행복하게 살고 있다고 생각한다. 펑크록 멤버가 대학 총학생회 간부보다 더 훌륭하지 않다. 펑크록 공연이 민주화투쟁보다 아름답지도 않다. 빈 맥주캔에 다이빙하는 것이 0.7평 교도소에 있는 것보다 숭고할리 없다. 문제는 무슨 일을 했냐는 것이다. 왜, 어떤 생각으로 그 일을 했는지가 중요하다. 크라잉넛 맴버들은 원하는 인생을 스스로 설계했고 옳다고 생각하는 방식으로 살았다. p.27

그들은 좋아하는 놀이를 직업으로 삼았다. 이것만으로도 인생의 절반을 성공했다고 할 수 있다. 나머지 절반은 사랑과 연대라고 나는 믿는다. p.28

## 1_2. 내 인생은 나의 것

청년기의 핵심은 하고 싶은 일을 찾고 그 일을 잘 할 수 있는 준비를 하는 것이다. p.29

기숙사 룸메이트들은 모두 미팅에 가버렸다. 나는 홀로 남아 [싯다르타]를 읽었다. 삶의 의미를 알고 싶어서 읽었지만 그리 큰 도움이 되지 않았다. 내 문제는, 꼭 하고 싶은 일이 없다는 것이다. 그걸 하지 못하면 삶이 깜깜해질 거 같은, 그렇게 간절하게 하고 싶은 일이 없었다. p.30

환갑의 아버지는 이상주의를 추구했지만 고작 열아홉살이던 나는 현실주의를 추구했다. 법학과에 가서 사법시험을 볼 작정이었다. 그때 아버지의 말씀을 진자하게 받아들였더라면 더 기쁜 삶을 살 수 있었을지도 모르겠다. 아버지가 얼마나 깊게 사랑하셨는지 그때는 미처 몰랐다. p.32

평범한 삶이 아름답고 행복하지 않다는 게 아니다. 평범해도 평범하지 않아도 인생은 훌륭하거나 비천할 수 있다. 인생의 품격은 평범함과 비범함과는 상관없다. 내게는 무엇인가 꼭  이루고 싶은 목표가 없었다. 인생을 어떤 색으로 꾸미고 싶다는 소망도 없었다. 그저 현실에 잘 적응했을 뿐이다. p.33

어떻게 살 것인가? 이 나이에 아직도 이런 질문을 껴안고 있는 내가 한심해 보인다. 그러나 이것이 여태껏 살아온 내 삶의 결과임을 인정한다. 만약 지금까지 살아온 그대로 계속 살면 좋겠다고 생각한아면 그 사람의 삶은 이미 훌륭한 인생이다. p.34

법학과 진학과 사법고시를 포기한 것은 악과 싸워 세상을 바꾸기 위한 결단이라고 보다 죄악으로 가득한 세상에서 내 자신을 확실히 지키기 위한 선택이었다. '성인은 못 되더라도 괴물은 되지 말자' 그렇게 생각했다. 그것도 나름 의미 있는 선택이라고 생각한다. p.36

인생에서 가장 중요한 것은 '자기 결정권'을 행사하는 일이다. 자기 결정권이란 스스로 설계한 삶을 옳다고 믿는 방식으로 살아가려는 의지이며 권리이다. 철학자 존 스튜어트 밀의 표현을 가져다 쓰면 "사람은 누구든지 자신의 삶을 자기 방식대로 살아가는 것이 바람직하다 그 방식이 최선이 아니라 자기 방식대로 사는 길이기 때문에 바람직한 것이다." p.37

내가 진정 원하는 삶을 들여다 보지 않았다. 무엇인가 즐거움과 행복을 느낄 때마다 근원을 알 수 없는 죄의식을 느꼈다. 행복을 느끼는 순간 누구엔가 잘못을 저지르는 것 같았다. 생각해보면 꼭 그래야 할 이유는 없다. 내가 괴로워한다고 해서 누군가 더 행복해지는 것도 아니었다. 내게는 내가 원하는 대로 살 권리가 있다. p.39

## 1_3. 왜 자살하지 않는가

열등감은 삶을 갉아먹는 부정적인 감정 중에서 가장 고약한 것이다. 열등감에 깊이 빠지면 자기 자신을 비천한 존재로 느끼게 된다. 그래서는 기쁜 삶을 살지 못하고 기쁘지 않은 삶은 훌륭하기 어렵다. 열등갑과 자기 비하에 빠지지 않으려면 '정신승리법'이 필요하다. p.42

여우의 '신 포도 논리'는 다소 진부하게 보이지만 삶을 긍정적인 태도로 살아가는데 도움이 된다. 도전하지 않고 포기하는 것도 어리석지만, 오르지 못할 나무와 넘을 수 없는 벽에 매달려 인생을 소모하는 것 또한 어리석인 일이다. 나한테 적합한 나무와 적당한 벽을 잘 골라야 인생이라는 너무 짧은 여행을 후회 없이 즐길 수 있다. p.43

살아있는 순간 동시에 우리는 죽어간다. 삶과 죽음은 처음부터 끝까지 동행하며 함께 완성된다. 어자피 죽을 것이기에 삶이 허무하다고 말하지 말자. 그것은 틀린 말이다. 오히려 그 역이 옳다. 죽을 수 밖에 없기 때문에 삶은 아름다운 것이다. p.46

죽음 가운데 극소수만이 관심을 끈다. 죽음으로써 말하려고한 내용을 들려준다. 때론 자살은 고통에서 벗어나 자유와 존엄을 찾는 수단이 되기도, 삶의 의미를 완성하는 행위가 되기도 한다. 구한말 지직인 매천 황현이 자결, 청년 노동자 전태일의 분신, 노무현 대통령의 투신이 그런 것이라고 생각한다. 그들의 죽음은 살아있는 사람들한테 던진 질문이다. 들은 준비가 된 사람들은 그들과 속깊은 이야기를 나눈다. p.48

## 1_4. 위로가 힘이 될까

아프고 지친 사람이 많아서 그런지 요즘에는 힐링이 대세이다. 그런데 나는 이런 말을 잘 하지 못한다. 어떤 이야기가 위로와 치유가 되는지 몰라서가 아니다. 자기의 삶이 어떤 의미가 있는지 치열하게 고민하지 않은 사람에게 타인의 위로가 별 도움이 되지 않는다고 보기 때문이다. 청년은 아기가 아니다. 삶의 의미는 사회가 찾아주지도 않고 찾아줄 수도 없으며 찾아주어서도 안 되기 때문에 자기가 직접 찾아야 한다. 각자 남들을 조금 더 배려하고 제도를 좀 더 합리적으로 바꾸면 모두가 존엄하고 의미 있는 삶을 살 수 있다는 착각을 일으키지 않을까 걱정이다. p.51

자기가 원인을 제공하지 않은 문제 때문에 고통을 받는 것은 부당하다고 주장할 수 있다. 그러나 그렇다고 해서 그 고통을 피해갈 수 있는 것은 아니다. 내 책임이든 사회의 책임이든 닥쳐온 고통은 일단 내가 견디고 이겨내야 한다. 세상을 원망한다고 달라질 것은 없다. 누구도 그 짐을 대신 짊어져주지 않는다. p.52

노동자들을 폭행한 용역들이 조폭들일 것이라고 생각하면 오산이다. 거기에는 대학생들도 섞여 있었다. 일당 7만원 짜리 대학생 아르바이트는 흔하지 않다. 거기에 있던 청년은 등록금을 벌기 위해 용역업체 일을 계속하겠다고 말했다. 도덕적 평가는 차치하더라도 용역 폭력은 명백한 범죄임에도 불구하고 그 청년은 아무런 죄책감이 없었다. 사람을 패는 일 말고는 등록금을 마련할 다른 일자리가 있었다면 결과는 달랐을 지도 모른다. 그 청년을 꼭 나쁘다고 단정지을 수 없다. 그러나 어쨌든 그는 범죄에 가담했다. 돈을 받고 타인을 폭행한 것은 형법상 범죄일 뿐만 아니라 타인과 자신의 존엄을 동시에 해친다. p.54

삶의 의미와 그것을 실현하는 것은 자유의지를 가진 개인의 존엄함 권리이지만, 그 자유의지를 발현하는 데는 어떤 상황에서도 지키려고 노력해야 마땅한 이성의 원리 또는 도덕법이 있다. '그것은 내 책임이 아니다'라고 용역 폭력에 가담한 대학생을 위로할 수 없다는  것, 그렇게 위로함으로써 그가 자신과 남에게 준 상처를 치유하거나 인간적 존엄을 회복하게 할 수 없다는 것만큼은 분명히 해두자. p.55

상처 받지 않는 삶은 없다. 상처받지 않고 살아야 행복한 것도 아니다. 누구나 다치면서 살아간다. 우리가 할 수 있는 일은 그 어떤 날카로운 모서리에 부딪쳐도 치명상을 입지 않을 내면의 힘, 상처 받아도 스스로 치유할 수 있는 정신적 정서적 능력을 기르는 것이다. 자신의 인격적 존엄과 인생의 품격을 지키려고 분투하는 사람만이 타인의 위로를 받아 상처를 치유할 수 있으며 타인의 아픔을 위로할 수 있다. p.56

## 1_5. 놀고 일하고 사랑하고 연대하라

카뮈는 우월한 유전자, 지적인 가풍, 훌륭한 가정 교육, 위대한 인물의 탄생과 관련해 거론되는 요소가 하나도 없었다. 카뮈의 유년기는 스스로 책임질 이유도 방법도 없는 불행과 고통의 연속이었다. 하지만 카뮈는 낭만적으로 열정적인 삶을 찾아나갔다. 그 길을 열어준 것은 소년의 재능이었다. p.58

일을 잘하는 사람은 놀듯이 일한다. 좋아하는 일을 잘 하면 일이 놀이만큼이나 즐거워 질 수 있다. 정치투쟁, 글쓰기, 연극, 연출 이 모든 것들이 카뮈에게는 일이자 놀이였다. 세상에는 다른 일, 다른 놀이, 다른 방식의 사랑이 얼마든지 있지만 카뮈는 자신이 하고 싶은 일에 최선을 다했고, 자기가 가치 있다고 여기는 놀이와 사람에 열정적으로 임했다. 그런 것에서 삶의 의미를 찾았기에 자살하지 않았다. 마틴 셀리그만이 주장한 삶의 위대한 세 영역은 사랑, 일, 놀이이다. 실제로 사람들이 이 셋으로 삶을 채우며 여기에서 살아가는 의미를 찾는다는 것이다. 그래서 위대란 세 영역이라고 하는 것이다. p.61

나는 셀리그만의 견해에 "연대"를 추가해야 한다고 생각한다. 연대란 좁게 보면 동일한 가치관과 목표를 가진 누군가와 손잡는 것이고 넓게 보면 기쁨과 슬픔, 환희와 고통에 대한 공감을 바탕으로 삼아 어디엔가 함께 속해 있다는 느낌을 나누면서 서로 돕는 것을 의미한다. 세금을 납부하고, 병역의 의무를 이행하고, 투표하고, 정당을 만들고, 이웃을 돕고, 시위를 하고, 유기동물을 보살피고, 아프리카 어린이를 후원하고, 멸종위기에 처한 동식물을 구하고, 온실가스 배출과 에너지 소비를 자제하는 등의 행위는 모두 사회, 국가, 인류, 지구에 대한 귀속감을 느끼고 표현하는 일이다. p.61
