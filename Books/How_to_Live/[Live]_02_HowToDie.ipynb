## 2_1. 죽음이라는 운명

나도 어른이 될 텐데 결국에는 죽을 것이 아닌가? 내가 세상에 온 데에는 무슨 특별한 목적이나 이유가 있는 걸까? 어른이 되면 무슨 일을 하면서 살게 될까? 구서 없이 생각을 하는데 팔뚝에 소름이 돋았다. 집 뒤 대나무 숲을 쓸어가는 바람 소리가 무서워졌따. 정체 모를 두려움과 슬픔이 밀려온다. p.68

'피할 수 없으면 즐겨라'라는 광고 카피가 유행한 적이 있따. 인생의 좌우명으로 삼아도 손색이 없을 정도로 좋은 말이다. 그러나 예외가 있는데 바로 죽음이다. 피할수도 즐길 수도 없다. 죽을 일을 생각하면 누구나 두려움을 느낀다. 죽음을 좋아하는 사람은 없다. 자살하는 사람은 좋아서 하는 것이 아니다. 젊은이들한테는 죽임이 멀리 있는 것처럼 보인다 삶이 유한하다는 것을 알지만 아직은 살날이 무한정 남아 있는 것 같다. 사실 청년들한테 시간은 아직 '회소적 자원'이 아니다. 조금쯤은 낭비해도 괜찬다. 방황가 시행착오를 격어도 될 만큼의 여유가 있다. 이것을 가리켜 청춘의 특권이라고 한다. p.69

죽음 다음에 무엇이 있을까? 만약 내일 죽는다면 오늘 무엇을 할까? 잘 죽으려면 어떻게 준비해야 할까? 이런 저런 대답을 혼자 생각해본다. 답을 꼭 찾아야 할까? 아무래도 그런 것 같다. 어떻게 대답하느냐에 따라 내 남은 생이 달라질 수 있을 거 같은 생각이 든다. 죽음은 삶의 끝을 의미하는 것이 아니다. 죽음은 삶의 완성이다. p.71

## 2_2.남자의 마흔살

젊은 사람은 공감하기 어렵겠지만 나이가 들면 젊은 그대들도 인생의 마지막 페이지를 예감하는 각성의 시간이 반드시 찾아온다. p.73

하지만 내가 예외적인 노인이 된다고 해도 젛은 세대에 나보다 더 능력있는 사람이 많을 것이다. 연장자를 공경하고 윗사람이 결정권을 가지고 있어 말을 하지 못할 뿐, 젊은이들은 언제나 세대교체의 순간이 오기를 기다린다. 그러니 나이가 들면 한 걸음 뒤로 물러나 있으면서 후배들이 재혜를 구하러 오면 조심스레 조언을 하는 선에서 머무르는 것이 바람직할 것 같다. p. 77

무역 전문가로 사는 것이 매력적일지 생각해보았다. 나름 의미는 있겠지만 가슴이 두근거리지는 않았다. 설렘이 없으니 열정이 불타오를리 없었다. 마음의 설렘이 없는 일에 인생을 쓰고 싶지 않았다. 어떻게 사는 인생이 훌륭할까? 하고 싶어서 마음이 설레는 일을 하자. 그 일을 열정적으로 남보다 잘하자. 그걸로 밥도 먹자. 이것이 성공한 인생이 아니겠는가. p.78

## 2_3. 나도 죽고 싶었던 때가 있었다

소충중대에 배치되어 해발 1000미터 산꼭대기 막사에서 한 달을 지낸 다음 철책선 근무에 투입되었다. 그 무렵 땔나무를 하고, 물을 져 나르고, 본대에서 부식을 수령하느라 눈 덮인 계곡의 낭떠러지를 지나다녔다. 종종 뛰어내리고 싶은 충동을 느꼈다. p.80

"죽을 용기가 있다면 그걸로 살아 볼 일이지!" 그러나 자살을 용기로만 하는 것이 아닌 것 처럼, 삶도 용기만 있다고 해서 마냥 잘 사는 것도 아니다. 사는 데도 죽는 데도 다른 것이 있어야 한다. 바로 확신이다. 그것이 없으면 삶도 죽음도 주체적일 수 없다. 삶은 습관이고 죽음은 패배일 뿐이다. p.83

자살을 생각한 5명 가운데 1명은 실제로 자살을 시도한다. 자살은 인간이기 때문에 할 수 있는 철학적 실존적 선택이다. 특별히 못나서 자살을 선택하는 것이 아니다. 동물은 먹을 것을 구하지 못하면 최후까지 버티다가 굶어 죽는다. 그러나 사람은 최소한의 인간적 존엄조차 지길 수 없을 때 스스로 목숨을 끊기도 한다. p.83

자살은 단순한 충동이 아니다. 누구도 가벼운 마음으로 자살하지 않는다. 겉보기에는 마치 한 순간의 분노나 충동을 억제하지 못해 목숨을 끊은 것처럼 보이지만, 실제로는 죽음이 직접 동반하는 것보다 더 혹심한 몸과 마음의 고통을 겪은 뒤에 자살을 감행한다. 학업 성적, 경제적 궁핍, 질병의 고통, 가족간의 불화, 명예 실추, 타인의 비난, 풀 길 없는 억울함, 그 동기가 무엇이든 다르지 않다. 그런 것들이 자존심을 회복할 수 없을 정도로 파괴할 때, 인간의 존엄성을 회복할 수 없다고 느낄 때 자살은 탈출구가 된다. p.85

우리들 각자는 사회적인 것이든 개인적인 것이든 부정적인 생활 사건이 주는 스트레스를 관리하고 극복할 수 있는 능력을 길러야 한다. 내 나름대로의 비법이 있기는 하다. 바로 거리감이다. 세상에 대해서 타인에 대해서 내 자신에 대래서 일정한 거리감을 유지하는 것이다. 나는 좋은 세상을 원하지만 그 소망이 이루러지지 않는다고 해서 세상을 저주하지 않는다. 좋은 사람들을 사랑하지만 무조건적이고 절대적인 사랑을 믿지 않는다. 내 생각이 옳다고 확신해도 모두가 그것을 받아들여야 한다고 주장하지 않는다. 내가 하는 일들은 의미가 있다고 믿지만 그건 어디까지나 내 생각일 뿐임을 인정한다. 삶이 사랑과 환희와 성취감으로 채워져야 마땅하다고 생각하지만 좌절과 슬픔, 상실과 이별 역시 피할 수 없는 삶의 한 요소임을 받아들인다. p.89

내가 기쁨을 느낄 수 있는 방식으로 살자. 타인의 시선이나 평가에 얽매이지 말자. 내 스스로 삶에 가치를 부여하는 꼭 그만큼만 내 죽음도 의미를 가질 것이다. 나는 이렇게 생각하며 산다. p.90

## 2_4. 찬 이성 더운 가슴

철학적으로 볼 때 인간의 죽음은 고유한 정체성을 가진 자아의 소멸을 의미한다. 일반적으로 생물학적, 철학적 죽음은 동시에 일어난다. 하지만 자아 정체성을 상실한 치매환자의 경우 철학적으로는 죽었지만 생물학적으로는 살아 있는 사람도 있다. 인간의 몸과 마음에 대해 아는 것이 적었던 시절에는 마음이 심장에 있다고 믿었다. 알프레드 미셜 이래 많은 경제학자들이 인용한 '찬 이성 더운 가슴'이라는 표현은 그런 믿음을 멋지게 표현한 말이다. 세상사를 냉정하게 관찰하고 추론하고 계산하는 일은 두뇌가 한다. 그러나 그것을 어떤 목적에 사용할지 결정하는 것은 심장의 몫이다. 공감, 연민, 나눔, 정의, 열정 이런 것들은 심장에 있다. p.94

심장이 멈추면 사람은 몇 분만에 죽는다. 심장은 느끼거나 생각하지 못한다. 차가운 계산도 뜨거운 헌신도 모두 두뇌가 하는 일이다. 이성도 마음도 모두 거기에 있다. '찬 이성 더운 가슴'은 사람이 갖추어야할 좋은 덕목이지만 그 둘은 따로 있지 않다. 뇌가 죽으면 찬 이성과 더운 가슴음 함께 사라진다. p.96

## 2_5. 타인의 죽음과 나의 죽음

인간의 죽음은 모두 똑같다. 아인슈타인의 죽음과 노숙자의 죽음, 나의 죽음과 내가 싫어하는 사람의 죽음은 아무런 차이가 없다. 그러나 철학적인 관점에서 보면 그렇지 않다. 근본적으로 다른 종류가 있는데 바로 나의 죽음과 타인의 죽음이다. 타인의 죽음에 대해서는 객관적으로 받아들이지만 나의 죽음에 대해서는 주관적, 감성적으로 대응한다. p.97

죽음을 표현하는 말은 다양한다. 하지만 자기의 죽음은 어떤 지위를 누리고 어떤 일을 하든 상관없이 그저 죽을 뿐이다. 타인의 죽음의 세상의 한 조각이 사라지는 것이지만 나의 죽음은 나의 삶과 나 자신, 내 주의 세상이 통째로 사라지는 것이다. p.98

나는 무엇인가. 나는 욕망과 감정, 기억과 소망의 덩어리이다. 심리학자 지그문트 프로이트는 이것을 에고라고 불렀다. 에고는 이드와 슈퍼에고의 통일이다. 이드는 오로지 욕망을, 슈퍼에고는 양심와 이상을 쫒는다. 에고는 과거의 사건과 행위를 비판적으로 기억하고 평가하면서 미래를 기대하고 상상하는 가운데 현재의 행위를 설계하고 실행한다. 살아가면서 사람은 좋은 쪽이든 나쁜 쪽이든 변할 수 있는 것이다. p.90

인간의 대뇌피질에는 타인의 고통과 기쁨을 인지하고 공감하게 만드는 거울뉴런이 있다. 우리가 이런 사실을 알게 된 것은 20세기 막바지에 MRI와 PER등 첨단장비로 무장하고, 과학의 무대 위로 올라와 '두뇌 지도'를 보여준 과학자들 덕분이다. p.103


유전자는 자식에서 손자로, 그 자식으로 끊임없이 이어지는 불멸의 코일이다. 하지만 유전자는 생물학적으로 의미가 있을 뿐 철학적 가치는 없다. 유전자는 기억하지 않으며 사유하지도 않고 자아도 없다. 오로지 자신을 '나'라고 인식하는 자아는 언제나 단 한 번 존재할 수 있다. 유전자는 유전자일뿐 나가 아니다. p.104

## 2_6. 나는 무엇인가

주어진 환경에서 나름 최선을 다해 살아왔다. 하지만 그 모든 것을 스스로 결정한 것이 아닐 지 모른다는 생각이 들었다. 종종 하고 싶은대로 하지 못한다는 느낌을 받는다. 무엇인가 해야 한다고 생각하지만 몸이 거부하고, 다르게 해야한다고 하지만 그렇게 하지 못할 때가 있다. p.108

리처드 도킨스는 모든 생물은 '생존 기계'라고 주장했다. 내가 아내와 자식들에게 느끼는 사랑이 유전자가 만들어낸 감성에 지나지 않을 지도 모른다. 논리적으로 그럴 듯하지만 이것을 인정하면 왠지 슬퍼진다. 유전자가 삶을 지배하는 것을 전면 부인하기는 어렵다. 하지만 같은 환경에서도 아이들은 다르게 자란다. 사람은 환경의 지배를 받는다. 사람은 스스로 자신의 성격이나 인생관, 행동 양식을 바꿀 수 있다. p.108

나는 무엇인가? 나는 과거를 비판하면서 미래를 꿈꾸고 현재를 살아가는 '철학적 자아'이다. 거칠게 대답하면 자는 뇌이다. 내 자아는 뇌에 기거하면 모든 행동과 생각은 뇌에서 비롯된다. 뇌의 구조는 지하실이 딸린 2층집과 비슷하다. 지하실인 뇌관은 파충류의 뇌로써 생명활동을 담당하고 1층은 변연계로 감성을 조절하는 편도, 기억을 저장하는 해마, 호로문 분비를 조절하는 사상하부, 운동을 제어하는 기저핵이 있다. 2층에는 대뇌피질은 지식인의 거실로 영장류가 가장 발달한 부위이다. 뇌의 무게는 1.4킬로그램정도인데 이중 80%가 대뇌피질이다. 자아가 담기는 공간이다. 프로이트는 대충 이런 구조를 이해한건 같다. 이드는 지하실과 1층을 오르내리며 살고, 슈퍼에고는 2층에서 양심과 이상을 담당한다. p.111

이타성이라는 것도 스스로 새운 규범을 따를 행위일 때 기쁨이 되는 것이지 않는가. 욕망을 억압하면서 규범을 따르는 일이 참기 어려울 만큼 어색하고 불편하고 고통스럽다면 욕망을 표출할 수 있는 문을 더 넓게 열어주는 것도 나쁘지 않다고 생각한다. 규범은 자기 자신이 기쁜 마음으로 자연스럽게 감당할 수 있는 만큼만 따르면 된다.

## 2_7. 레이건의 작별인사

남의 일이 아니다. 예순다섯 살 넘은 대한민국 국민 가운데 52만명이 치매를 앓고 있다. 더 젋은 환자도 적지 않다. 환자의 자녀, 며느리, 사위, 손주까지 합치면 수백만 명의 국민들이 생물학적 죽음보다 먼저 찾아온 철학적 죽음을 경험하고 있다. 고령 치매 환자 숫자는 앞으로 10년안에 1백만 명을 넘어설 전망이다. p.117

로널드 레이건은 아흔네 살 나이로 2004년 세상을 떠났다. 그는 미국 대통령을 2번 지냈으며 죽기 전 십 년동안 알츠하이머병을 앓았다. 그는 보기 드문 방식으로 이 병을 맞았다. 바로 국민들한테 알린 것이다. 같은 병을 겪는 사람들과 그 가족들에 대한 관심을 높이는 데 도움이 되기 바란다고 했다. p.119

레이건은 미국의 경제를 망쳤다. 부자를 더 부자로 만들기 위해 정부와 국민을 가난하게 만들었다. 곳곳에서 전쟁을 별여 많은 죄없는 사람까지 죽게 만들었고 전쟁터로 내몰렸던 미국 청년들고 숱하게 죽었다. 그러나 그는 뚜렷한 정체성을 지니고 자기의 삶을 자기가 설계한 방식으로 살았다. 레이건은 철학적 죽음에 의연하게 대처하였다. p.120

## 2_8. 존엄한 죽음

죽음을 강을 이루는 물방울들은 모두 자기마다의 인생 스토리가 있는 하나뿐이었던 자아이다. 자신의 죽음을 분명하게 이해하고 해석하고 미리 준비하는 사람은 몇이나 될까? 의식이 꺼지려는 마지막 순간에 그들의 자기의 삶을 어떻게 평가하면서 죽음을 맞이했을까? 사람은 대부분 병에 걸려서 죽는다. 압도적 다수가 암, 뇌혈관 질환, 심장 질환, 당뇨병, 폐렴, 고혈압, 간 질환 등의 질병으로 사망한다. p.123

오래 사는 것이 좋은 일인가? 꼭 그렇지는 않다. 장수는 기회인 동시에 위험이다. 운이 없거나 잘 대비하지 못하면 재앙이지만 일정한 조건이 갖추어 진다면 축복이 될 수 있다. 무엇보다 삶이 의미있고 즐거워야 하며 건강해야 한다. 그리고 돈이 있어야 한다. p.125

이제는 세상이 달라졌고 삶이 바뀌었다. 물론 아직 완전한 평화가 아닌 전쟁이 위험이 남아있어 종종 군인들이 전투를 벌이고 포탄이 날아다니지만 정부와 건설업체들은 초고층 신도시와 산업시설을 계속 만들고 있다. 이제는 단순히 오래 살아남는 것만이 삶의 목표일 수는 없다. 죽음은 기피 대상이 아닌 차분히 생각하고 준비해야할 것이 되었다. p. 128

생명은 존엄하다. 그러나 죽음 역시 존엄해야 한다. 의학적 연명치료는 단순히 의미가 없는 게 아니라 사람을 모욕하고 존엄을 짓밟는 것이라고 생각한다. 김옥경 할머니에 대한 '존엄사'가 세상에 받아들여 진 것은 대다수의 사람들이 그렇게 생각하기 때문이었다. p.131

영원히 지속될 수 없는 삶의 마지막 단계에서 의미없는 날들이 계속되면 어떻게 해야 할까? 만약 그런 상황에 직면한다면 나도 내가 원하는 방식으로 삶을 마감할 것이다. 사랑하는 사람들, 추구하던 가치들, 한때 기쁨과 의미를 주었던 모든 것들과 스스로 선택한 방식으로 작별하는 것은 누구도 막아서는 안 될 자유이며 존엄한 권리라고 나는 믿는다. p.135

## 2_9. 자유 의지

그렇다면 제 몸을 가누지 못하는 삶은 의미가 없는 것일까? 그렇지 않다. 그런 조건에서도 삶을 끈질기게 이어갈 권리가 있다. 스스로 손가락 하나 움직이지 못해도 끈질기에 삶을 지속하는 사람이 있다. 바로 스트브 호킹 박사이다. 그는 자신의 힘으로 어떻게 할 수 없는 최악의 조건에서도 나름의 방식으로 삶을 이어나갔다. 그 동력이 무엇이었을까? 지적 호기심, 깨달음의 즐거움, 자기가 알게 된 것을 사람들과 나누는 기쁨, 이런 것들이 그의 삶을 밀고 나간 원동력이 아니었나 싶다. p.136, 137

사람은 존엄을 지키기위해 살기도 하고 죽기도 한다. 그것이 인간이다. 존엄이란 무엇인가? 이 단어의 어원은 라틴어 '다그니타스 Dignitas'이다. 존엄은 일상생활에서 존엄과 고귀함을 의미한다. 철학자 임마누엘 칸트에 따르면 존엄한 것은 가치를 따질 수 없다. 가치는 사람들이 인정한 가치이며 인정한다면 얼마나 높게 평가하는지에 좌우된다. 그러니 그 자체가 목적인 것은 가치를 따질 수 없다. 옳은 것과 그른 것 사이의 선택을 나타내는 것만이 그 자체로 목적이 된다. 인간다움, 존엄성이 그러 것들이다. p.138

그러나 자유의지가 제멋대로 살고 죽는 것을 무조건 정당화 하지 않는다. 자유의지를 발현할 때 지켜야 할 규칙 또는 도덕법이 있다. 칸트는 이 규칙을 이성이 내리는 '정언명령'이라고 했다. 칸트의 도덕법은 2가지로 첫째, 스스로 세운 준칙에 따라 행동하되, 보편적 법칙이라고 주장할 수 있는 준칙이라야 한다. 둘째, 나 자신이든 다른 어떤 사람이든 인간을 절대로 단순한 수단으로 다루지 말고 언제나 한결같이 목적으로 다루도록 행동하라. 존엄한 인간이 자유의지를 옳게 발현하려면 이 두 가지 규칙을 따라야 한다. p.139, 140


라몬은 자유 없는 삶은 존엄성에 대한 모독이라고 여겼다. 그는 자신의 결정이 칸트의 도덕법에 어긋난다고 생각하지 않았다. 신성한 것은 삶 그 자체가 아니라 삶의 존엄성이며 자유로운 판단에 따라 삶과 죽음을 선택할 수 있는 인간의 권리라고 생각했다. 빗소리를 들을 수 있었지만 몸에 흐르는 빗물은 느낄 수 없었다. 사랑을 머리로 알 수 있지만 몸으로 느낄 수 없는 삶, 라몬에게 그것은 지옥이었다. 그는 지옥을 떠나 자유를 얻고 싶었다. p.144,145

라몬 삼베드로가 남긴 글을 읽으면서 육체, 욕망, 충동이 가지는 의미를 다시 보게 되었다. 우리는 많은 것들로 삶을 채우며 그런 것에서 삶의 의미를 발견한다. 일, 놀이, 사랑, 이념, 지식, 돈, 명예, 권력 ... 무엇이든 기쁨의 원천이 될 수 있고 삶의 의미가 될 수 있다. 그러나 그런 것보다 중요한 것은 삶의 의미를 아는 것이라는 생각이 들었다. 당신은 무엇으로 인생을 채우고 있는가? 그것이 당신의 삶에 충분한 의미를 부여하는가? 살아있는 순간마다 당신을 기쁨을 느끼는가? 라몬 삼베드로가 이렇게 묻는 것만 같다. p.146
