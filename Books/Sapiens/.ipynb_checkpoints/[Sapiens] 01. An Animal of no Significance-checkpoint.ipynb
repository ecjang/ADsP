#### 1부. 인지혁명

# 1. 별로 중요치 않은 동물


약 7만 년 전 호모 사피엔스 종에 속하는 생명체가 좀 더 정교한 구조를 만들기 시작했다. 문화가 출현한 것이다. 그 후 인류문화가 발전해온 과정을 역사라고 부른다. 역사의 시작은 약 7만 년 전 인지 혁명, 역사의 발전은 약 1만 2천 년 전 농업혁명, 그리고 현재의 과학혁명은 500년 전에 일어났다.

- 종 : 서로 교배하는 경향이 있고 후손을 낳을 수 있다. 
- 속 : 같은 조상에서 진화한 각기 다른 종들.
- 과 : 속의 상위 개념. 같은 선조를 가지는 속들이 포함.

- 표기할 때는 종, 속으로 표기. 예> 사자 Panthera leo : Panthera 속에 속하는 leo 종이라는 의미


### 1-1. 인류가 스스로 숨겨온 비밀

인류 : 호모 속에 속하는 현존하는 모든 종.

인류는 약 250만 년 전 동부 아프리카의 오스탈로피테쿠스에서 진화했다. '남쪽의 유인원'이라는 의미. 그 후 여러 지역으로 여행하며 각기 다른 방향으로 진화했다.

1. **호모 네안데르탈렌시스** : 네안데르 골짜기에서 온 사람. (=네안데르탈인). 덩치가 크고 근육이 발달함.
1. **호모 에렉투스** : 똑바로 선 사람. 동아시아 지역. 그 지역에서 200만 년 생존.
1. **호모 솔로엔시스** : 솔로 계곡에서 온 사람. 인도네시아 지역. 열대지방에 잘 적응.
1. **호모 데니소바인** : 시베리아 데니소바 동굴에서 발견.
1. **호모 사피엔스** : 슬기로운 사람. 우리 인류가 되는 종.

### 1-2. 생각의 비용

인간의 여러 종은 다양한 차이점이 있지만 공통적으로 뇌가 크다.  고인류는 뇌가 커지면서 식량을 찾아다니는데 많은 시간을 썼고, 근육이 퇴화했다. 그리고 직립보행을 하였다. 이동에 사용하던 팔을 다른 섬세한 작업을 할 수 있도록 진화하였다. 특히 여성은 산도가 좁아졌는데 이로 인해 미발달 상태에서 출산하게 되었다. 그래서 갓 태어난 아기는 부모와 다른 사람들의 부양이 필요하며 이는 곧 사회적 능력의 발전으로 이어졌다. 

### 1-3. 익혀 먹는 종족

먹이사슬의 정점에 올라서는 핵심적인 계기는 바로 불이다. 80만 년 전쯤 일부 인간은 불을 사용했으며 30만 년 전쯤에 호모 에렉투스, 호모 네안데르탈인, 호모 사피엔스가 불을 일상적으로 사용했다. 

불의 사용으로 조리가 가능해지면서 생으로 먹을 수 없었던 밀, 쌀, 감자 등이 인간의 주식이 되었고 원래 즐겨먹던 과일, 견과류, 벌레, 죽은 고기고 불에 익히면 소화가 잘 되었다. 불을 길들임으로써 무한한 잠재력을 통제할 수 있게 되었다. 언제 어디서든 사용할 수 있고, 수많은 용도로 사용 가능하며, 힘의 한계를 극복하였다.

### 1-4. 호모 사피엔스 - 형제 살해범

호모 사피엔스가 아라비아 반도에 상륙했을 때 거기엔 이미 다른 인간들이 정착해있었다. 서로 다른 인종은 교배했거나 아님 한쪽을 전멸시켰다는 설이 있다. 

- 교배 이론 : 사피엔스는 네안데르탈인과 서로 교배하여 서로 하나가 됨.

- 교체 이론 : 서로 성적인 관심이 없으며 사피엔스가 네안데르탈인을 살해.

교체 이론이 우세하였으나 2010년에 네안데르탈인 게놈 연구 결과 중동과 유럽 인구의 1~4%는 네안데르탈인의 DNA인 것으로 밝혀졌다. 약 5만 년 전 당시의 사피엔스와 네안데르탈인, 데니소바인 들은 유전자의 회색지대에 있어 어느 정도 교배가 가능했을 거라고 추측할 수 있다. 즉 완전히 다른 종은 아니지만 대체로 별개의 종이 었으며 일분 몇몇 개체들은 서로 교배를 하였을 것이다. 

사피엔스가 네안데르탈인을 전멸시킨 이유로는 몇 가지가 있지만 자원을 둘러싼 경쟁이 주된 이유라고 생각된다. 현대의 사람들도 피부색이나 종교, 언어 차이로 다투기 일수인데 과거 사피엔스도 다르지 않을 것이다. 

우리가 무시하기에는 너무 친숙하고, 관용을 베풀기에는 너무 달랐다.

사피엔스의 성공의 비결에 대한 논쟁은 아직도 뜨겁게 진행되고 있지만 가장 그럴싸한 정답은 바로 언어이다. 우리에게만 있는 고유한 언어 덕분이다.
