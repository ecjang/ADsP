#### 1부. 인지혁명

# 2. 지식의 나무


네안데르탈인과 사피엔스의 첫 번째 승리는 네안데르탈인의 것이었다. 하지만 **약 7만 년 전 호모 사피엔스는 매우 특별한 일**을 하기 시작했고 지구 상 모든 지역으로 이동하며 네안데르탈인을 몰아냈다. 대부분의 연구자들은 이런 업적이 **사피엔스의 인지능력의 혁명의 결과**라고 주장한다.

인지 혁명이란 **약 7만 년 전부터 3만 년 전 사이에 출현한 새로운 의사소통 방식**을 의미한다. 우연히 일어난 유전자 돌연변이가 뇌의 내부 배선을 바꾸어서 생겨났다. 이것을 **지식의 나무 돌연변이**라고 한다.

우리의 언어가 동물들의 언어와 다른 점은 바로 **언어의 유연성**이다. 기호와 언어, 소리 등을 이용해 문장을 만들어 다양한 정보를 공유할 수 있다. 초기 언어는 단순히 위험을 알리거나 사냥을 위한 정보를 공유하는 정도였지만 언어의 발달을 통해 **세상에 대한 정보, 즉 사람에 대한 정보**를 이야기할 수 있었다.

사피엔스는 사회적인 동물이다. 무리 안에서 누가 누구를 미워하고, 누가 정직한지, 누가 속이는지 등의 정보가 중요한데 바로 이런 **뒷담화 능력**이 있어 무리 간의 협동이 가능했다. 누가 신뢰할 수 있는 사람인지 파악하여 더 큰 무리로 발전할 수 있었고 더 복잡한 협력이 가능했다. 

**전설, 신화, 신, 종교는 인지 혁명과 함께 등장했다.** 사피엔스는 사자를 보고 '우리 종족의 수호령이다'라고 말할 수 있게 되었다. **허구를 믿는 능력이야말로 언어의 가장 큰 특징이다.** 허구 덕분에 우리는 단순한 개인의 상상을 넘어 집단적으로 상상을 할 수 있게 되었고, 신화나 민족주의 같은 개념을 만들었다. 그래서 대규모 협력이 가능하였다. 

### 2-1. 푸조라는 신화

동맹의 구성원은 서로 많은 시간을 보내고 서로 도우며 살아가지만 이런 방식으로는 무리를 키우는데 한계가 있다. 서로 만난 적도, 싸운 적도, 서로 털을 골라 준 적도는 두 침팬지는 서로 믿을 만한 존재인지 알 수 없다. 침팬지는 서로 다른 무리들은 협력하지 않으며 영토와 먹을거리를 두고 경쟁한다. 



사피엔스는 인지 혁명을 통해 침팬지보다 더 크고 안정된 무리를 이룰 수 있었다. 이런 능력으로 **150명 정도의 무리를 유지**할 수 있었다. 오늘날 집단의 임계치 또한 이 정도이다. **오늘날의 사피엔스는 이보다 더 큰 공동체를 유지한고 있다. 바로 허구의 등장이다.** 국가는 공통의 국가적 신화에 기반을 두고 있다. 



기업들의 세계를 예로 들어보자. 현대의 기업가와 원시 샤먼들의 차이는 현대 법률가들이 훨씬 더 이상한 이야기를 한다는 점이다. **푸조라는 자동차 회사가 존재**한다고 말할 때 이것은 무슨 뜻일까? 푸조라는 자동차는 많지만 이런 자동차들의 집합이 푸조는 아니다. 공장의 설비, 정비공, 회계사들을 합쳐도 푸조가 되는 것은 아니다.

반대로 판사가 해산 판결을 내리면 공장도, 직원도, 경영자 모두 다 온전해도 푸조는 사라진다. 푸조는 정말로 존재하는 것일까? **푸조는 우리의 집단적 상상력이 만들어낸 환상이다. 변호사들은 이를 법적인 허구라고 부른다. 물리적 실체는 없지만 법적인 실체가 있다.**



사피엔스는 막강한 힘을 가지게 된다. 서로 모르는 사람 수백 명이 힘을 모아 한 가지 목표를 향해 매진할 수 있게 되기 때문이다. 만일 우리가 실제 하는 것만 이야기했다면 국가와 교회, 법 체계를 만들 수 없었을 것이다. 



**가상의 실재란 거짓말을 의미하지 않는다.** 거짓말에는 전혀 특별함이 없다. 가상의 실재는 모든 사람이 믿는 것을 의미한다. 이런 공통의 믿음이 지속되는 한, 이런 가상의 힘은 현실세계에서 힘을 발휘한다. 대부분의 백만장자는 돈과 유한회사의 존재를 신봉한다. 시간이 흐르면서 가상의 실재는 점점 더 강력해졌고, 오늘날에는 강변의 사자라는 실체는 미국이나 구글 같은 가상의 실재들의 자비에 좌우될 정도이다.

### 2-2. 게놈 우회하기

원시인류의 행동 패턴이 수십만 년간 고정되어 있던 반면 사피엔스는 불과 **10년, 20년만에 사회구조, 인간관계 속성, 경제활동을 비롯한 수많은 형태를 바꿀 수 있었다.** 급속하게 바뀌는 외부의 도전에 맞게 자신들의 사회 형태를 바꿀 수 있었다.

사피엔스 외에는 **교역을 하는 동물들이 없고,** 상세한 증거를 가지고 있는 사피엔스의 교역망은 모두 픽션에 근거를 둔다. 교역은 신뢰 없이 존재할 수 없고 모르는 사람을 믿기는 매우 어렵다. 오늘날의 교역 또한 마찬가지로 연방준비은행, 기업의 토탬적 상표와 같은 허구적인 실체들에 대한 신뢰를 기반으로 하고 있다.

또 다른 증거로는 사냥이 있다. 네안데르탈인은 소수로 사냥한 반면 사피엔스는 무리를 지어 대규모 사냥을 한 흔적들이 있다. 인공적인 함정과 도살장을 만든 유적지도 있다. 

인지 혁명으로 무엇이 일어났나?

- 더 많은 정보를 전달할 수 있는 능력 ⇨ 복잡하고 긴밀한 협동과 행동이 가능.
- 사회적 관계 형성, 뒷담화 능력 ⇨ 규모가 더 크고 더 강한 집단. 50명에서 150명으로 증가.
- 부족 정신, 국가, 인권 등 가상의 실체 ⇨ 낯선 사람들끼리의 대규모 협력, 사회적 행태의 급속한 혁신.

### 2-3. 역사와 생물학

사피엔스의 **가상의 실재는 엄청난 다양성과 행동 패턴을 문화라고 한다.** 일단 등장한 문화는 끊임없이 변화하고 발전하였으며 **그것을 역사라고 부른다.** 그러므로 유전자나 호르몬과 생물의 상호작용만이 아닌 개념과 이미지의 환상ㅇ이 벌이는 상호작용 또한 고려해야 한다.



생물하근 호모 사피엔스의 행동과 능력의 기본 한계를 결정한다. 하지만 이 영역은 극도록 넓기 때문에 사피엔스는 엄청나게 다양한 일을 할 수 있다. 픽션을 창조하는 능력으로 더 복잡하고 정교한 작업이 가능했다. 사피엔스의 행동을 이해하려면 그들의 행동이 역사적으로 진화해온 경로를 알아야 한다.
