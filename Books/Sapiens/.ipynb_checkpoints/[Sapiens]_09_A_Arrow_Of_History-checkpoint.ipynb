####  3부. 인류의 통합

# 9. 역사의 화살

수백만명이 서로 협력할 수 있는 인공적 본능을 창조했다. 그게 바로 **문화**다.

프랑스 혁명이래 세계 모든 곳의 사람들은 점차 평등과 개인의 자유를 근본적인 가치라고 생각하게 되었다. **하지만 두 가치는 모순된다.** 평등을 보장하는 방법은 형편이 더 나은 사림의 자유를 제한하는 것 외에는 방법이 없다.

중세문화가 기독교와 기사도를 어떻게든 조화시키는데 실패했듯이, 오늘날 사회는 자유와 평등을 조화시키는데 실패하고 있다. 그 모순은 인간문화에서 땔래야 땔 수 없는 부분이다. 사실 이것은 문화의 엔진으로서, 우리 종의 **창의성과 활력의 근원이기도 하다.** 

이것은 핵심적인 측면이기 때문에 이름까지 있다. **'인지부조화'**이다. 만일 사람에게 모순되는 신념과 가치를 품을 능력이 없다면, 인간의 문화자체를 건설하고 유지하는 것 자체가 불가능했을 것이다. **두가지 지상명제에서 흔들리고 있는 지점이야말로 당신이 그들을 가장 잘 이애할 수 있는 지점이다.**

----

### 9_1. 정찰위성

역사의 방향을 인지하는 것은 시점의 문제다. 역사를 거시적으로 보면 통일하는 방향인지, 다양성으로 가는 방향인지 알기 어렵다. 그보다는 정찰위성처럼 수백년이 아닌 수천년이라는 단위를 스캔하는 시점을 취하는 것이 낫다.

역사의 전반적인 방향을 이해하는데 가장 좋은 방법은 어떤 한 순간에 지구하는 행성 위에 **각기 분리된체 공존했던 인간 세상의 개수를 세는 것이다.** 지구에는 얼마나 많은 인간 세상들이 공존했을까? 기언전 1만년경에는 수천개였다. 기원전 2000년이 되자 수백개로 줄어들었고, 유럽인의 세계탐사 직전인 시기에는 많은 독립된 세계가 있었지만 90%에 가까운 인류는 아프로아시아라는 단 하나의 큰 세상에 살았다. 나머지 10%정도의 인류는 상당한 규모와 복잡성을 지닌 네 개의 세계로 분리되어 있었다. 바로 메소아메리카, 안데스, 호주, 대양세계이다. 이후 3백년간 아프로아시아 거인은 세계의 나머지 지역을 전부 집어삼켰다.

우리는 고유문화에 대해 이야기 하지만 '고유성'이 독자적으로 발전시킨 무엇을 의미한다면, **오늘날 지구상에는 고유한 문화가 하나도 없다.** 지난 몇 세기동안 모든 문화는 지구적 영향에 의해 거의 알아볼 수 없을 정도로 변했다. 이런 지구화의 가장 흥미로운 사례는 이름바 **'민속요리'**이다. 이탈리아 식당에서는 토마토소스를 넣은 스파게티를, 아르헨티나 식당에서는 수십 종의 스테이크를, 인도 식당에서는 거의 모든 요리에 고추가 들어갈 것이라고, 모든 스위스 카페의 하이라이트는 크림을 잔뜩 넣은 뜨겁고 진한 코코아를 예상한다. **하지만 이 중 어떤 음식도 이들 국가가 원산지가 아니다.**

### 9_2 지구적 비전

**어떠한 사회적 동물도 자신이 속한 전체의 이익에 이끌려 행동하지 않는다.** 침팬지 족의 이익에 관심을 가지는 침팬치는 한 마리도 없고, 지구적 달팽이 공동체를 위해 촉수 한 쪽이라도 까딱하는 달팽이는 없으며, 알파 수컷 사자들 중에 모든 사자의 왕이 되고자 나서는 사자는 한 마리도 없고, 벌집 입구에 "만국의 일벌들이여 단결하라"는 구호가 붙어 있는 경우도 없다.

기원전 첫 밀레니엄 동안, 보편적 질서가 될 잠재적인 후보 세 가지가 출현했다. **첫번째 질서는 바로 경제적인 것, 즉 화폐 질서였다. 두번째 질서는 정치적인 것, 즉 제국의 질서였고, 세번째 질서는 종교적인 것, 바로 불교, 기독교, 이슬람교 같은 보편적인 종교의 질서였다.** 인류의 잠재적 통일을 내다볼 수 이는 사람들은 상인, 정복자, 예언자들이었다. 이들은 어디에서나 누구에게나 적용되는 질서를 세우려고 노렸했다.
