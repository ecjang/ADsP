####  3부. 인류의 통합

# 10. 돈의 향기

"나와 내 동료들은 금으로만 나을 수 있는 마음의 병을 앓고 있기 때문이다." 스페인 사람들이 떠나온 아프로아시아 세계에는 금에 대한 직찹이 만연해 있었다. 서로 원수인 사람도 이 쓸모 없는 금속을 갈구하는 데는 한마음이었다. 

이교도인 기독교인들을 상대로 성전을 벌였던 무슬림 통치자들조차 경배의 표시로 예수와 성모 마리아를 새겨 넣은 주화로 세금을 받았다. 

----

### 10_1. 가격이 얼마인가요?

어떤 마을이 정말로 품질이 좋은 와인, 올리브오일, 도자기를 만드는 것으로 명성을 얻었다면, 그 마을은 거의 전적으로 해당 상품만을 생산하고, **필요한 다른 상품은 다른 정착지의 사람들과 교역을 통해 얻는 것이 낫다**는 사실을 알아 차렸다. 하지만 이런 전문화는 문제를 하나 일으켰다. **전문가 사이의 물품 교환을 어떻게 관리할 것인가?**

물물교환 경제에서 상대가격을 매일매일 새로 알아야한다. 만일 1백종의 각기 다른 상품 시장에서 거래된다면, 구매자와 판매자는 4,950가지 서로 다른 교환율을 알아야 할 것이다. 그걸 어떻게 다 안단 말인가?

옛 소련에서 중앙집권적 물물교환 시스템을 만들었지만 비참한 실패로 끝났다. 하지만 잉카 제국에서는 이보다 온건하고 성공적인 실험이 이루었졌다. 하지만 대부분의 사회는 많은 수의 전문가를 연결시키는 좀 더 쉬운 방법을 찾아냈다. 바로 **돈을 개발한 것이다.**

----

### 10_2. 조가비와 담배

화폐는 주화와 지폐가 아니다. **재화와 용역의 가치를 체계적으로 표현** 할 수 있게끔 사람들이 기꺼이 사용하려고 하는 모든 것을 의미한다.그 목적은 재화와 용역을 교환하는데 있다.

복잡한 상업 체계가 제대로 기능하려면 **모종의 화폐가 반드시 있어야 한다.** 화폐 경제하의 제화공은 다양한 종류에 매겨지는 가격만 알면 족하지, 신발과 사과, 신발과 염소의 교환율을 암기할 필요가 없다.

돈은 거의 모든 것을 다른 거의 모든 것으로 바꿀 수 있게 해주는 보편적인 교환수단이다. 이상적인 형태의 돈은 사람들로 하여금 어떤 것을 다른 것으로 바꾸게 해줄 뿐만 아니라 부를 축척할 수 있게도 해준다. 

부를 이용하려면 단순히 저장해두는 것만으로는 충분하지 않다. 이 장소에서 저 장소로 이동시킬 필요가 있다. 예컨데 부동산 같은 것들은 전혀 이동할 수 없다. 돈은 이런 문제를 해결해준다. 농부는 재산을 팔고 별보개고등 껍대리 한 자루를 받아서 어디든 쉽게 가지고 갈 수 있다. 이는 **복잡한 상거래망과 역동적 시장이 출현**하는데 결정적으로 기여했다. 

----

### 10_3. 돈은 어떻게 작동하는가? 

**돈은 물질적 실체가 아니라 심리적 구조물이다.** 사람들이 돈을 위해 수고하려는 것은 자신들의 집단적 상상의 산물을 믿기 때문이다. 신뢰는 온갖 유형의 돈을 주조하는 데 쓰이는 원자재이다.

달러 지폐 한 장을 꺼내어 잘 살펴보라. 그것의 앞면에는 미국 재무성 장관의 서명이 있고, 뒷 면에는 '우리는 하나님을 믿는다'는 구호가 쓰여 있는 알록달록한 종이조각에 지나지 않음을 알 수 있다. 우리가 달러는 받는 이유는 우리가 하느님을 믿고 미국 재무성을 믿기 때문이다. 

최초의 화폐가 만들어졌을 때 사람들은 이런 신뢰가 없었기 때문에 **그 자체가 내재적 가치를 지닌 물건을 '화폐'로 정의할 필요가 있었다.** 역사상 최초의 화폐로 알려진 수메르인의 **보리 화폐가 좋은 사례이다. 

가장 흔한 단위가 실라였는데 대략 1리터에 해당하는 양이었다. 한 실라를 담을 수 있는 표준화된 그릇이 대량생산되어 사람들은 물건을 사고 팔 때 필요한 양만큼의 보리를 쉽게 잴 수 잇었다. 월급역시 '보리 몇 실라'식으로 결정되고 지불되었다. 

보리가 실질적인 가치를 지닌다고 해도 이것을 그냥 하나의 생필품이 아니라 돈으로 사용하도록 확신을 주는 것은 쉽지 않았다. 그럼에도 불구하고 보리를 첫번째 유형의 화폐로 받아들이고 신뢰하는 것은 어느 정도 쉬운 일이었었는데, **그 이유는 보리에는 내제된 생물학적 가치가 있기 때문이다. 즉 사람은 그것을 먹을 수 있기 때문이다.** 한편 보리는 저장하거나 운반하기에는 어려웠다. 

화폐의 역사에 진정한 돌파구가 생긴 것은 기원잔 3000천년에서 2000년의 중간쯤 고대 메소포타미아에서 출현한 **은으로 된 세겔이었다.** 세겔은 은화가 아니라 은 8.33그램을 의미하며 20세겔은 은화 20개가 아니라 은 166그램을 말한다.

**정해진 무게의 귀금속은 결국 동전, 즉 주화를 탄생시켰다.** 역사상 최초의 주화는 기원잔 640년경 아나톨리아 서부에 있던 **리디아의 왕 알뤼아테스**가 만들었다. 이 주화는 표준화된 무게의 금이나 은으로 만들어졌고, 식별 표식이 새겨져 있었다. 이 식별표시는 2가지를 증명했다. 해당 주화에 귀금속이 얼마나 들어 있는지, 그 내용물을 보증한 당국이 누구인지를 확인시켜주었다. 오늘날의 모든 주화는 이 리디아 주화의 후손들이다. 


----

### 10_4. 금이라는 복음

로마 주화에 대한 신뢰는 매우 강력해서 국경 바깥 사람들조차 데나리우스 주화를 받았다. **기원전 1세기 로마 주화는 인도 시장에서 교환 수단으로 받아들였다.** 가장 가까운 로마땅이 수천 키로미터나 떨어져 있는데도 말이다. 

리디아 스타일의 주화 제조법이 지중해에서 인도양까지 퍼져나가고 있을 때, 중국은 이와 조금 다르게 청동 동전과 표식이 없는 은괴와 금괴를 기반으로 한 화폐 시스템을 개발했다. 하지만 두 시스템은 공통점이 많아 중국과 리디아 사이에는 금전적, 상업적 관계가 발전했다. 

사람들은 예나 지금이나 서로 알아들을 수 없는 언어로 말했고, 각기 다른 통치자의 지배를 받았고, 각기 다른 신을 숭배했지만, 모두 금과 은, 금화와 은화를 신뢰했다. 이런 공통의 신념이 없었다면 세계 무역망은 사실상 존재 할 수 없었을 것이다. 

두 지역이 일단 무역으로 연결되면, 운송가능한 물품의 가격은 **수요와 공급의 힘에 의해 평준화되는 경향이 있다.** 인도사람에게 금을 사용할 실질적인 가치가 없더라고, 지중해 사람들이 그것을 원한다는 사실만으로도 인도 사람들은 그 가치를 높게 평가하게 된다.

종교는 우리에게 무언가를 믿으라고 요구하는 반면, 돈은 다른 사람들이 뭔가를 믿는다는 사실을 믿으라고 요구하낟. 철학자와 사상가와 예언가는 수천년에 걸쳐 돈을 흉보면서 돈이 모든 악의 근원리가 매도했다. 물론 그렇기도 하지만, 한편 **돈은 인류가 지닌 관용의 정점이다.** 종교나 사회적 성별, 인종, 연령, 성적 지향을 근거로 사람을 차별하지 않는 유일한 신뢰 시스템이기도 하다. 돈 덕분에 서로 알지고 못하고 심지어 신뢰하지도 않는 사람들이 효율적으로 협력할 수 있다.


----

### 10_5. 돈의 대가

돈은 두 가지 보편적인 원리를 기반으로 하고 있다.
1. **보편적 전환성** : 돈이 있으면 당신은 마치 연금술사처럼 땅을 충섬심으로, 사법을 건강으로, 폭력을 지식으로 변환할 수 있다.
1. **보편적 신뢰** : 돈을 매개로 삼으면 임의의 두 사람은 어떤 프로젝트에도 협력할 수 있다.

**인간 공동체는 늘 명예, 충성심, 도덕, 사랑처럼 '돈으로는 살 수 없는 것'들에 대한 믿음을 기초로 삼았다.** 부모는 아이를 팔아서는 안되고, 경건한 기독교인은 대죄를 범해서는 안되고, 충성스러운 기사는 영주를 배반해서는 안되며, 조상 대대로 물려받은 부족의 땅을 낯선 사람에게 팔아서는 절대로 안된다.

부모는 자식 몇 명을 노예로 팔아 나머지 자식들에게 먹일 식량을 사야 하는 처지에 내몰렸다. 독실한 기독교인은 살인과 도둑질과 사기를 저질렀으며 그렇게 얻은 돈으로 교회에서 면죄부를 샀다. 야망에 찬 기사들은 자신의 충성심을 경매에 붙여 가장 높은 값을 부르는 사람에게 팔았으며 자신을 따르는 시종들의 충섬심도 현금 지불로써 확보했다. 부족한 땅은 지구 반대편애서 온 나천 사람에게 팔렸다. 글로벌 경제에 진입하는 티켓의 대가로서.

우리는 이방인이나 이웃집 사람을 신뢰하는 것이 아니라 그들이 지닌 주화를 신뢰할 뿐이다. 그들에게서 주화가 떨어지면 우리의 신뢰도 사라진다.

인간의 경계사는 미묘한 춤과 같다. 이방인과의 협력을 위해 돈에 의존하지만 그것이 인간적 가치와 친밀한 관계를 손상시킬까봐 두려워한다. 

어떻게 수천개의 고립된 문화가 세월이 가면서 점차 합쳐져서 오늘날 지구촌을 형성했는지를 이해하혀면, **물론 금과 은의 역할을 고려해야 하지만 강철도 이에 못지않게 중요한 역할을 했다는 사실을 간과해서는 안 된다.**
