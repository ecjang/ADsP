####  3부. 인류의 통합

# 11. 제국의 비전

기원전 134년 로마의 인내심이 바닥이 났다. 원로원은 스키피오 아이밀리아누스를 파견하기로 했다. 로마의 제일가는 장군이자 카르카도를 폐허로 만든 인물을 보내 누만시아를 손봐주기로 한 것이다. 그 장군은 누만시아를 포위하여 외부와의 접촉을 차단했고, **1년 뒤 누만시아의 식량이 바닥났다. 모든 희망이 사라졌다는 것을 깨달은 그들은 스스로 마을을 불태웠다. 로마의 기록에 의하면 사람들은 대부분 노예가 되지 않기 위해 자살했다.**

누만시아인들이 실제로 남긴 것은 폐허밖에 없다. 심지어 그들이 이야기가 지금까지 전해지는 것도 로마인 역사가들 덕분이다. **로마는 누만시아를 상대로 너무나 완벽한 승리를 거둔 나머지, 패자들이 기억마저 자기들 것으로 만들었다.**

역사에 정의란 없다. 과거의 대부분의 문화는 어떤 무자비한 제국의 군대에 의해 희생되었고, 문화를 망각 속으로 밀어넣었다. 언젠가 제국도 무너지지만 화려하고 풍성하고 지속적인 유산을 남긴다. 21세기를 사는 거의 모든 사람은 어디가 되었든 제국의 후예이다.

----

### 11_1. 제국이란 무엇인가?

제국이라는 정치질서는 두가지 중요한 특징을 지닌다. 첫째, 그런 명칭으로 불리려면 **서로 다른 문화적 정체성, 서로 다른 지역, 서로 다른 민족을 충분히 많이 지배해야 한다.** 둘째, 제국의 특징은 **탄력적인 국경과 잠재적으로 무한한 식욕이다.** 기본 구조와 정체성을 변화시키지 않은 채 갈수록 더 많은 국가와 영토를 집어삼키고 소화할 수 있다.

제국이 그 기원이나 형태, 영토의 범위, 인구의 크기에 의해서 정의되는 것이 아니라 오로지 **문화적 탄력성으로만 정의된다.** 반드시 군사적 정복으로 등장하지 않는다. 아테네 제국은 자발적 동맹으로 생명을 얻었고, 합스부르크 제국은 혼인으로 탄생해 일련의 영리한 결혼동맹에 의해 꿰맞춰졌다. 

반드시 독재적 황제가 통치하는 것도 아니다. 역사상 최대 규모였던 대영제국의 통치체계는 민주주의였다. 다른 민주적 제국으로는 근현대의 네덜란드, 프랑스 벨기에, 미국이 이다. 근대 이전에는 노브고로드, 로마, 카르타고, 에테네도 속한다.

크기 역시 문제되지 않는다. 에테네 제국은 최전성기에도 크기와 인구가 오늘날 그리스보다 작았다. 하지만 그리스 제국이 아니다.

----

### 11_2. 사악한 제국

현대에 제국에 대한 비판은 두가지가 있다. **첫번째. 제국은 제대로 작동하지 않는다.** 수많은 피정복 민족을 효과적으로 다스리는 것은 결국 불가능하다. **두번째. 그것이 가능하더라도 실행해서는 안된다.** 제국은 파괴와 착취의 사악한 엔진이기 때문이다. 모든 민족은 자결권이 있고 다른 민족의 지배를 받아서는 안된다.

기원후 476년 서로마 제국이 게르마족의 침공으로 마침내 무너졌을 때, 누만시아, 아르베르니, 헬베티아, 삼늄, 루시타이나, 움브리아, 에트누리아 종족등 **로마에 정복당했던 민족들은 제국의 시체에서 살아돌아지 못했다. 이들은 모두 사라졌다.** 스스로 그런 국가의 민족이라고 믿었고 그 나라의 언어를 썼고 그 나라의 신을 섬겼고 그 나라의 신화와 전설을 읊었던 사람들의 생물학적 후손들은 **이제 로마인처럼 생각하고 숭배했다.**

현대 유대인이 지닌 정치, 경제, 사회적 관스븐 고대 유대 왕국에서 유래한 부분이 크지 않다. 그보다는 지난 2천 년간 자신들이 그 휘하에서 살았던 제국들에서 유래한 부분이 더 크다. 만일 다윗왕이 오늘날 예루살렘의 초정통파 시나고그에 모습을 드러낸다면, 동유럽식 복장에 그르만어로 말하여 바빌로니아 문서의 의미에 대해 끊임없이 논쟁을 벌이는 사람들을 보고 그는 크게 당황할 것이다. 



---

### 11_3. 너를 위해서 하는 일이야

분명한 정보가 남아있는 최초의 제국은 **사르곤 대제(기원전 2250년경)의 아카드 제국**이다. 메소포타미아의 모든 도시국가뿐만 아니라 바깥에 있는 넓은 영역까지 정복했다. 사르곤 대제는 자신이 온 세상을 정복했다고 뽐냈다. 아카드 제국은 오래가지 못했지만 사르곤이 남긴 제국을 물려받을 주인이라고 주장하는 자가 끊임없이 나왔다. 그러다 기원전 550년경 페르시아의 키루스 대왕이 등장해 이보다 더 인상적인 자랑을 늘어놓았다.

**키루스는 전 세계를 지배한다고 주장했을 뿐 아니라 이것은 모든 사람을 위한 것이라고 주장했다.** 페르시안인들은 "우리가 너희를 정복하는 것은 너희를 위해서다."라고 말했다. **온 세상을 그들의 위해 지배한다는 것은 놀라울 정도로 주제넘은 것이었다.** 사피엔스는 본능적으로 '우리'와 '그들'의 두 부류로 나눈다. 우리란 너와 나, 언어와 종교와 관습이 같은 사람들을 말한다. 이런 인종적인 배타성과 대조적으로, 키루스 제국의 이데올로기는 모든 것을 아우르는 경향이 있었다. 인류는 하나의 대가족으로 인식되었고, 부모의 특권은 자녀의 복지에 대한 책임과 함께 존재하는 것이었다. 

이런 새로운 제국관은 알렉산드로스 대왕에게로, 그에게서 다시 고대 그리스의 왕, 로마의 황제, 무슬림의 칼리프, 인도의 세습 군주, 그리고 심지어 소련의 지도자와 미국의 대통령에게로 이어졌다. 이런 자애로운 제국관은 제국의 존재를 정당화하고, 피지배 민족의 반란 식도를 무효화했으며, 독립된 민족들이 제국의 팽창에 대항하려는 시도까지도 무효화했다.



----

### 11_4. '그들'이 '우리'가 될 때

제국은 수많은 작은 문화를 서로 융합해 몇 개의 큰 문화로 만드는데 결정적인 역할을 했다. 그런 하나의 이유는 그들 스스로가 편하기 위함이다. 표준화는 황제에게 대단히 유용했다.

두번째 이유는 **전통성을 얻기 위함이다.** 제국의 행동이 우월한 문화를 전파하기위해 꼭 필요하다고 정당화하였다. 자기네 문화가 피정복민에게 더 큰 이익이 된다고 주장했다. 황제에게 천명을 부여하는 것은 착취가 아니라 인간성을 기르치라는 의도에서 였다.

로마인들도 비슷한 주장을 폈다. 미개한 게르만족과 얼굴에 물감을 칠한 갈리아족은 천하고 무지하게 살다가 로마인 덕을 보게 되었다고 한다. 기원전 3세기의 마우리아 제국은 몽매한 세계에 부처님의 가르침을 퍼뜨리는 것을 사명으로 삼았다. 

자유주의와 자유무역이라는 두가지 복음을 퍼뜨리겠다는 영국의 사명에는 해가 지는 일이 없다. 오늘날 미국은 제3 세계에 민주주의와 인권의 혜택을 가져다 줄 의무가 있다고 주장한다. 비록 그것을 위해 순항 미사일과 F16 전투기로 배달해야 하더라도.

문화의 용광로가 패자의 문화적 동화 과정을 쉽게 만드는 것은 아니었다. **그런 혼성의 결과는 대다수 사람들한테 여전히 낯설었다. 동화의 과정은 고통스럽고 큰 정신적 충격을 동반하는 일이 많았다.** 정복과 수용 사이에 끼인 세대는 소외되고 배제되었다. 이들은 스스로 사랑했던 지역문화를 이미 잃었음에도 불구하고 제국주의 세계에 동등하게 참여할 자격을 받지 못했다. 모한다스 카람찬드 간디는 영국령 남아프리카 기차에서 끌어내려 졌다.

몇몇 경우에는 문화적 동질화 과정에서는 신참과 옛 엘리트가 뒤섰이기도 한다. 로마의 모든 백성은 몇 세기에 걸친 제국의 통치 후에는 로마 시민권을 부여받았다. 로마 출신이 아닌 사람도 로마 군단 장교단의 최상위 계급으로 승진해서 원로원 의원에 임명되었다. 

아랍의 제국 프로젝트는 대성공을 거두었다. 스스로 창조한 제국의 문화가 수많은 비아랍인에게 진심으로 받아들여진 것이다. 이들은 그 문화를 계속 유지하고 발전시키고 퍼뜨렸다. 심지어 원래의 제국이 무너지고 인종집단으로서 아랍인이 통치권을 잃어버린 다음에도.

지난 수십년간 진행되온 식민지 해방과정도 이와 비슷하게 이해할 수 있다. **근대 유럽인들은 지구의 많은 지역을 정복하면서 우월한 서구 문화를 전파한다는 것을 구실로 삼았다.** 이는 성공하여 지구상 많은 사람들이 영어, 프랑스어, 스페인어를 배웠다. 이들은 인권과 민족자결주의를 신봉하기 시작했으며, 서방의 자유주의, 자본주의, 페미니즘, 민족주의 같은 이데올로기를 받아들였다.

이런 가치를 서구의 유산이다. 오늘날 인도, 아프리카, 중국 사람들은 예쩐에 자신들을 지배했던 서구 군주의 제국 문화에서 많은 것을 받아들이고 이를 자신의 필요와 전통에 맞추어 변형시키려 노력했다.

-----

### 11_5. 역사상의 선인과 악당

제국의 정의상 나쁜 것이라면 우리는 어떻게 되는가? 세상에는 제국주의르 제거하고 죄에 더럽혀지지 않은 소위 순수하고 진정한 문명만을 남기자는 취지의 학파와 정치운동이 있다. **하지만 역사의 여명 이후에는 그런 주장을 합리적으로 펼칠 수 없을 것이다. 적어도 오늘날 존재하는 문화중에서는 없는 것이 분명하다.**

인류의 모든 문화는 적어도 부분적으로는 제국과 제국주의 문명의 유산이며, 어떤 학술적, 정치적 외과수슬을 받는다해도 환자를 죽이지 않고 제국 주의의 유산만 도려낼 수 없다.

오늘날 영국은 인도를 정복하고 점렴하는 과정에서 수백만 명의 목숨을 희생시켰다. 식민 정부는 수억 명 이상의 인도인을 지속적으로 모욕하고 착취한 책임이 있다. 하지만 많은 인도인들이 개종의 기쁨을 누리면서, 민족자결이나 인권 같은 서구의 개념들을 받아들였다. 또한 그들은 인도인들과 영국인에게 동등한 권리나 독립을 인정하지 않는 데 대해 실망했다. **그럼에도 인도라는 현대 국가는 대영제국의 자식이다.**

영국인은 인도 사법제도의 초석을 놓았으며, 행정부의 구조를 창건하고, 경제적 통합에 극히 중요한 철도망을 건설했다. 독립 인도는 영국에서 구현된 형태의 서구식 민주주의를 정부 형태로 받아들였다. 

만일 어떤 극단적 힌두 민족주의자가 있어서 뭄바이 기차역을 비롯해 영국 정복자가 남긴 모든 건물을 파괴한다면,인도의 무슬림 정복자가 남긴 타지마할 같은 건물은 어떻게 할 것인가?






----

### 11_6. 새로운 지구 제국

전 세계를 지배하는 제국이라는 환상이 실현될지도 모른다. 특정 국적을 지닌 사람이 아닌 구성원 모두가 정치권력의 합법적인 근원이며, 인간 종 전체의 이익을 지키는 것이 정치의 원칙이 되어야한다고 믿고 있다. 많일 그렇다면 2백여개의 독립국가는 도움이 아니라 방해가 될 것이다. 

어떤 주권국가도 혼자서는 지구 온난화 문제를 극복할 수 없다. 중국의 천명은 이런 인류의 문제를 해결하라고 하늘이 부여한 것이다. 현대의 천명은 하늘의 문제(오존, 온실가스)를 해결하라고 인간이 내린 것이다. 

오늘날 국가는 빠른 속도로 그 독립성을 잃어가고 있다. **어떤 국가도 독자적인 경제정책을 시행하거나 마음대로 전쟁을 선포하고 수행할 실지적인 능력이 없다.** 심지어 국내의 문제가 자기들이 좋다고 생각대로 운영할 수 도 없다. 국가는 글로벌 마켓의 책략에, 글로벌 회사와 글로벌 NGO의 간섭에, 여론의 가독에, 국제 사법제도에 점점 더 문호를 열고 있다. 

전 세계에 걸쳐 점점 더 많은 기업가, 엔지니어, 학자, 법률가, 경영인이 이 제국에 동참하려는 요청을 받고 있다. 이들은 세계 제국의 부름에 응답할 것인가, 아니면 자기 국가와 민족에 충성을 바치며 남을 것인지 심사숙고 해야 한다. 그리고 점점 더 많은 사람들이 제국을 선택하고 있다.
