####  4부. 과학혁명

# 14. 무지의 발견

지난 5백 년간 인간의 힘은 경의적으로 유래가 없을 만큼 커졌다. 1500년 지구 전체에 살고 있는 호모 사피엔스는 5억 명인 반면 오늘날에는 70억 명인 산다. 

| 시기 | 인구 | 재화 가치 | 하루 소비 에너지 | 
| ---- | ---- | --------- | ------------- | 
| 1500년 | 5억명 | 총 2,500억 달러 | 약 13조 칼로리 |
| 현대 | 70억명  | 연간 60조 달러 | 1,500조 칼로리 |


현대의 전함 한 대가 콜롬버스 시대로 옮겨졌다고 상상해보자. 이 배는 니냐, 핀타, 산타마리아 호를 몇 초 만에 널판지 조각으로 만 들 수 있고, 당시 열강들의 모든 군함을 격침시키면서도긁힌 자국 하나 없을 것이다. 현대의 화물선 5척이면 당시 세계의 모든 상단이 실어 나르는 짐을 실을 수 있다. 현대의 컴퓨터 한 대면 중세의 모든 도서관에 있는 사본과 두루마리에 있는 단어와 숫자들을 저장하고도 공간이 넉넉히 남았을 것이다. 

역사의 대부분 기간동안, 인상은 지구상의 생명체 중 약 99.99%에 대해 아무것도 몰랐다. 미생물말이다. 인간의 눈이 미생물을 처음 본 것은 1674년이 되어서였다. 안톤 판 레이우엔훅이 집에서 만든 현미경으로 엿본 세계는 놀라웠다. 그 후 3백 년간 인류는 현미경으로만 보이는 엄청난 숫자의 미생물을 알게 되었다. 

하지만 지난 5백 년간 가장 눈에 띄는 단 하나의 결정적 순간은 **1945년 7월 16일 오전 5시 29분 45초**였다. 정확히 그 때 미국 과학자들은 앨러머고도 사막에 첫 원자폭탄을 터뜨렸다. 그 순간 이후 인류는 역사의 진로를 변화시킬 능력뿐 아니라 역사를 끝장낼 능력도 가지게 되었다.

**왜 과학혁명은 혁명일까? 약 1500년 이전까지 새로운 의학적, 군사적, 경제적 힘을 얻을 능력이 있는지 의심했기 때문이다.** 당시 교육과 학문에 대한 투자는 새로운 것이 아닌 이미 존재하는 능력을 얻기 위함이었다. 사제와 철학자, 시인들한테 돈을 주면서 사회 질서를 유지하려고 했지, 새로운 의약품이나 신무기, 혹은 경제성장을 촉진시키라고 하지않았다.

하지만 지금은 부자와 정부는 과학에 더 많은 투자를 하며 스스로의 능력을 증가시킬 수 있다고 믿게 되었다. 그런 투자덕분에 미생물을 조작하거나 원자를 쪼갤 수 있었다. 미국 정부는 핵물리학에 수 십년동안 수십억 달러를 배정했다. 그 연구로 인해 원자력 발전소를 세울 수 있었고 미국 산업에 싼값으로 전력을 공급했으며, 그 산업으로 세금이 커지고 이 세금이 핵물리학을 더욱 깊이 연구할 연구 자금으로 이어졌다.

### 14_1. 우리는 모른다

현대과학은 과거의 모든 전통 지식과 3가지가 결정적으로 다르다.

1. **무지를 기꺼이 인정하기**<br>
현대 과학은 라틴어로 표현하면 '이그노라무스(Ignoramus)'이며 '우리는 모른다'라는 의미이다. 알고 있다고 생각하는 것이 더 많은 지식을 가지게 되면 틀린 것일 수도 있다는 것을 받아들이는 것이다. <br><br>

2. **과찰과 수학이 중심적**<br>
현대 과학은 수학적 도구로 관찰을 연결해 포괄적인 이론을 만들어낸다.<br><br>
    
3. **새 힘의 획득**<br>
이른을 창조하는 것에 그치지 않고, 그 이론을 사용해서 새로운 힘을 얻거나 기술을 개발하려고 한다.

과학혁명은 **지식의 혁명이 아닌 무지의 혁명이다.** 근대 전통적인 이슬람이나 기독교, 불교, 유교는 세상에 알아야하 모든 것은 이미 알려져 있다고 단언했다. 평범한 사람들은 그런 문헌과 전통에서 지식을 얻었다. 이런 전통적인 지식은 오직 두가지의 무지만을 인정했다.

1. **한 개인은 뭔가 중요한 것에 대해 무지할 수 있다.**<br>
필요한 지식을 얻으려면 현명한 자한테 묻기만 하면 되었다. 동네의 한 농부가 인류의 기원에 대해 궁금해 했다면 그 지방의 사제한테 물어보는 것이 전부였다.<br><br>
    
1. **뭔가 중요하지 않은 것에 대해 무지할 수 있다.**<br>
위대한 신들과 과거의 현자들이 우리에게 애써 말해주지 않는 것은 그게 무엇이든 중요하지 않았다. 거미가 어떻게 거미줄을 치는지 궁금하다고 사제한테 물어봐도 무의미했다.


이것은 꽤나 심각한 문제이다. 이런 종교와 신화덕분에 서로 모르는 수백만의 사람들이 협력할 수 있는데, 이런 신화의 많은 부분에 의심이 든다면 공동체, 국가, 국제 시스템은 어떻게 기능할 수 있을까? 정치사회적 질서를 유지하기 2가지의 비과학적인 방법에 의지할 수 밖에 없었다.

1. **하나의 과확이론을 택하여 궁극적인 진리로 선포하는 것.**<br>
이것은 나치당원과 공산주의자들이 사용한 방법이다. 그들의 인종정책이 생물학적 사실들의 필연적인 귀결이라고 주장했다. 그리고 공산주의자들과 레닌의 경제적 진리는 절대적이고 신성한 것으로 결코 반박이 불가능하다고 주장했다.
<br><br>

1. **과학과 무관한 절대진리에 따르는 것**<br>
이것은 자유주의적 인본주의의 전략이다. 인간 고유의 가치와 권리에 대한 도그마적 신조를 토대로 건설된 이념이다. 하지만 그 신조는 호모 사피엔스의 연구 결과와는 전혀 공통점이 없다. 

현대의 사회질서를 지탱해준 요인 중 하나는 기술과 과학적 연구 방법에 대한 종교적인 믿음의 확산이었다. 이것은 절대진리에 대한 믿음을 어느정도 대체해 주었다.

### 14_2. 과학의 도그마

현대인들은 중요한 몇몇 질문에 대한 대답을 모른다는 사실을 인정하였다. 그러면서 완전히 새로운 지식을 찾아볼 필요가 있다는 것을 꺠달았다. 옛 전통을 연구하기 보다는 새로운 관찰과 실험을 하는 쪽으로 치중하고 있다.

**하지만 더 많은 관찰이 곧 더 많은 지식은 아니다.** 우주를 이해하기 위해서는 관잘들을 연결하여 포괄적인 이론을 만들 필요가 있다. 과거 코란, 베다, 유교의 경전에는 방정식, 그래프, 계산이 거의 없다. 수식이 아닌 이야기 형태로 전통적 신화를 구술했다. 미니교의 예언자는 세상을 선과 악의 전쟁터라고 이야기하지만 "사람에게 작용하는 힘은 그의 정신의 정신의 가속도를 신체 질량으로 나눈 값과 같다"는 식으로 이야기 하지 않는다.

**1687년 아이작 뉴턴은 현대 역사에서 가장 중요한 [자연철학의 수학적 원리]를 출간했다.** 떨어지는 사과부터 별똥별에 이르기까지 우주의 모든 물체의 운동을 설명할 수 있었다. 뉴턴은 자연이라는 책이 수학의 언어로 쓰여 있음을 보여주었다. 하지만 생물학, 경제학, 심리학을 깔끔한 뉴턴의 방정식으로 환원하려고 시도했던 학자들은 실패했다. 지난 2백년간 이런 복작합 측면을 다루기 위한 새로운 수학 분과가 개발되었다.

**1744년 스코틀랜드 장로교 목사인 알렉산더 웹스터와 로버트 월리스는 생명보험기금을 만들어 사망한 목사의 미망인과 고아에게 연금을 지급하기로 했다.** 그러기 위해서는 목사의 사망수, 미망인과 고아의 수, 미망인이 수명 등을 예측해야 했다. 이들은 답을 알기 위해 하느님께 기도하지 않았다. 스코틀랜드인 답게 실용적이었다. 에든버러 대학의 수학 교수인 콜린 매클로린과 만나 수학을 사용하여 이를 계산하였다.

이들은 통계와 확률을 사용하였고 그 중 하나가 야코프 베르누의의 '큰 수의 법칙'이었다. 단일사건을 예측하기 힘들지만 수많은 사건의 평균은 비교적 정확하게 예측할 수 있다는 것이다. 즉 자료만 충분하다면 내년에 얼마나 많은 목사가 죽을지 예측할 수 있다는 것이다. 에드먼드 핼리가 작성한 생명표의 1,238건의 출생기록, 1174건의 사랑 기록을 토대로 50세인 사람이 사망할 확률은 39분의 1이라는 것을 알아내었다. 이 수치를 가공하여 연평균 27명의 목사가 사망할 것이며 18명의 미망인이 생길 것이라고 예측하였다.**이 계산은 놀라울 만큼 정확하였다.**

오늘날 웹스터와 얼리스의 기금은 간단히 **스코틀랜드 미망인**이라는 이름으로 알려져 이는데, 세계 최대의 연금 및 보험회사로 꼽힌다. 이런 확률계산은 연금과 보험산업의 핵심이 되는 보험통계학뿐 아니라 인구통계학의 기초가 되었다. 특히 **인구통계학은 찰스 다윈이 세운 진화론의 초석이 되었다.**


오늘날 점점 많은 학생들이 수학을 배우도록 강요받는다. 정밀과학을 향하는 흐름은 거부할 수 없는 대세가 되었다. 인간 언어 및 심리 연구는 인문학 중심에서 점점 더 수학에 의존하고 있다. 이제 통계학은 물리학이나 생물학을 넘어 사회학, 경제학, 정치학의 필수 과목이 되었다.

### 14_3. 아는 것이 힘이다

70억 인구 중에 양자역학, 세포생물학, 미시경제학을 이해하는 사람은 드물지만 과학은 막대한 특권을 누리고 있다. **1620년 프랜시스 베이컨은 [신기관 The New Instrument]이라는 과학 선언문을 출간했다. 이 책에거 그는 '아는 것이 힘'이라고 주장했다.** 진정한 시금석은 유용성이다. 우리에게 새로운 일을 할 수 있는 능력을 주는 이론은 지식이다.

실제로 과학과 기술인 관련을 맺은 것은 매우 최근에 일어난 현상이다. 1500년 이전에 과학과 기술은 별개의 분야였다. 17세기 베이컨이 양자를 연결시킨 것은 혁명적인 아이디어였다. 이 규칙에 전혀 예외가 없는 것은 아니다. 근대 이전의 지배자들은 새로운 기술을 개발하기 위해 우주의 속성에 대한 연구를 지원하지 않았다. 마찬가지로 교육기관에 자금을 댔지만, 기존의 질서를 유지하려는 목적이었다.

사람들은 여기저기서 새로운 기술을 개발했지만 이는 시행착오를 통해 만들어낸 것이지, 체계적인 과학적인 연구를 통해 나온 것은 아니었다. 현대 국가들은 에너지, 공중보건 심지어 쓰래기 처리에도 과학자를 필요로 한다. 1차 세계대전 당시 끝없는 전쟁을 끝내기 위해 과학자를 불러들여 끊임없이 새로운 무기들을 만들었다. 전투기, 독가스, 탱크, 잠수함등은 물론 성능을 개선하여 기관총이나 대포, 소총, 폭탄등이 개발되었다. 

과학은 2차 세계대전에서는 더 큰 역할을 했다. 1944년 후반 독일은 패색이 짙었지만 계속 싸웠다. 이들은 자국의 과학자들한테 마지막 기대를 걸었다. V2로켓이나 제트기 같은 기적적인 무기를 만들어 대세를 바꿀 수 있다고 생각한 것이다. 동시에 미국의 맨해튼 프로젝트는 원자폭탄 개발에 성공했다. 이 폭탄은 결국 독일의 항복으로 1946년 일본으로 떨어졌다.

과학은 공격무기 뿐만 아닌 방어에도 중요한 역할을 하였다. 오늘날 많은 미국인들은 테러리즘의 해결책은 정치가 아닌 기술에 있다고 믿는다. 나노기술과 생체공학적인 스파이 파리에 수백만 달러를 투자한다. 뇌 과학에 더 많이 투자한다면 모든 공항에 fMRI 스캐너가 설치되어 사람들의 본노와 증오를 스캔할 수 있을지도 모른다.

**19세기까지만 해도 군사분야의 혁명은 기술의 변화가 아닌 조직적 변화의 산물이었다.** 아랍인들이 사산 제국을 무너뜨린 것은 우월한 칼과 활 덕분이 아니었고, 셀주크 사람들이 바진틴 사람들에 비해 기술적 위의를 가지지 않았으며, 몽골이 중국을 지배한 것은 독창적인 신무기의 도움을 받은 것은 아니었다. 사실 이 모든 경우에서 군사기술과 민간기술의 우월한 것은 오히려 패배자 쪽이었다. 

당시 최정예였던 로마군이 특히 좋은 예다. 기술적으로 다른 국가들에 비해 나을 것이 없었지만 효율적인 조직, 강철같은 규율, 막대한 예비인력이 있었다. 하지만 연구개발 부서를 만든 일이 없었으며 이들의 무기는 몇 세기동안 똑같았다. 

고대의 중국도 마찬가지다. 장군과 철학자 대부분은 시무기 대발을 자신의 임무로 생각하지 않았다. 중국 역사상 가장 위대한 발명은 화약이지만 이는 생명의 영약을 찾는 도교 연금술사에 의해 우연히 발명된 것이다. 세계의 주인이 될 수 있는 막강한 힘을 발명했지만 그 마저도 폭죽으로 사용하였다. 

나폴레옹은 포병이었음에도 신무기에 관심이 거의 없었다. 과학자들과 발명가들의 비행기계, 잠수함, 로켓을 개발할 자금을 달라고 그를 설득했지만 먹히지 않았다. **과학과 산업과 군사기술은 자보주의 체제와 산업혁명이 등장하면서 본격적으로 서로 얽히기 시작했고, 일단 그 관계가 정립되자 세상은 급속도로 변했다.**

많은 문화권에서 번개는 분노한 신이 죄인을 처벌하기 위한 망치로 여겨졌다. 18세기 중반, **벤자민 프랭클린이 번개는 단지 전류에 불과하다는 것을 증명하기 위해 폭풍속에서 연을 띄웠다.** 결국 피뢰침을 발명하고 신들을 무장해제시켰다. 

가난도 저절한 사례이다. 예수는 은화 3백 데나리우스의 비싼 향료를 축성한 여인을 옹하며 말했다. "가난한 자들은 너희와 함께 있으니 아무 때라도 원하는 대로 도울 수 있거니와 나는 너희와 함께 있지 아니하리라". 식량과 집이 없어 삶을 위협받는 생물학적인 가난은 많은 국가에서 옛말이 되었다. 오늘날 세계 대부분의 사람들의 발 밑에는 안전망이 쳐져있다. 보험, 국가의 사회보장, 국제적 NGO들이 가난으로 보호하고 있어 최악의 사태를 피할 수 있게 되었다.

### 14_5. 길가매시 프로젝트

우루크의 왕 길가매시는 세상에서 가장 힘세고 유능한 남자로, 전쟁에 나가면 그 누구인든 승리를 거두었다. 하지만 그의 오랜 친구 엔키두가 죽었을 때, 그의 시신 옆에 앉아 사색에 잠겨 있는 사이 시신의 콧구멍에서 벌래 한 마리가 떨어지는 것을 보았다. 그 순간 길가매시는 공포에 사로잡혔고 자신은 결코 죽지 않겠다고 다짐한다. 그는 죽음을 물리치기 위해 우주의 끝을 향한 여행을 떠난다. 하지만 그의 원정은 실패한다. 그는 죽음을 받아들여야하는 존재로써 빈손으로 고향에 돌아온다. 하지만 이 원정에서 그는 한 가지를 깨달았는데 바로 신들이 인간을 창조할 때 죽음을 필연적인 숙명으로 정했으며 인간은 그 숙명과 함께 사는 법을 배워야 한다는 것이다.

전세계에서 약 25~40세이던 평균 수명은 약 67세로 성큼 뛰었다. 선진국은 약 80세가 되었다. 특히 어린이와 유아 사망률이 특히 낮아졌다. 20세기 이전의 농경사회헤서는 3분의 1에서 4분의 1이 성인이 되기 전에 사망했다. 오늘날 영구에서 출생 첫해에 사망하는 아기는 1천명단 5명, 15세가 되지 전에 죽는 아이는 1천 명당 7명에 불과하다.

### 14_6. 과학을 보살피는 다정한 아빠

**과학은 여타 인간활동보다 상위에 있는 고도의 도덕적, 정신적 차원에서 벌어지는 사업이 아니다.** 우리 문화와 마찬가지로 과학은 결제적, 정치적, 종교적 이해관계에 의해 형성된다. 

**과학은 돈이 매우 많이 든다.** 인간의 면역계를 이해하려고 연구하는 생물학자한테는 실험실, 시험관, 전자현미경, 실험물질이 필요하다.  거기다 실험실 조수, 전기기술자, 배관공, 청소부는 말할 것도 없다. 이 모든 것에는 돈이 든다. 지난 5백년간 현대 과학이 놀라운 업적은 정부, 재단, 기업, 민간 기부자들이 과학 연구에 기꺼이 수십억 달러를 투자한 덕분이다. 

천재들이 아니더라도 뛰어난 통찰은 누군가의 머리속에서 떠오를 수 있지만, **적절한 자금지원이 이루어지지 않으면 아무리 뛰어난 지성으로도 그것을 보충할 수 없다.** 만일 다윈이 태어나지 않았다면 앨프리드 러셀 월리스한테 진화론의 공이 돌아갔을 것이다. 하지만 유럽 열강들이 지리학적, 동물학적, 식물학적 연구에 자금을 대지 않았다면 다윈도 윌리스도 진화론을 뒷바침할 실증적 자료를 손에 넣지 못했을 것이다. 어쩌면 그런 시도조차 하지 못했을 가능성이 크다.

대부분의 과학연구는 모정의 정치적, 경제적, 종교적 목적을 달성하는데 도움이 된다고 믿기 때문에 자금이 지원되는 것이다. 예를 들어 16세기의 왕과 은행가들은 세계를 누비는 지리적 탐험대에 막대한 자원을 투입했지만 아동심리학 연구에는 한푼도 대지 않았다. 1960년대 미국과 소련 정부가 수중고고학이 아니라 핵물리학 연구에 막대한 지원을 한 것은 원자폭탄을 기대할 수 있지만, 수중고고학은 전쟁에서 승리하는데 큰 도움이 될 것 같지 않았다.

**우리의 자원은 한정되어 있다. 정치, 경제, 종교적 이해관계와 무관한 순수과학연구를 지원하고 싶다고 해도 불가능할 것 이다.** 기초 연구를 위해 국립과학재단에 1백만 달러를 요청한다면 의원은 그 돈으로 교사 교육이나 어려운 사람들을 위한 복지에 사용하는 것이 더 좋지 않느냐고 반문할 것이다. 제한된 자원을 끌어오려면 '무엇이 더 중요한가?', '무엇이 더 좋은가?'를 고민해야 한다. 그러고 이런 것은 과학저긴 질문이 아니다. 이런 질문은 종교와 이데올로기에 관련된 것이다.

**과학은 자신의 우선순위를 스스로 정할 수 없다. 자신이 발견한 것으로 무엇을 할지 결정할 능력도 없다.** 발전한 유전학적 지식을 가지고 우리는 무엇을 해야 할까? 암 치료에 사용하거나 유전자 조작으로 슈퍼맨을 만들거나, 아니면 슈퍼 사이트 젖통이 달린 유전자 조작 젖소를 만드는데 써야할까? 자유주의 정부, 공산 정부, 나치 정부, 자본주의 기업은 동일한 과학적 발견을 완전히 다른 용도로 사용할 것이다.

**과학은 모종의 종교나 이데올로기와 제휴했을 때만 번성할 수 있다. 이데올로기는 연구비를 정당화한다.**  그 대신 이데올로기는 과하적 의제에 영향을 미치고, 과학의 발견을 어떻게 사용할지를 결정한다. 그러므로 인류가 어떻게 해서 달에 도달했는지를 이해하려면 심리학자, 생물학자, 사회하자의 업적을 조사하는 것만으로는 충분하지 않다. 이데올로기적, 정치적, 경제적인 힘을 고려해야 한다.
