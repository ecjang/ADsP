# 15. 과학과 제국의 결혼

'자연지식의 개선을 위한 런던 왕립협회'는  북구 캐나다와 캘리포니아까지 과학자를 파견한 것으로는 충분하지 않다 머나먼 남서 태평양까지도 천문학자를 보내야 한다고 생각했다. 그 결과 찰스 그린을 일곱가지 다른 분야의 과학자들과 함께 타히티로 보내기로 한다. **식물학자 조지프 뱅크스**가 팀을 이끌고 **뱃사람이면 뛰어난 지리학자이자 민족지학자, 그리고 해군 장교인 제임스 쿡** 선장의 지휘를 받아 먼 여행을 떠난다.

당시 괴혈병의 원인가 해결책을 찾지 못하여 장거리 항해는 불가능하였지만 쿡 선장은 당시의 민간요법인 감귤이 괴혈병을 치료할 수 있다고 믿었기 때문에 채소와 과일로 괴혈병을 극복, 먼거리 항해를 할 수 있었다.

이 탐험 비용의 대부분은 왕립협회가 댔지만 선박은 영국 해군의 함선이었다. 85명의 무장 선원과 대포, 머스킷 총, 화약 등의 무기가 실렸으며 탐사대가 수집한 지리학, 기상한, 인류학 등의 데이터는 군사적인 목적이 뚜렸했다. 쿡은 자신이 발견한 섬들에 대한 영국의 소유권을 주장했는데 대표적인 곳이 바로 호주이다. **그곳에 정착하며 그곳의 토착문화를 파괴하고 원주민 대부분을 박멸할 기초를 닦아주었다.**

쿡의 배는 군대의 보호를 받는 과학적 탐사대였을까? 아니면 소수의 과학자가 따라붙은 군사원정대였을까? 이것은 연료통이 반쯤 찼는지 반쯤 비었는지 묻는 것과 다릅없다. 둘다 해당한다.

### 15_1. 어째서 유럽인가

영국 제도와 서유럽 전반은 지중에 세계에서 멀리 떨어진 벽지에 지나지 않았다. 근대 이전의 로마제국은 북아프리카, 발칸, 중동 지방에서 부를 얻었다. 로마에게 서유럽은 초라하고 황량한 서부에 지나지 않았다. 유럽이 미 대륙을 정복하고 바다의 패권을 획득한 것은 주로 아시아의 강대국들이 그런 지역에 관심을 보이지 않았기 때문이다. 당시 유럽은 다른 강대국에 비해 경제적 난쟁이였다.

세계의 권력이 유럽으로 이동한 것은 1750년에서 1850년 사이에 이르러서이다. 1900년에는 유럽은 세계 경제와 대부분의 땅을 점령했다.  1950년 서유럽과 미국을 합친 생산량은 세계 전체 생산량의 절반을 넘었고, 중국이 차지하는 비율은 5%로 축소되었다. 오늘날에는 유럽식의 질서와 문화가 세계에 영향을 미치고 있다.

어째서 유라시아 변방에 있던 이들은 그 오지에서 뛰쳐나와 전 세계를 정복할 수 있었을까? 보통은 그 공이 유럽의 과학자들한테 있다고 한다. 근대 이후 많은 나라가 혁신을 꿈꾸며 무기, 의학, 기술의 발전에 지원하였다. 통조림은 병사들을 먹여 살렸고, 철도와 증기선은 군대와 장비를 수송하였으며, 의약품은 많은 병사들을 치료했다. 병참 부분에서의 이 같은 진보는 유럽인이 아프리카 정복에 기관총보다 더욱 중요한 역할을 하였다. 

하지만 1850년 이전만 하더로 이렇지 않았다 과학혁명은 여물지 않았으며 유럽과 아시아와 아프리카 강대국들간의 격차는 크지 않았다. 그런데 왜 왕롱안 선장이나 후세인 파샤 같은 선장은 호주를 정복하지 못했을까? 만일 1770년 유럽인이 무슬림, 인도인, 중국인 보다 기술적 우위를 점하지 못했다면 그렇게나 큰 격차를 낼 수 있었을까?

중국인과 페르시안인에게 부족했던 것은 증기기관 같은 기술적 발명이 아니었다. **이들에게는 서구에서 여러세기에 결쳐 형성되고 성숙한 가치, 신화, 사법기구, 사회정치적 구조였다. 이런 것들을 빠르게 복사하거나 내면화할 수 없었다.**

1500~1850년을 보면 유럽은 독창적인 잠재력을 구축했고, 1850년이 되자 그 중요성은 뚜렷해졌다. 나무와 진흙 벽돌을 사용하는 사람과 강철과 콘크리트를 사용하는 사람에 비교해 볼 수 있다. 두 사람이 건물을 짖는다면 처음에는 큰 차이가 없지만 결정적인 문턱을 지나면 나무와 진흙은 무게를 버티지 못하고 무너진다. 이런 **강철과 콘크리트가 바로 현대 과학과 자본주의이다.** 유럽은 과학적이고 자본주의적인 방식으로 사고하고 행독하는 습관이 있었다.

### 15_2. 정복의 사고방식

극동과 이슬람 세계에서도 유럽 못지않게 지적인 호기심이 많은 사람들도 있었지만, 1500년~1950년 사이 이들은 뉴턴의 물리학이나 다윈의 생물학 비슷한 것 조차 만들어내지 못했다. 이것은 유럽인이 물리학과 생물학을 영원히 지배할 것이라는 것이라는 의미가 아니다. **현대과학은 유럽인의 전문 분야로 시작했지만 오늘날 다민족의 사업이 되고 있다.** 무엇이 현대 과학과 제국주의 사이의 연대를 구축했을까? 과학자와 정복자 둘 다 무지를 인정하는 것에서 출발했다. 이들은 "저 밖에 무엇이 있는지 모른다"고 생각했고 밖으로 나가서 새로운 것들을 발견해야 한다고 생각했다. 

로마인, 몽골인, 아즈텍인들이 새로운 땅을 정복한 것은 권력과 부를 찾아서였다. 이와 대조적으로 유럽 제국주의자들은 새 영토와 새로운 지식을 찾기 위해 포르투갈의 엔히크 왕자와 바스코 다 가마는 아프리카를, 크리스토퍼 콜럼버스는 아메리카를, 페르디난도 마젤란은 세계일주를 하였다. 

19세기에는 이런 항해에 과학적 지식을 발견을 찾기 위해 과학자가 추가된다. 찰스 다윈은 칼라파고스 섬의 지도를 작성하는 일이 계기가 되어 실증적인 자료를 수집하고 통창력을 형성했으며 종국에는 진화론을 꽃피웠다.

### 15_3. 비어있는 지도

근대 이전의 지도에는 낯선 지역을 빼버리거나 상상속의 괴물로 채워넣는 경우가 많았다. 이런 지도는 빈 공간이 없었다. **그러다 15~16세기 유럽인인들은 빈 공간이 많은 세계지도를 그리기 시작했다.** 1492년 10월 12일 오전 2시쯤 콜롬버스의 탐사대는 미지의 대륙과 맞닥뜨렸다. 오늘날 바하마라고 부르는 곳이지만 콜롬버스는 거기가 동아시아 연안의 인도의 작은 섬이라고 생각했다. 그래서 그곳의 사람들을 인디언이라고 불렀다. 그런데 곰곰히 생각해보면 **어떻게 위대한 사상가는 물론, 오류가 있을 수 없는 성경까지도 수천 년 동안 유럽과 아프리카와 아시아 밖에 모를 수 있었을까?** 이것은 마치 1969년 달을 항해 비행하던 아폴로가 그때까지 전혀 관측되지 못한 새로운 달과 충돌했다는 것과 마찬가지였다.

최초의 근대인은 아메리고 베스푸치였다. 1502~1054년 자신의 탐험의 내용을 담은 두 건의 문서가 유럽에 출간된다. 그 문서에 따르면 콜럼버스 새로 발견한 섬은 동아시아 연안의 섬들이 아닌 완전히 새로운 대륙이었다. 이런 주장에 영향을 받은 지도 제작자 마르틴 발트제뭘러는 최신판 세계지도를 출간하는데 신대륙을 표시한 최초의 지도였다. 

**아메리카 대륙의 발견은 과학혁명의 기초가 되는 사건이었다. 이는 유럽인들에게 과거의 전통보다 관찰 결과를 더 선호하라고 가르쳐주었다.** 방대한 새 영토를 통제하기 위해서는 지리, 기후, 식물, 동물, 언어, 문화 많은 부분의 정보를 수집해야 했지만 기독교 성경이나 옛 지리서, 고대 구비 전통은 도움이 되지 않았다.

비어있는 지도는 마법처럼 유럽인을 이끌었고 15~16세기 동안 유럽 탐험대는 아프리카를 일주하고 아메리카를 답사했으며, 태평양과 인도양을 횡단하고, 세계 전역을 지배하였다. 


그런데 과거에는 이러한 일이 전혀 없었다. 정거리 정복 전쟁을 일으킨다는 것은 자연스럽지 않은 일이었다. 로마인들은 로마에서 런던까지 넓혀가는데 4백년이 걸렸다. 배를 타고 바로 브리튼으로 가서 점령할 생각을 하지 않았다. 

**많은 학자들은 중국 명 왕조의 정화 제독이 벌인 대항해가 유럽인의 발견의 항해를 예고했으며 그 못지않은 성취였다고 주장한다.** 1405년~1433년 까지 일곱차례에 결쳐 대함대를 이끌고 중국에서 인동양까지 먼 항해를 하였다. 가장 규모가 컸던 함대는 3백척에 가까운 배에 3만명이 탑승했다. 함대는 인도네시아, 스리랑카, 인도, 페르시아 만, 홍해, 동아프리카를 방문했다. 

하지만 둘 사이에는 큰 차이가 존재한다. 정화 제독은 자신이 방문한 나라를 정복하거나 식민지로 삼으려고 하지 않았다. 심지어 1430년대 파벌의 바뀌자 대함대는 해체되었으며 이후 그 같은 위상과 자산을 지닌 탐험대가 항해를 떠난 일이 없다. **정화 제독의 원정은 유럽이 뛰어난 기술적 우위를 누리지 못했다는 것을 증명한다. 유럽인의 이례적인 점은 탐험과 정복의 야망이 탐욕스러웠다는 데 있다.** 근대 초기의 유럽인들은 열병에 걸린것이다. 그 병은 해변에 한 발을 디딘 뒤 "이 땅은 모두 우리 왕의 것이다!"라고 선언하게끔 만들었다.
