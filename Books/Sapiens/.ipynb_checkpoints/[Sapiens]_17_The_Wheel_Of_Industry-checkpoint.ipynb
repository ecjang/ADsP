####  4부. 과학혁명

# 17. 산업의 바퀴

경제 성장에는 에너지와 원자재가 필요한데 이는 유한하다. 만일 이것들이 고갈된다면 경제는 붕괴할 것이다. 하지만 과거를 돌이켜 보면 그런 위험이 있을 때마다 과학적, 기술적 연구에 투자하였고 더 효과적인 방법이나 새로운 에너지 원을 찾아내었다. 

운송수단을 생각해보자. 누군가는 오랜 시간 동안 엄청난게 많이 만들어진 기차, 자동차, 제트기, 우주왕복선 때문에 에너지원과 원자재가 고갈될 것이라고 생각했을지도 모른다. 하지만 현실은 그 반대다. 1700년전 지구의 운송수단은 거의 대부분 나무나 철에 의존하는 반면 현재는 플라스틱, 고무, 알루미늄, 티타늄 등 조상들이 듣도 보도 못한 재료들을 사용한다. 그리고 오늘날의 도요타와 보잉사의 기계들은 내연기관과 핵발전소로부터 동력을 얻는다.

**과거 중요한 동력중 하나는 바로 인간의 힘이었다.** 인간과 동물의 몸은 음식을 먹고 그 영양분을 태워 에너지를 얻는다. 근력은 그 당시 유일한 에너지 전환 장치였다. 자연을 이용하에 에너지를 얻는 경우도 있었지만 바람은 필요할 때만 부는 것이 아니고, 주변에 강이 없으면 수력으로 동력을 얻을 수 없었다.  반면 근력은 언제 어디서든 활용할 수 있었다. 필요할 때 황소로 밭을 갈 수 있었고, 말은 물건을 운송할 수 있었다.

**그 결과 인류의 역사는 두 가지의 주요 주지의 지배를 받았다. 식물의 성장 주기와 태양에니지의 변화 주기(낮과 밤, 여름과 겨울)이다.** 햇빛이 충분치 않고 곡식이 아직 초록색일 때 창고는 비어있고 왕들은 평화적인 통치를 하는 경향을 보였다. 반면 태양을 충분히 받아 곡식이 잘 여물면 창고는 가득 채워지고, 세금 징수원은 바빠지고 병사들은 힘을 과시하며 왕은 전투계획을 세우기 시작한다.

### 17_1. 부엌의 비밀

9세기 중국에서 화약이 발명된 이후, **열을 운동으로 전화하는데 부분적인 돌파구가 열리기 시작했다.** 물이 끓는 주전자에서 뚜껑이 튀어 오르는 것을 시작으로 결국 대포까지 발명하게 된다. 아마 어느 화약 전문가가 막자사발에 화약을 넣고 갈다가 터져서 막자가 포탄처럼 튀어오른 것을 발견한 다음의 일일 것이다. 

1700년경 영국의 광산 갱도에서 이상한 소음이 울려 퍼진다. 소리는 해가 살수혹 점점 커졌고 마침내 전 세계에서 들릴 정도로 큰 소음이 났다. 그것은 증기기관에서 나는 소리였다. 증기기관의 피스톤은 갱도 바닥에서 물을 퍼 올렸는데 처음에는 무척 비율적이었지만 영국의 기업가들이 개선을 거듭하였고, 나중에서는 이 증기기관으로 섬유를 생산하려는 생각에까지 이른다. 눈 깜짝할 사이에 영국의 세계 최대의 섬유 생산 공장이 되었다. 

여기서 중요한 것은 증기기관을 갱도에서 끄집어 내면서 심리적 장벽이 깨졌다는 사실이다. 직조기를 위해 석탄을 땔 수 있다면 다른 것들도 움직일 수 있지 않을까하는 호기심이었다. **1825년 한 영국인 엔지니어는 석탄 운반차량에 증기기관을 연결한다. 이윽고 증기기관은 석탄을 20km 떨어진 항구로 끌고 갔다. 이것이 역사상 첫 증기기관차였다.**

이 때 사람들은 기계와 엔진이 한 유형의 에너지를 다른 유형의 에너지로 변환시킬 수 있다고 생각하였다. 이후 수 많은 연구과 실험이 이루어지게 되었고 **아인슈타인이 '모든 종류의 질량은 에너지로 전환될 수 있다'라고 한 뒤 40년 만에 원자폭탄은 히로시마와 나가사키를 날려버렸고 햇 발전소는 전 세계적으로 퍼져나갔다.**

또 다른 중요한 발명은 **내연기관**으로 석유를 액체정치로 바뀌어놓았다. 석유는 불과 한 세기 전만해도 방수처리나 회전축을 매끄럽게 하는 용도로만 사용되었다. 그리고 **전기**는 이보다 더욱 놀랍다. 2 세기전에는 아무런 역할을 하지 못했지만 일련의 발명이 이어지자 도처에 존재하는 램프속의 거인이 되었다. 전기가 어떻게 작동하는지 아는 사람은 거의 없지만 전기기가 없는 삶을 상상할 수 있는 사람은 더더욱 없을 것이다.

### 17_2. 에너지의 바다

산업혁명의 핵심은 바로 **에너지 전환**이었다. 산업의 발전마다 새로운 에너지원이 발견되고 기존 에너지의 효율을 높이는 등 에너지의 총량은 증가했지만 그럼에도 불구하고 에너지 고갈을 걱정하는 사람이 많은 것은 무엇 때문일까? 세상에는 에너지 결핍이 존재하지 않는다. 부족한 것은 에너지를 찾아내 우리의 필요에 맞게 전환하는 지식이다. 

지구에 저장된 에너지의 총량은 태양에너지에 비하면 무시할 만한 정도이다. 지구에 도달하는 태양 에너지는 극히 일부분임에도 불구하고 376만 6,800 엑사줄(1줄의 10억의 10억배)이나 된다. 하지만 세상의 모든 식물이 광합성을 통해 받아들이는 태양에너지는 약 3천 엑사줄 밖에 되지 않는다. 반면 인간이 매년 소비하는 양은 5백 엑사줄 가량으로 지구가 태양으로 부터 90분간 받는 양에 불과하다. 이밖에도 핵에너지, 중력에너지(밀물, 썰물 등)도 있다.

**에너지를 보다 효율적으로 변환하는 방법을 알아내면서 원자재 문제도 해결되었다.** 1820년도 화학자들은 광석에서 알루미늄을 발견했지만 이것을 추출하기는 무척이나 힘들고 많은 비용이 소모되었다. 수십년간 알루미늄은 금보다 비쌌다. 1860년대 프랑스의 나폴레옹 황제는 가장 신분이 높은 사람한테 알루미늄 식기를 놓았고, 그보다 신분이 낮은 사람한테는 금으로 된 포크와 나이프가 놓였다. 
19세기에 들어 알루미늄을 대량으로 값싸게 추출하는 방법을 알아냈고 오늘날에는 연간 3천만톤을 생산하고 있다. 만일 나폴레옹 대왕이 오늘날 후손들이 남은 음식을 알루미늄 호일에 싸가지고 다닌다는 것을 알면 무척 놀랄 것이다.

### 17_3. 컨베이어 벨트 위에서의 삶

산업혁명의 성장은 농업에서 가장 먼저 느낄 수 있었다. 지난 2백 년간 산업적 생산의 주류는 농업이었다. 트랙터는 사람의 일을 대신하였고, 인공비료, 살충제, 성장 호르몬 등의 무기로 생산량이 폭발적으로 증가하였다. 그리고 냉장고, 선박, 항공기의 도움으로 유럽인들은 신선한 아르헨티나 쇠고기와 일본 스시로 식사를 할 수 있었다.

**심지어 동물들도 기계화되었다.** 오늘날 공장 비슷한 시설에서 대량으로 생산되며, 산업의 요구에 따라 다양한게 변화했다. 가축의 삶의 질은 해당 기업의 이익과 손해에 따라 결정되었다. 돼지는 포유동물 중 가장 지능과 탐구심이 뛰어난 축에 속한다. 이를 능가하는 동물은 유인원뿐이다. 하지만 산업화 속의 돼지는 너무 좁은 우리에서 몸을 돌릴 수 도 없으며, 암퇘지는 출산 후 4주 동안 밤낮으로 우리에 갇혀 있여 계속해서 다음번 새끼를 임신한다. 많은 젖소는 좁은 울타리에서 생애를 보내며 자기가 싼 대소변 위에서 잠을 잔다. 기계 장치는 가축들에게 호르몬, 약품을 공급하며, 동시에 우유를 짜낸다. 이 사이에 있는 젖소는 원자재를 받아들이는 입과 상품을 생산하는 젖통 이상의 취급을 받지 못한다. 

**대서양 노예무역이 아프리카인을 향한 증오의 결과가 아니었던 것 처럼, 현대의 동물 산업도 악의에서 출발한 것이 아니다. 이번에도 그 원천은 바로 무관심이었다.** 진화심리학에 따르면 가축의 정서적, 사회적 욕구는 그것이 생존과 번식에 핵심적이었던 시절에 야생에서 진화한 것이다. 송아지는 어린 아이와 마찬가지로 놀이를 하려는 강한 본능이 심어져 있고 어미와 강한 유대를 형성하려는 강한 욕구도 있다. 어미의 젖과 보살힘은 생존에 결정적으로 중요하기 때문이다.

오늘날 농부는 이런 송아지를 어미한테서 때어내고 높은 우리에 가두어 키운다. 객관적으로 보면 송아지는 생존과 번식을 걱정할 필요가 없지만 여전히 강력한 욕구를 느낀다. 이런 욕구가 충족되지 않으면 송아지는 큰 고통을 느끼게 된다. **산업화된 농업의 비극은 동물의 주관적 욕구를 무시하고 객관적인 욕구만 챙긴다는 점이다.**

1950년도 미국 심리학자 할로는 원숭이를 통해 실험을 진행하였다. 어미와 때어낸 새끼 원숭이를 우유병이 있는 철사로 된 인형, 나무와 천으로 된 인형 2가지를 보여주었다. 결과는 예상과 달리 새끼 원숭이는 아무런 자양분을 얻을 수 없음에도 불구하고 나무와 천으로 된 모형에 매달렸다. 

이런 원숭이들은 성장후에도 정서장애가 생겼다는 사실이 밝혀졌다. 사회에 적응하지 못했고, 다른 원숭이와 소통에 어려움을 겪었고, 높은 수준의 분안과 공격성에 휩싸였다. 이 실험을 통해 **원숭이는 물질적 필요를 넘어서는 심리적 필요와 욕구를 지니고 있으며, 이 욕구가 충족되지 않으면 큰 고통을 겪는다**라는 것을 알 수 있었다. 이후 이 실험은 원숭이 외 다양한 동물들을 통해서도 진행되었는데 마찬가지 결과가 나타났다. 오늘날 수백만 마리의 가축이 할로의 원숭이와 동일한 환경에 처해있다.

산업화 이전에는 90%이상의 농부가 소수의 지배계층 위해 생산했다. 농업이 산업화 된 오늘날에는 적은 수의 농부로도 많은 사무원과 공장 노동자를 먹여 살리기에 충분한 양을 생산한다. **오늘날 미국에서는 농업으로 먹고 사는 인구는 2%에 불과하다. 이 2%의 사람들이 미국 인구으 ㅣ전체를 먹이고 남은 것을 수출할 만큼 생산하고 있는 것이다.** 오늘날 인류는 과거의 어느 때보다 많은 양의 철강과 의류를 만들고 구조물을 세운다. 과거에는 상상도 할 수 없었던 제품을 만든다. 전구, 휴대전화, 카메라, 식기 세척기, ... 인류 역사상 처음으로 공급이 수요를 넘어서기 시작했다.

### 17_4. 쇼핑의 시대

상어가 계속 헤엄을 치지 않으면 질식하여 죽는 것 처럼 현대의 자본주의 경제는 끊임없이 생산량을 늘려가고 있다. **하지만 만들기만 하고 누군가 사주지 않으면 의미가 없다. 이런 파국을 막기 위해 새로운 윤리가 등장했는데 바로 소비지상주의이다.** 근검과 절약을 강조했던 근대와 달리 소비지상주의는 점점 더 많은 재화와 용역을 소비하는 것을 긍정적으로 본다. 

대중 심리학(Just do it!)의 도움을 받아 탐닉은 좋은 것이며 검약은 스스로를 억합하는 것이라고 설득하려 한다. 하지만 그 설득은 성공했고 이제 우리는 무엇인가를 사들인다. 크리스마스 종교 휴일은 쇼핑의 축제가 되었고, 언제와 다른 최신품이 유행하고, 가전제품은 몇 년에 고장이나 주기적으로 새로 구매해야 한다. 

**소비지상주의는 식품 시장에서 가장 잘 알 수 있다.** 풍요로운 오늘날 사회에서 가장 심각한 문제는 바로 비만인데, 가난한 층일 수록 비만율이 높다. 미국인이 해마다 다이어트에 투자하는 돈은 나머지 나라의 배고픈 사람들을 먹이고도 남을 정도이다. 즉 사람들은 먹기 위해서도 돈을 쓰고, 다이어트를 위해서도 돈을 쓰는 시대이다.

과거에는 귀족들은 호화로운 생활을 위해 과소비를 하였지만 가난한 농부는 검소하게 살았다. 하지만 오늘날에는 상황이 바뀌었다. 부자들은 자산과 투자물을 조심스럽게 관리하는데 반해, 그 보다 가난한 사람들은 빛을 내어서라도 꼭 필요하지 않은 TV나 자동차를 구매한다. **자본주의와 소비지사주의는 동전의 양면이다.** 이 동전에는 두 가지의 계율이 새겨져 있다. 부자의 계율은 '투자하라!'이고 나머지 사람들의 계율은 '구매하라!'이다.
