####  4부. 과학혁명

# 20. 호모 사피엔스의 종말

사피엔스는 다른 어떤 생명체도 누리지 못한 거대한 운동장을 가지게 되었지만, 생물학적인 한계를 벗어날 수 없다. 하지만 21세기에는 이는 더 이상 사실이 아니다. 호모 사피엔스는 자연선택의 법칙을 깨면서 지적설계의 법칙으로 대체하고 있는 중이다.

약 1만 년 전 농업혁명으로, 언제든 편하게 고기를 얻고 싶어하는 사피엔스가 가장 통통한 동물끼리 **교배시키면서 부터 신이 아닌 인간이 지적으로 설계로 새로운 혈통을 만들었다.** 이것을 시작으로 40억 년에 걸쳐 이어져온 자연선택 체계는 현제 유전자 조작 기술로 큰 위협을 받고 있다. 2000년에는 생물예술가 에두아르도 카츠가 녹색 현광 토끼 '알바'를 창조했다. 알바의 존재는 자연선택의 법칙으로 설명이 불가능하다. 

현재에도 자연선택을 지적설계로 대체하는 일이 진행중이다. 그 방법은 크게 3가지로 **첫째가 생명공학, 둘째가 사이보그 공학, 셋째가 비유기물공학이다.**


----

### 20_1. 생쥐와 인간

#### 유전공학

생명공학은 유전체 이식과 같이 생물학적인 부분이 인간이 개입하는 것을 의미한다. 사실 생명공학은 새로운 것이 아니고 과거 농업혁명때 교배, 거세 등에서부터 시작되었다. **1996년 등에 귀가 달린 생쥐의 사진이 등장했다. 합성이 아닌 유전자 조작으로 만든 것이었다. 유전공학이 윤리적, 정치적, 이데올로기적 쟁점이 되는 이유가 여기에 있다.** 

동물권리 운동가들은 이런 유전공학의 실험 동물이 겪는 고통에 대해 강하게 비판한다. 인권 운동가는 이런 생물학자가 슈퍼맨을 만들어내는 것이 아닌지 걱정한다. 너무나도 많은 가능성의 문이 열려있고, 유전자 조작 기술은 무섭게 발전하고 있다.

그래서 유전공학의 잠재력 중 아주 일부분만 사용하고 있는데 식물, 미생물, 곤충 등이다. 인간의 장에 있는 대장균을 이용해 바이오 연료를 사용하거나, 곰팡이들을 조작해 인슐린을 생성하여 당뇨병 치료에 기여하거나, 북극에 사는 물고기의 유전자와 감자를 결합해 서리 저항력을 높이는 식으로 사용하고 있다. 

지금의 기술로도 벌래의 수명을 크게 늘리고, 기억력이 좋은 천재 생쥐를 만들었다. 차세대의 유전공학은 동물을 복제하는 수준을 우숩게 해낼수 있을 것이다. 일부일처를 유발하는 유전자를 바람둥이한테 결합하면 충실하고 애정 깊은 배우자로 변할지도 모른다.

### 20_2. 네안데르탈인의 귀환

#### 멸종 동물의 복원

유전공학의 대표적인 영화로 '쥐라기 공원'이 있는데 영화처럼 멸종위기의 동물들한테는 이런 유전공학이 꼭 필요할 것이다. 최근에는 여러 나라의 과학자가 모여 메머드의 유전자 지도를 작성한 일이 있는데, 머지않아 메머드를 복원하는 것을 계획하고 있다. 

하버드의 조지 처치 교슈는 네안데르탈인의 유전체 프로젝트를 완성하여, 고작 3천만 달러로 직접 네안데르탈인을 만들 수 있다고 주장하기도 하였다. 네안데르탈인과 호모 사피엔스의 뇌를 비교하면 인류학과 생물학적 부분에 큰 도움이 될 것이다.

#### 점점 무너지고 있는 윤리

앞으로 머지 않아 유전공학 더분에 인간의 생리기능, 면역계, 수명, 지적 및 정서적인 능력까지 크게 변화시킬 수 있을 것이다. 천재 생쥐를 만들 수 있다면 천재 인간도 충분히 만들 수 있을 것이다. **하지만 기술적인 장애는 몰라도 윤리적, 정치적 반대가 있다. 인간에 대한 연구가 느려진 것도 이 때문이다. 하지만 이 장애가 그리 오래가지는 않을 것이다.**

만약 알츠하이머 치료제를 개발했는데, 마침 그 약이 기억력을 증진시키는 부수효과가 있다면 누가 그 약의 개발을 중단시킬 수 있을까? 또한 이 약을 제한하는 법이 통과될 수 있을까? **생명공학이 네안데르탈인을 정말 부활 시키지 않을지 몰라도 호모 사피엔스의 막을 내리게 할 가능성은 크다. 유전자 공학으로 인류가 반드시 멸종하진 않아도 더 이상 호모 사피엔스가 아니게 될 것이다.**
