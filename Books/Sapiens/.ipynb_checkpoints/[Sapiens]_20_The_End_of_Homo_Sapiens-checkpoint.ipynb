####  4부. 과학혁명

# 20. 호모 사피엔스의 종말

사피엔스는 다른 어떤 생명체도 누리지 못한 거대한 운동장을 가지게 되었지만, 생물학적인 한계를 벗어날 수 없다. 하지만 21세기에는 이는 더 이상 사실이 아니다. 호모 사피엔스는 자연선택의 법칙을 깨면서 지적설계의 법칙으로 대체하고 있는 중이다.

약 1만 년 전 농업혁명으로, 언제든 편하게 고기를 얻고 싶어하는 사피엔스가 가장 통통한 동물끼리 **교배시키면서 부터 신이 아닌 인간이 지적으로 설계로 새로운 혈통을 만들었다.** 이것을 시작으로 40억 년에 걸쳐 이어져온 자연선택 체계는 현제 유전자 조작 기술로 큰 위협을 받고 있다. 2000년에는 생물예술가 에두아르도 카츠가 녹색 현광 토끼 '알바'를 창조했다. 알바의 존재는 자연선택의 법칙으로 설명이 불가능하다. 

현재에도 자연선택을 지적설계로 대체하는 일이 진행중이다. 그 방법은 크게 3가지로 **첫째가 생명공학, 둘째가 사이보그 공학, 셋째가 비유기물공학이다.**


----

### 20_1. 생쥐와 인간

#### 유전공학

생명공학은 유전체 이식과 같이 생물학적인 부분이 인간이 개입하는 것을 의미한다. 사실 생명공학은 새로운 것이 아니고 과거 농업혁명때 교배, 거세 등에서부터 시작되었다. **1996년 등에 귀가 달린 생쥐의 사진이 등장했다. 합성이 아닌 유전자 조작으로 만든 것이었다. 유전공학이 윤리적, 정치적, 이데올로기적 쟁점이 되는 이유가 여기에 있다.** 

동물권리 운동가들은 이런 유전공학의 실험 동물이 겪는 고통에 대해 강하게 비판한다. 인권 운동가는 이런 생물학자가 슈퍼맨을 만들어내는 것이 아닌지 걱정한다. 너무나도 많은 가능성의 문이 열려있고, 유전자 조작 기술은 무섭게 발전하고 있다.

그래서 유전공학의 잠재력 중 아주 일부분만 사용하고 있는데 식물, 미생물, 곤충 등이다. 인간의 장에 있는 대장균을 이용해 바이오 연료를 사용하거나, 곰팡이들을 조작해 인슐린을 생성하여 당뇨병 치료에 기여하거나, 북극에 사는 물고기의 유전자와 감자를 결합해 서리 저항력을 높이는 식으로 사용하고 있다. 

지금의 기술로도 벌래의 수명을 크게 늘리고, 기억력이 좋은 천재 생쥐를 만들었다. 차세대의 유전공학은 동물을 복제하는 수준을 우숩게 해낼수 있을 것이다. 일부일처를 유발하는 유전자를 바람둥이한테 결합하면 충실하고 애정 깊은 배우자로 변할지도 모른다.

----

### 20_2. 네안데르탈인의 귀환

#### 멸종 동물의 복원

유전공학의 대표적인 영화로 '쥐라기 공원'이 있는데 영화처럼 멸종위기의 동물들한테는 이런 유전공학이 꼭 필요할 것이다. 최근에는 여러 나라의 과학자가 모여 메머드의 유전자 지도를 작성한 일이 있는데, 머지않아 메머드를 복원하는 것을 계획하고 있다. 

하버드의 조지 처치 교슈는 네안데르탈인의 유전체 프로젝트를 완성하여, 고작 3천만 달러로 직접 네안데르탈인을 만들 수 있다고 주장하기도 하였다. 네안데르탈인과 호모 사피엔스의 뇌를 비교하면 인류학과 생물학적 부분에 큰 도움이 될 것이다.

#### 점점 무너지고 있는 윤리

앞으로 머지 않아 유전공학 더분에 인간의 생리기능, 면역계, 수명, 지적 및 정서적인 능력까지 크게 변화시킬 수 있을 것이다. 천재 생쥐를 만들 수 있다면 천재 인간도 충분히 만들 수 있을 것이다. **하지만 기술적인 장애는 몰라도 윤리적, 정치적 반대가 있다. 인간에 대한 연구가 느려진 것도 이 때문이다. 하지만 이 장애가 그리 오래가지는 않을 것이다.**

만약 알츠하이머 치료제를 개발했는데, 마침 그 약이 기억력을 증진시키는 부수효과가 있다면 누가 그 약의 개발을 중단시킬 수 있을까? 또한 이 약을 제한하는 법이 통과될 수 있을까? **생명공학이 네안데르탈인을 정말 부활 시키지 않을지 몰라도 호모 사피엔스의 막을 내리게 할 가능성은 크다. 유전자 공학으로 인류가 반드시 멸종하진 않아도 더 이상 호모 사피엔스가 아니게 될 것이다.**

### 20_3. 생체공학적 생명체

#### 사이보그 공학

사이보그은 생물과 무생물을 부분적으로 합친 존재를 의미하는데 어쩌면 우리 모두는 생체공학적인 존재인지도 모른다. 안경, 심장박동기, 의료보장구, 컴퓨터, 휴대전화 등으로 우리의 감각과 기능을 보완하고 있기 때문이다. 

미국의 군사 연구기관이 국방고등연구기획청은 곤충 사이보그를 개발하고 있다. 파리나 바퀴벌레에 전자칩이나 연산장치를 심어서 멀리서 정보를 수집한다고 한다. 2006년 미 해군잠수전센터는 사이보그 상어를 개발하겠다는 의도를 밝힌 바있다. 상어의 뛰어는 자기장 탐지 능력으로 잠수함이나 기뢰들을 식별하는 목적이다. 

#### 사이보그로 변하는 사피엔스

망막 임플란트 회사는 시각장애인의 눈에 작은 마이크로칩을 삽입하여 빛을 전기 에너지로 전환하여 시신경세포를 자극하는 장치를 개발중이다. 미국의 전기기술자 제시 설리반 2001년 사고로 두 팔을 모두 잃었지만 시카고 재활연구소 Rehavilitation Institute of Chicago의 도움으로 생각만으로 움직이는 기계팔을 가지게 되었다. 

현재로써는 이런 기계 팔이 실제 팔에 미치지 못하지만 더 강한 힘을 가지고 있다. 미국 노스캐롤라이나 주 듀크 대학교에서는 붉은털원숭이인 오로라와 아이도야의 실험으로 원격으로 원거리의 기계 팔과 다리를 움직일수 있다는 것을 입증했다. 현재 뇌와 컴퓨터를 직접 연결하는 연구가 활발히 진행중인데 그런 사회가 되면 어떻게 될까? 우리의 뇌를 연결해 컴퓨터를 만들고, 기억은행이 생긴다면 기억과 의식은 어떻게 될까? 자아난 성정체성 같은 개념은 어떻게 변할까?

### 20_4. 또 다른 삶

#### 또 다른 생명의 법칙

 유전적 프로그래밍은 유전자의 진화를 모방한 것으로 혼자서 학습, 진화할 능력을 가진 프로그램이 목표이다. 하지만 그 피조물은 창조자의 초기 방향과 다른 방향으로 진화할 지도 모른다. 이런 프로그램의 대표적인 것이 바로 컴퓨터 바이러스이다. 백신에 쫒기는 한편 다른 바이러스와 경쟁하기 위해 수없이 스스로를 복제하는데, 그 과정중에 돌연변이가 생길 수 있다. 만일 이 돌연변이 바이러스가 백신에 내성이 있고 다른 바이러스보다 더 감염능력이 좋다면 더 잘 살아남고 번식할 것이다. 

#### 블루브레인 프로젝트

당신의 뇌를 휴대용 하드드라이브에 백업해서 노트북에서 실행한다고 가정해보자. 이 경우 그 노트북은 사피엔스 처럼 느끼고 생각할 수 있을까? 당신이 노트북이라고 할 수 있을까? 노트북 사용자가 완전히 새로운 마음을 창조할 수 있다면 어떻게 될까? 온갖 감정을 코드로 구성할 수 있을까? 그렇다면 그 코드들을 전부 지우면 살인을 한 것일까? 

머지 않안 이런 질문에 대한 답이 나올지도 모른다. 2005년 시작된 블루브래인 프로젝트 Blue Brain Project는 인간의 뇌를 전부 컴퓨로 재창조하는 것을 목표로 하고 있다. 만일 성공한다면 40억 년의 세월 끝에 비유기물의 세계가 펼쳐진다는 것을 의미한다. 인간의 마음이 디지털 컴퓨터와 비슷한 방식으로 작동한다고 100% 동의할 수 없지만 그 가능성을 무시하는 것은 어리석은 일이 되었다. 2013년 유럽연합은 이 프로젝트에 10억 유로의 보조금을 지급하기로 결정하였다.

### 20_5. 특이점

#### 생물학적 족쇄로부터 해방

이런 가능성 중에서 실현된 것은 일부분에 지나지 않지만 그 속도는 위험할 정도로 빠르다. 법률가는 프러이버시와 정체성 이슈를, 정부는 의료 및 평등에 대한 문제를 다시 생각해볼 필요가 있다. **생명공학, 사이보그, 비유기적 생명체 등에 대한 난제를 대면하게 될 것이다.** 인간 유전자 지도를 처음 만들 때는 30억 달러의 15년의 시간이 필요했지만 오늘날에는 몇백달러와 몇 시간이면 충분해졌다.


#### DNA 프라이버시
현제는 맞춤 의학의 시대가 되었다. 하지만 이런 진화에 맞추어 윤리적 난제가 대두될 것이다. 윤리학자와 법률 전문가들은 벌써부터 DNA 프라이버스라는 난제로 서로 씨름하고 있다. 보험회사는 고객의 DNA를 분석하여 문제가 될 유전적 성향이 있다면 보험료를 인상할까? 취업을 원하는 회사에 이력서가 아닌 DNA 자료를 전송할까? 유전적 차별을 이유로 소송 할 수 있을까? 

이 같은 딜레마는 21세기 길가메시 프로젝트의 윤리적, 사회적, 정치적 함의에 비하면 사소한 것들이다. 세계인권선원, 전 세계의 의료 지원 프로그램, 국민 건강보험 프로그램, 세계 각국의 헌법은 인간사회의 그 구성원 모두에세 공평한 의학적 피료를 제공하고, 상당 수준의 건강을 유지하게 할 의무가 있다는 것에 동의한다. 하지만 의학이 몰두하는 주제가 인간강화라면 어떻게 될까? 모든 인간은 강화된 능력을 부여받을 자격이 있을까? 초인간 엘리트 족속이 새로 생겨날까?

#### 평화로운 차별

오늘날은 역사상 유래가 없을 정도로 평화로운 시대지만, 동시에 그만큼의 불평등을 창조할 만한 만반의 태세를 갖추고 있다. 오래전 상류계급은 자신들이 언제나 하류계급보다 우수하다고 자신들을 속여왔다. 가난한 농부의 아이나 황태자 다를바 없지만 현대에서는 의학적 도움을 받는다면 그 차이는 객관화 될 수 있다. 

미래의 기술은 호모 사피엔스를 변화시킬 것이다. 단순히 신체적 변화가 아닌 감정과 욕망까지도 말이다. 미래의 주인공이 될 수 있는 사이보그와의 차이는 우리와 네안데르탈인의 차이보다 더욱 클 것이다. 적어도 네안데르탈인은 같은 인간이었지만 미래의 사이보그는 신에 가까운 존재가 될 지도 모른다. 



#### 프랑캔슈타인의 예언

1818년 메리 셜리의 소설 [프랑켄슈타인]은 우월한 존재를 창조하려는 과학자가 괴물을 만들어 버리는 이야기인데, 여기에는 우리가 종말의 날에 생각해볼 이야기가 있다. 머지않아 우리들은 완전히 다른 존재로 대체되는 시대가 올 것이다. **이런 생각은 사피엔스를 극단적으로 불안하게 만든다.** 미래에는 우리와 같은 감정과 정체성을 가진 존재가 더 이상 존재하지 않고, 더 능력이 뛰어난 생명체가 차지할 것이라는 불안감이다. 

그래서 우리는 **프랑켄슈타인 박사가 창조할 수 있는 것은 오직 괴물뿐이며, 그 괴물을 파괴해야 세상을 구할 수 있다는 스토리에 위안을 얻을 수 있는 것이다.** 하지만 이는 우리가 모든 존재 중에 가장 우수하며, 우리보다 우수한 존재는 과거에도 미래에도 있을 수 없다는 것을 의미하기도 한다. 과학자들이 신체뿐만이 아닌 정신까지도 조작할 수 있다는 사실을 받아들이려면 힘든 시간을 거쳐야 할 것이다. 미래의 프랑캔슈타인 박사는 정말로 더 우월한 존재를 만들어 낼 지도 모른다.

#### 알 수 없는 미래

역사를 살펴보면 한 치 앞도 예상할 수 없을 정도로 변화무쌍하게 이루어졌다는 것을 알 수 있다. 1940년대 원자력이 도래했을 때 2000년도 쯤에는 원자력을 활용한 새로운 세계가 펼쳐질 것이라고 예상하였고, 아폴로 11호가 우주에 대한 상상력에 불을 지폈을 때 20세기에는 화성에서 살게 될 것이라고 예상하였다. 

앞의 환상들은 상상력을 일께우기 위한 자극제일 뿐이다. 우리가 진지하게 받아들여야 할 부분은 **다음 역사에서는 인간의 의식과 정체성에도 근본적인 변화가 일어날 것이라는 점**이다. 어쩌면 '인간적'이라는 용어에 대해서도 의문을 품게 될지도 모른다. 만일 사피엔스가 정말로 막을 내리고 있다면, **우리는 "우리는 무엇이 되고 싶은가?"라는 마지막 남은 질문에 답을 내리는데 시간을 보내야 한다.** 아마 온갖 종류의 논쟁은 호모 사피엔스의 종말과 함께 사라져 버릴 것이다.

대부분의 사람들은 이런 생각을 하지않는 편을 선호한다. 심지어 생명윤리에서조차 '무엇이 금지된 행위인가?'하는 질문에 관심을 쏟는다. 각종 불멸을 향한 탐구는 이제 떌려야 땔 수 없을 만큼 엮여 있다. **과학이 하는 모든 일은 사람의 목숨을 살리고 모두의 행복을 위하여 진행되며 막을 도리가 없다. 단지, 우리가 할 수 있는 것은 이 방향에 영향을 미치는 것 뿐이다.** 앞으로 우리한테 남겨진 가장 중요한 질문이 빠뀔지도 모른다. **"우리는 무엇을 원하고 싶은가?"**
