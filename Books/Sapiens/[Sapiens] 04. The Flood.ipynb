#### 1부. 인지혁명

# 4. 대홍수

인지 혁명 이전 인간은 **아프로 아시아 육괴(아프리카와 아시아가 합쳐진 고대륙)에서 살았다.** 하지만 이들은 큰 바다로 나가지 않았으며 아무도 아메리카, 호주 혹은 더 먼 곳인 마다가스카르, 뉴질랜드, 하와이에는 가지 못했다. 약 4만 5천 년 전 호주에 최초로 정착한 기록이 있는데 그 사이에 있는 수많은 해협과 도착했을 때 완전히 새로운 환경에 어떻게 적응한 한 것인지 아직 불분명하다. **가장 합리적인 설은 인도네시아 제도에서 살던 어떤 사피엔스가 항해술을 익혀 장거리 탐험가가 되었다는 것이다. **



최초의 인류가 호주까지 여행을 한 것은 콜럼버스의 아메리카 발견, 아폴로 11호 탐험대가 달에 착륙한 것 못지않게 중요한 사건이다. 하지만 이보다 더 중요한 것은 바로 이들이 호주에 발을 디딘 이후에 일어난 사건이다. 이로부터 몇천 년 지나지 않아 대형동물은 사실상 모두 사라졌다. **몸무게 50kg이 넘는 호주의 동물 24종 23종이 멸종했다. 이보자 작은 종도 대량으로 사라졌다.** 이 호주 정복자들은 호주의 먹이사슬 최상층부로 올라가 행성 연대기에서 가장 치명적인 종이 된 순간이다. 

### 4-1. 기소 내용대로 유죄

일부 학자들은 동물들의 멸종이 기후변화라고 한다. 하지만 호모 사피엔스가 결백하다고 하기에는 어렵다.

1. 4만 5천 년 전쯤 호주의 기후가 변한 것은 사실이지만 동물이 멸종할 만큼 급격한 변화는 아니었다. 

2. 기후변화가 대량 멸종을 초래할 경우 해양생물 또한 육지생물 못지않은 타격을 받는다. 하지만 그 당시 해양 동물이 크게 줄었다는 증거가 없다. 

3. 호주에서 일어난 대량 멸종 사건이 인간이 다른 외부세계에 정착할 때마다 일어났다.



당시가 석기 시대임을 고려하면 돌도끼를 위두느느 정도의 기술로 대량 멸종시켰다는 내용 또한 믿기 쉽지 않다. 하지만 이에 대해 3가지 설명이 있다.

1. 호주의 대형동물은 번식 속도가 느리다. 호주의 동물은 인간의 사냥으로부터 도망쳐야 한다는 것을 배울 시간이 없었다.

2. 당시 사피엔스는 이미 화전법에 능했다. 당시 유칼립투스 나무는 화재에 강해 살아남을 수 있었지만 다른 식물과 그 식물을 먹는 동물들 또한 연쇄적으로 멸종의 길을 걸었다.

3. 당시의 기후변화가 인간의 사냥과 합쳐져서 더욱 다양한 방법으로 공격했다. 

### 4-2. 나무늘보의 종말

기원전 1만 4천 년 전쯤 일부의 사피엔스가 사냥감을 쫒아 시베리아 북동부에서 알래스카까지 이동하였다. 기원전 1만 2천 년 전쯤 지구 온난화로 얼음이 녹고 이번에는 인류는 때를 지어 대륙 전체로 퍼져나갔다. 시베리아인의 후예들은 미국 동부의 울창한 숲, 미시시피 삼각주의 늪지대, 멕시코의 사막, 중미의 밀림에 정착했다. 이 모든 일이 1천 년 전이나 2천 년 전에 일어났다.  기원전 1만 년이 되자 인류는 미대륙 최남단 티에라 델 푸에고 제도까지 정착했다.



**사피엔스가 지나간 자리에는 희생자들의 흔적이 길게 남았다.** 사피엔스가 도착한 지 2천 년이 지나지 않아 이들 유일무이한 종은 대부분 사라졌다. 

### 4-3. 노아의 방주

가장 심한 타격을 받은 것은 털북숭이 대형동물들이었다. 인지 혁명이 일어날 즈음 45kg 이상의 동물들이 2백여 속 정도가 살고 있었지만 농업 혁명이 일어날 쯤에는 이들 중 약 1백 속만이 살아남았다. **호모 사피엔스는 문자나 바퀴, 금속 도구를 발명하기 한참 전부터 지구 대형동물 절반가량을 멸종시킨 것이다. **



수렵채집인의 확산과 함께 벌어졌던 멸종의 제 1 물결 다음에는 농부들의 확산과 함께 벌어졌던 멸종의 제 2 물결이 왔고, 이 사실은 오늘날 산업활동이 일으키고 있는 멸종의 제3의 물결에 대한 중요한 관점을 제공한다. 우리 조상이 자연과 더불어 행복하게 살았다는 급진적인 환경운동가의 말을 밎지 마라. 산업 혁명 훨씬 이전부터 호모 사피엔스는 모든 생물들을 아울러 가장 많은 동물과 식물을 멸종으로 몰아놓은 기록을 보유하고 있다. **우리는 생물학 연대기에서 단연코 가장 치명적인 종이라는 불명예를 갖고 있다.**
