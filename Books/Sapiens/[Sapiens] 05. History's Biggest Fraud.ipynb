#### 2부. 농업혁명

# 5. 역사상 최대의 사기

호모 사피엔스는 동아프리카에서 중동으로, 유럽과 아시아로, 마지막으로 호주와 미 대륙으로 퍼져나갔다. 하지만 이들은 가는 곳마다 야생식물을 채취하고 야생동물을 사냥하면서 사는 방식을 유지했다. **하지만 이 모든 것은 1만 년 전쯤부터 달라졌다.** 



인류가 농업으로 이행한 것은 기원전 **9천5백 년 ~ 8천5백 년 경 터키 남동부, 서부 이란, 에게 해 동부 지방에서였다.** 밀을 재배하고 염소를 가축화한 것은 기원전 9000년경, 완두콩과 렌즈콩은 기원전 8000년경, 올리브 나무는 기원전 5000년경, 포도는 기원전 3500년 재배되었고, 밀은 기원전 4000년 전부터 기르기 시작했다. **오늘날 인류를 먹여 살리는 주요 곡물인 밀, 쌀, 옥수수, 감자, 보리는 기원전 9500년~3500년 전에 작물화되었다.**



우리 조상들이 잡거나 채취했던 수처 종의 동물과 식물 중에 농업과 목축에 맞는 후보는 몇 개 되지 않았으며, 이들이 모여 살던 특정 장소에서 농업 혁명은 싹을 틔었다. 하지만 농업혁명 덕분에 생산량이 늘어난 것은 맞지만 여분의 식량이 꼭 더 나은 식사나 여유를 의미한 것은 아니다. **농부는 수렵채집인 보다 더 열심히 오랫동안 일 했으며 더 열 약한 식사를 했다. **



식물이 호모 사피엔스를 길들였지 호모 사피엔스가 이들을 길들인 것이 아니다. 사피엔스의 신체는 이런 과업에 맞게 진화하지 않았다. 사과나무에 기어오르고 가젤을 뛰어서 뒤쫓는 데 적응했지, **바위를 제거하고 물이 든 양동이를 운반하는데 적합한 몸이 아니었다. 인간의 척추와 무릎, 목과 발바닥의 장심이 대가를 치렀다.**



곡류를 중심으로 하는 식단은 미네랄과 비타민이 부족하다. 경제적 안정을 제공하지도 않았다. 비가 내리지 않거나 메뚜기 떼가 덮치거나, 곰팡이가 감염시키면 농부는 수천, 수백만 명이 죽었다. 인간 사이이의 폭력에 대한 안전망을 제공하지도 않았다. 위협을 당해 목초지에서 쫓겨나면 굶어 죽었다. 



요즘 우리는 풍요와 안전을 누리고 있고 그러한 것들은 농업혁명 기초 위에 놓인 것이기 때문에 농업혁명은 놀라운 개선이라고 가정한다. 하지만 수천 년 전의 역사를 오늘날의 관점에서 판단하는 것은 잘못이다. 제정신인 사람이라면 호모 사피엔스의 DNA 복사본을 늘리기 위해서 삶의 질을 포기할 사람이 있겠는가? **그런 거래에 동의한 사람은 아무도 없었다. 농업혓명의 덫이다.**

### 5-1. 사치라는 덫

영구 정착촌에 살면서 **식량 공급이 증가하자 인구가 늘기 시작했다.** 방랑하는 삶을 포기하자 여성은 매년 아기를 가질 수 있게 되었다. 하지만 먹을 입이 늘면서 식량 소진은 빠르게 증가했고 영양을 제대로 공급받지 못한 아이들은 죽어 나갔으며, 동시에 사람들은 **더 많은 아이를 낳기 시작했다. 역설적이게도 일련의 개선이 합쳐져서 농부들의 어깨에 더 무거운 짐으로 얹혔다.** 각각의 삶을 좀 더 좋게 바꾸려고 노력했는데도 말이다.



좀 더 쉬운 삶을 추구한 결과 더 어렵게 되어버린 셈이었고, 이것이 마지막도 아니었다. 오늘날 우리에게도 똑같은 일이 벌어지고 있다. 대학을 졸업한 젊은이 상당수는 돈을 벌어 35세에서 은퇴해서 진짜 자신이 원하는 것을 하겠다고 다짐하며 유수 회사들에 들어가 힘들게 일한다. 하지만 막상 그 나이가 되면 거액의 주택융자, 학교에 다니는 자녀, 차가 있어야 다닐 수 있는 교외의 집, 정말 좋은 와인과 휴가가 없다면 삶은 살만한 가치가 없다고 느끼게 된다. 이들은 뭘 어떻게 할까? 뿌리채소나 캐는 이전의 삶을 되돌아 깔까? **이들은 노력을 배가해서 노예 같은 노동을 계속한다. **



**농업혁명은 해당 지역의 모든 무리의 동참을 필요로 하지 않는다. 한 무리가 정착을 하면 농업은 저항할 수 없는 것이 된다.** 급속한 인구 성장 조건을 만들어 주었기 때문에 농부들은 순수한 머릿수의 힘만으로 언제나 수렵채집인을 압도할 수 있었다.  인류가 좀 더 편한 생활을 추구한 결과 막강한 변화의 힘이 생겼고 **이것이 아무도 예상하거나 희망하지 않았던 방향으로 세상을 변화시켰다. **

### 5-2. 신성한 개입

**1995년 고고학자들은 터키 남동부의 괴베클리 테페 지역 유적지를 발견했다.** 돌기둥은 최대 7톤, 높이는 5미터에 달했다. 이 유적지를 조사 결과 기원전 9500년 전의 것으로 나타났다. 즉 수렵채집인들이 세운 것이라는 의미이다. 이들은 막대한 노력과 시간을 투자하였고 이를 위해 수많은 부족이 연합했을 것이다. 이런 협력을 위해서는 종교나 이데올로기 시스템가 필수적이다. 그리고 이 유적지에서 30킬로 떨어진 카라사다 그 언덕은 이 유적지에서 발견된 작물화된 밀의 발상지이다.



이런 2가지 사실은 우연이 아닐 것이다. **인류에 의한 밀의 작물화, 밀에 의한 인간 길들이기가 서로 연관되어 있을 것이다. 먼저 사원이 세워지고 나중에 그 주위에 마을이 형성되었다는 추측할 수 있다.**


### 5-3. 혁명의 희생자들.

**불행히도 진화적 관점은 성공의 척도로서는 불완전하다.** 그것은 모든 것은 생존과 번식이라는 기준으로 판단할 뿐, 개체의 고통이나 행복은 아랑곳하지 않는다. 동물의 가축화는 일련의 야만적인 관행을 기반으로 이루어졌고, 관행은 수백수천 년이 흐르면서 더욱 잔인해졌다. 야생 상태에서 닭은 7~12년, 소는 20~25년 정도 사는데 반해 가축화된 닭이나 소는 몇 주 내지 몇 개월만에 도살당한다. 



황소나 말, 당나귀와 낙타를 순종적으로 바꾸려면 이들의 자연적인 본능과 사회적 유대관계를 파괴해야 한다. 

뉴기니에서는 돼지 코에서 큰 살덩이를 잘라내는데, 그러면 돼지는 냄새를 맡을 때마다 심한 통증을 느껴 인간에게 의존하게 된다. 낙농산업에서는 암소, 염소, 양한테서 많은 젖을 얻기 위해 새끼를 도살하고 가능한 많은 양의 젖을 짜내는데 이 방법은 현재까지도 사용되고 있다. 수단의 누에르족은 송아지의 입가에 가시를 씌어 새끼가 젖을 빨리 못하게 막고, 사라하 사막의 투아레그족은 낙타의 새끼에 코에 구멍을 내거나 일부를 잘라내어 젖을 많이 먹지 못하게 했다. 



화적 성공과 개체의 고통 간의 괴리는 농업혁명에서 얻을 수 있는 가장 중요한 교훈이다. **진화적 성공이 개체의 경험에 어떤 영향을 미칠지 고려할 필요가 있다. 우리는 우리 종의 성공이 개객인의 큰 고통과 나란히 진행되었다는 사실을 거듭 확인하게 될 것이다.**
