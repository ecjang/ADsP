#### 2부. 농업혁명

# 6. 피라미드 건설하기

농경 사회 덕분에 인구가 빠르게 증가했고 **너무 커져버린 덩치는 다시는 이전의 수렵채집 생활로 돌아가 스스로를 유지할 수 없었다.** 기원전 10000년경 5백만~8백만의 수렵채집인이 있었지만 **기원전 1세기가 되자 농부들은 2억 5천만 명으로 커져있었다.** 

지구 표면은 약 5억 1천만 ㎢인데 이중 1억 5천만 ㎢가 육지다. 다수의 농부들은 약 경작에 알맞은 지역인 1.1천만 ㎢, 즉 **지표면의 약 2%에 불과한 좁은 지역에 모여 살았다. 이 좁은 지역에서 역사가 펼쳐진 것이다.**

----

### 6_1. 미래의 도래

**농업으로 미래를 예측하는 것은 무엇보다 중요해졌다.** 수확을 마친 농부는 그날 마음껏 축하를 할 수 있지만 며칠 지나지 않아 다시 빠른 일상으롤 되돌아가야 했으며 식량은 오늘, 다음 주, 다음 달까지는 충분하지만 다음 해에 먹을거리를 걱정해야 했다.

그동안 유프라테스, 황화, 인더스 유역의 농부들은 강물의 수위를 점검했다. 이들에게 강물은 불어날 필요가 있었다. 그래야 고지대에서 씻겨 내려온 비옥한 흙이 퍼지고 광대한 관개 시스템에 물을 댈 수 있었다. 하지만 홍수가 너무 심하거나 시기가 잘못되면 가뭄만큼이나 들판을 모두 망칠 수 있었다.

**농사 스트레스는 막대한 영향을 미쳤다. 그것은 대규모 정치 사회의 토대였다.** 슬프게도 부지런히 일했음에도 불구하고 농부는 그토록 원했던 경제적 안정을 얻지 못했다. 모든 곳에서 지배자와 엘리트가 출현했다. 그들이 빼앗은 잉여 식량은 정치, 전쟁, 예술, 철학의 원동력이 되었고 왕궁과 성채, 기념물과 사원을 지었다. 근대 후기에 이르기까지 인류의 90%는 아침마다 일어나 구슬 같은 땀을 흘리며 땅을 가는 농부였다. 하지만 역사에 기록된 것은 이들 엘리트의 이야기다. **역사란 다른 모든 사람이 땅을 갈고 물을 운반하는 동안 극소수의 사람들이 해온 무엇이다.**

### 6_2. 상상 속의 질서

새로운 수송기술의 발전으로 더 많은 사람이 모여 마을을 이루어 읍이 되었고 도시가 되었다. 왕국과 상업망이 이들을 하나로 묶었지만 이런 새로운 기회를 활용하기 위해서는 **잉여 생산과 수송 수단으로는 충분하지 않았다.** 어떻게 토지와 물을 나눌지, 불화와 분쟁을 조절할지, 가뭄이나 전쟁에 어떻게 대비할지 합의가 이루어질 보장이 없었다. 합의가 이루어지지 못하면 분쟁이 번지기 마련이다. **도시와 왕국과 제국이 출현한 것은 농업혁명 이후 몇천 년밖에  걸리지 않았다. 협력하는 본능이 진화하기에는 너무나 짧은 기간이었다.**

강력한 제국이 형성될 가능성이 열리자 사람들은 위대한 신들, 조상의 땅, 주식회사 등등의 이야기를 지어냈다. 꼭 필요한 사회적 결속을 위해서였다. 인간의 상상력은 지구 상 유래 없는 거대한 협력의 네트워크를 만드어나갔다. 



기원전 8500년경 수백 명이 마을이 가장 큰 정착지였다. **기원전 7000년 소아시아의 차탈휘유크** 주민인 5천~1만 명으로 당시 세계 최대의 도시였다. **기원전 5천 ~ 4천 년 나일강의 초승달 지역**에 수만 명의 도시가 생겼다. **기원전 3천 년에는 하부 나일강 유역**까지 통합되면서 파라오가 다스리는 수십만 명의 최초의 이집트 왕국이 생겼다. **기원전 2250년경 사르곤 대제**는 1백만 명이 넘는 신민과 5천 명의 상비군을 가진 최초의 제국인 아카드를 건설했으며 **기원전 1천 ~ 5백 년 사이에는 아시리아, 바빌론, 페르시아 등 최초의 거대 제국이 등장**했다. 이들은 수백만명의 시민과 수만의 군대를 거느렸다. 협력이라는 말은 매우 이타적으로 들리겠지만 항상 자발적인 것은 아니었으며 평등주의 적인 경우는 드물었다. 대부분 압제와 착취에 적합하도록 맞춰져 있었다. 



기원전 1776년 바빌론은 세계 최대의 규모였다. **오늘날 가장 유명한 바빌론의 왕은 함무라비다.** 그의 이름을 딴 함무라비 법전은 정의로운 왕의 역할 모델을 제시하였다. 이 법전은 바빌론 사회적 질서는 보편적으로 영원한 정의의 원칙에 뿌리를 두고 있으며 이 원칙은 신이 읊어준 것이라고 단언한다. 

**계급제도의 중요성은 엄청나다.** 이 법에 따르면 인간은 두 개의 성별과 귀족, 평민, 노예 3개의 계급으로 나뉜다. 사람은 성별은 계급에 따라 각기 다른 가치를 지닌다. 평민 여성의 목숨 값은 30 세겔이고, 노예 여성은 20 세겔인데 반해 평민 남성의 눈은 60 세겔의 가치가 있다.

**가족 내의 엄격한 위계질서를 규정한다.** 어린이는 독립된 개인이 아니라 부모의 재산이다. 그러므로 어떤 귀족이 다른 다른 귀족의 딸을 죽이며 그 벌로 살해자의 딸이 사형당한다. 왕의 신민 모두가 위계질서상 자기 자리를 받아들이고 그에 맞게 행동하면 모두가 효과적으로 협력할 수 있다고 믿었다. 



함무라비가 사망한 지 약 3,500년 후 북미에 있는 영국 식민지 열세 곳의 주민들의 대표들은 필라델피아 시에 모여 **1776년 7월 4일 자신들은 더 이상 영국 왕의 신민이 아니라고 선언했다.** 이들의 독립선언은 함무바리의 것과 마찬가지로 신이 영감을 내린 것이라고 했다. 하지만 미국 신이 불러준 정의는 바빌론의 신들이 불러준 내용과는 상당히 달랐다.



"우리는 다음의 진리가 자명하다고 믿는다. 모든 사람은 평등하게 창조되었으며, 이들은 창조주에게 생명, 자유 행복의 추구를 포함하는 양도 불가능한 권리를 부여받았다." 미국 독립선언문은 함무바리 법전과 마찬가지로 원칙에 따라 행동하면 안정하고 평화롭게 살 수 있을 것이라고 약속한다. 



하지만 함무라비 법전과 미국 독립선언서는 우리에게 명백한 딜레마를 제공한다. **보편적으로 영원한 정의의 원리를 약속하지만 미국인은 모두가 평등한 반면 바빌론은 그렇지 않다. 사실은 모두가 틀렸다. 그런 보편적인 원리가 존재하는 장소는 오직 한 곳, 사피엔스의 풍부한 상상력과 그들이 지어낸 신화 속뿐이다. 이런 원리들의 객관적 타당성은 없다. **



생물학에 따르면 인간은 "창조"되지 않았다. 진화했다. 심지어 "평등하게 진화"하지도 않았다. 미국인들이 말하는 평등은 기독교 신앙에서 나온 것으로 기독교 신화를 믿지 않는 사람한테는 의미가 없다. 즉 "평등한 창조"는 "각기 다르게 진화"라고 바뀌어야 한다.

마찬가지로 **생물학에는 오로지 기관과 능력과 특질이 있을 뿐"권리"란 없다.** 이것들은 끊임없이 변화를 겪어 달라질 수 있으므로 "양도 불가능한 것"이 아니다. 타조는 나는 능력을 잃었다. "양도 불가능한 권리"는 "변이가 일어날 수 있는 특질"이라고 번역해야 한다.

따라서 미국의 독립선언문의 해당 구절을 생물학 용어로 번역하면 이렇게 된다.



"우리는 다음의 진리가 자명하다고 본다. 모든 사람은 각기 다르게 진화했으며, 이들은 변이가 가능한 모종의 특질을 가지고 태어났고 여기에는 생명과 쾌락의 추가가 포함된다."



함무라비도 자신의 위계질서 원리를 동일한 논리로 옹호할 수 있을 것이다. 가령 이렇게 말이다. 

**"나는 귀족, 평민, 노예가 날 때부터 다른 사람이 아니라는 사실을 알고 있다. 하지만 만일 그들이 다르다고 믿으면, 우리는 더 안정되고 번영한 사회를 창조할 수 있을 것이다."**

### 6_2. 진정한 신자들

자연의 질서는 안정된 질서다. 설령 사람들이 중력을 믿지 않는다 해도 내일부터 중력이 작용하지 않을 가능성은 없다. 이와 반대로 **상상의 질서는 언제나 붕괴의 위험을 안고 있다.** 이런 질서는 신화에 기반하며 신화는 사람들이 믿지 않으면 사라지기 때문이다. 이런 상상의 질서를 보호하기 위해 일부는 폭력과 강요를 사용한다.

군대, 경찰, 법원, 감옥은 상상의 질서에 맞추어 행동하도록 쉴 새 없이 돌아간다. 고대 바빌론 때는 '눈에는 눈'을 강요했고, 1860년 미국인들은 아프리카 노예의 자유권 때문에 전쟁을 치러야 했다.



하지만 상상의 질서는 폭력으로만 유지할 수 없으며 **진정으로 믿는 사람들이 필요하다. 주창자 대 다수가 인, 예, 효를 신봉하지 않았다면, 유교는 2천 년 넘게 이어질 수 없었을 것이다.** 역대 대통령과 의원 대다수가 인권을 신봉하지 않았다면 미국 민주주의는 250년간 지속되지 않았을 것이다. 만일 투자자와 은행가 대다수가 자본주의를 신봉하지 않았다면, 현대 경제 시스템은 단 하루도 유지하지 못했을 것이다.

### 6_3 교도소의 답장

사회를 지탱하기 위해서는 **위대한 신이나 자연법에 의해 창조된 객관적 실채라고 늘 주장해야 한다.** 사람이 평등하지 않은 것은 함무바리가 주장해서가 아니라 엔릴과 마르두크가 그렇게 명했기 때문이다. 사람이 평등한 것은 토마스 재퍼슨이 그렇게 말해서가 아니라 신이 그렇게 창조하였기 때문이다. 

또한 **사람들을 철저히 교육시켜야 한다.** 상상의 질서를 끊임없이 주입시켜야 한다. 평등을 믿는 요즘 사람들을 위해 노동계급이 입던 청바지를 부잣집 아이들도 입는다. 귀족한테 국한되었던 'sir, madam'는 공손한 편지에서 쉽게 볼 수 있다. 



**상상의 질서는 물질세계에 단단히 뿌리내리고 있다.** 오늘날 서구인인 개인주의를 신봉한다. 모든 인간은 개인이며 그 가치는 다른 사람이 어떻게 생각하느냐에 좌우되지 않는다는 것을 믿는다. 이런 신화는 현대의 집에도 적용된다. 아이들읜 자기만의 방이 있고 남의 눈에 띄지 않는 사적인 공간을 가져 최대한의 자율권을 가지도록 한다. 문을 닫고 자신이 좋아하는 스타의 포스터가 벽에 붙어 있으며, 옷을 여기저기 흩트려 놓고, 자기만의 공간을 만들면서 자기 내면의 가치를 표출한다.

반면 중세시대 아이들은 사자왕 리처드의 포스터를 붙일 수 없었고 잠금장치가 있는 개인 방이 없었다. 그런 환경에서 자란 아이들은 사람의 진정한 가치는 사회적 위계질서 속에서 어떤 위치에 속하느냐, 다른 사람이 어떻게 판단하느냐에 따라 결정된다고 결론을 내린다.



**상상의 질서는 우리 욕망의 형태를 결정한다.** 실상 모든 사람은 기존의 상상의 질서 속에서 태어났으며 태어날 때부터 지배적인 신화에 의해 욕망의 형태가 결정되었다. **"마음 내키는 대로 하라"는 권유 자체가 우리 마음속에 새겨진 것은 19세기 낭만주의 신화와 20세기 소비자 주의 신화의 결합을 통해서였다.** 20세기 소비자주의 신화에 의해 코카콜라는 "다이어트 코크, 기분 좋은 일을 하라"라는 문구를 내세웠다.

해외에서 휴가를 보내고 싶다는 흔한 욕망을 보라. 이런 욕망은 전혀 자연스럽지도, 당연하지도 않다. 침팬지 알파 수컷은 권력을 이용해 이웃 침팬지 무리로 휴가 갈 생각 따위는 하지 않을 것이다. 우리는 "새로운 경험이 어떻게 나의 시야를 넓히고 내 인생을 바꾸었는가"하는 낭만주의적 신화를 되풀이해서 듣는다. 다양성을 권하는 낭만주의는 소비지상주의와 꼭 들어맞는다. 양자의 결합은 현대 여행 산업이 기반으로 하고 있는 무한한 '경험의 시장'을 탄생시켰다. 고대 이집트 엘리트처럼, 대부분의 문화에 속하는 대부분의 사람들은 나름대로의 피라미드 건설에 삶을 바쳤다. 문화에 따라 피라미드의 이름과 형태와 크기가 달라질 뿐이다.



**상상의 질서는 상호 주관적이다.** 상상의 질서는 수 억 명 이상의 사람들이 공유하는 상호 주관적인 질서이며 이를 이해하려면 **'객관', '주관', '상호 주관'**이란 용어의 차이를 알아야 한다. 객관적 현상은 인간의 의식이나 믿임과는 독립적으로 존재한다. 주관이란 한 개인의 의식과 신념에 따라 존재한다. **상호 주관이란 많은 개인의 주관적 의식을 연결하는 의사소통망 내에 존재하는 무엇이다.** 그물망 속의 대부분의 사람이 죽거나 신념을 바꾸면 상호 주관적 현상은 변형되거나 사라진다.  **법, 돈, 신, 국가가 모두 그런 예이다.** 가령 푸조를 해체하려면 프랑스 법률체계처럼 그보다 더 강력한 뭔가를 상상해야 하고, 프랑스 법률 체계를 해체하려면 국가처럼 그보다 더 강력한 것을 상상해야 하고, 국가마저 해체하려면 그 보다 더 강력한 무언가를 상상해야 한다. 



**상상의 질서를 빠져나갈 수 없다.** 우리가 감옥의 벽을 부수고 자유를 향해 달려 나가도 실상은 더 큰 감옥의 더 넓은 운동장을 향해 달려 나가고 있는 것일 뿐이다.
