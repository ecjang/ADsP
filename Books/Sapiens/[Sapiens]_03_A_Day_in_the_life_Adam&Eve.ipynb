#### 1부. 인지혁명

# 3. 아담과 이브가 보낸 어느 날


오늘날의 진화심리학에 따르면 **현대인의 사회적, 심리적 특성 중 많은 부분이 농경을 시작하기 전 기나긴 시대에 형성되었다.** 식습관, 분쟁 성적 특징 모두 우리의 수렵채집 마인드가 현대의 산업사회와 상호작용한 결이다. 이런 환경 덕분에 물적 자원이 풍부해지고 수명도 길어졌지만 우리로 하여금 소외되고 압박받는다고 느끼게 만들었다.



고대의 수렵채집인들은 일부일처제가 아니며 **사유재산이나 아버지라는 개념이 없었다.** 무리의 여성은 동시에 여러 명의 남자와 성관계를 하고 밀접한 유대를 맺을 수 있었다. 또한 무리의 성인들은 모두 합쳐 아이들을 키웠다. 누가 자신의 친자식인지 확신할 수 없기 때문에 모든 아이에게 공평한 관심을 보였다.



고대 공동체 이론의 지지자들은 오늘날 결혼 생활을 특정 짓는 잦은 불륜, 높은 이혼율, 갖가지 콤플렉스들은 모두 **생물학적 소프트웨어와 맞지 않는 핵가족과 일부일처제로 살도록 강제한 탓이라고 한다.**



우리는 살면서 수백만 개의 **인공물**을 갖게 된다. 이런 인공물에 의존하면 고대 수렵채집인의 삶을 왜곡하게 된다. 따라서 이 문제를 시정하는 방법 중 하나는 현대의 수렵채집 사회를 살펴보는 것이다. 하지만 현대의 모든 수렵채집인은 이웃한 농경 및 산업 사회에 영향을 받았다는 점과, 해당 지역의 기후가 농경에 접합하지 않다는 문제점이 있다. 

고대의 수렵채집인들 사이의 인종적, 문화적 다양성이 상당했을 것이다. 농업혁명 전 수렵채집인은 5백만~8백만 명 정도였고 수천 개의 각기 다른 언어와 문화를 지닌 개별 부족으로 나누어져 있었을 것이다. 단 하나의 자연스러운 삶의 방식이란 없으며 **수많은 가능성 가운데 어떤 문화를 선택하느냐** 하는 문제가 있을 뿐이다.

### 3-1. 최초의 풍요사회

**개는 농업혁명 전 인간이 길들인 최초의 동물로 약 1만 5천 년 전에 가축화했다는 증거가 있다.** 어쩌면 이보다 더 이전일지도 모른다. 개는 사냥용, 경고용으로 이용했으며 시간이 흘러 보다 의사소통이 잘 되도록 진화했다. 그 결과 개가 죽으면 사람처럼 예식을 치르는 경우도 생겼다.



사피엔스는 계절의 변화, 동물들의 연례 이동, 식물의 성장주기에 따라 여기저기를 떠돌며 살았다. 식량이 풍부한 곳이라면 한 철을 같은 지역에서 보내기도 하였다. **그러면서 어떤 무리는 영구적인 정착지를 찾기도 하였다.** 식량을 말리고, 훈제하고, 냉동하는 기술 덕분에 오랫동안 머무는 것이 가능해졌다. 심지어 해산물이 풍부한 곳에서는 어촌이 형성되기도 하였다. 



생각해보면 고대 수렵채집인드른 현대인들에 비하면 **주변 환경을 파악하는 능력이 뛰어났을 것이다.** 계절이 어떻게 변하는지, 동물들의 습성이 어떤지, 어떤 식량이 영양가가 높은지, 언제 폭풍우가 오는지, 토끼 덧을 설치하는 방법, 옷을 고치는 법 등 넓고 다양한 지식을 지니고 있었다. 

또한 이들은 **후손에 비해 더 안락하고 보람찬 생활을 영위했을 것이다.** 현대인들은 일주일에 평균 40~45시간, 개발도상국은 60시간을 일하는 반면, 수렵채집인은 주 4일 한 번 정도 사냥을 하며 사냥은 3~6시간 정도밖에 걸리지 않았다. 그리고 모든 시기 대부분의 장소에서 다양한 영양소를 찾을 수 있어 굶어 죽거나 영양실조에 걸리는 일이 적었다. 건강에 유익한 다양한 음식을 먹고 주당 일하는 시간도 짧고, 전염병도 드물었으니 이를 두고 **농경 사회 이전의 수렵채집 사회를 최초의 풍요사회라고 불렀다.**

## 3-2. 말하는 유령

고대 수렵채집인들은 **애니미즘 신앙**이 일반적이었다. 애니미스트들은 인간과 다른 존재 사이를 가로막는 장애물이 없다고 믿는다. 사냥꾼은 사슴 무리 중 한 마리한테 말을 걸어 스스로 희생해달라고 요청할 수 있고 사냥에 성공하면 죽인 동물한테 용서를 구할 수 있다. 모든 소통의 행위의 특징은 말을 거는 대상이 **국지적 존재**라는 특징이 있다. 특정한 사슴, 나무, 시냇물, 유령 등이다.



우리는 고대 수렵채집인들에 대해서 무덤의 유물, 동굴벽화, 조각상들의 의미를 파 해치기보다 **그들에 대해 아는 것이 거의 없다는 것을 알아야 한다.** 그들이 애니미즘을 믿었다는 것을 가정할 수 있지만 이것으로 알 수 있는 정보는 많지 않다. 

### 3-3. 전쟁이냐 평화냐

규모가 크고 인구밀도가 상대적으로 조밀한 독립 수렵채집인을 관찰할 수 있는 기회는 2 차례뿐이었다. 19세기 북미 북서부와 19~20세기 초반 호주 북부에서이다. 고고학의 유물은 드물고 불분명하다. 인간의 유골이라고 해서 해석하기 쉬은 것은 아니다. 부러진 뼈는 전쟁에 의한 것일 수도 있고, 그냥 사고를 당한 것일 수도 있다. 



포르투갈에서 **농업혁명 이전의 유골 4백 구를 조사한 결과 폭력의 흔적은 단 2구뿐이었다.** 다뉴브 강 유적지에서는 4백여구 가운데 8구에서 폭력의 흔적이 나왔다. 만일 정말로 폭력에 의해 희생된 것이라면 4.5%가 폭력으로 죽었다는 의미이다. 20세기 2 차례의 세계대전과 인종학살로 인한 사망자의 비율이 5%였다. **20세기 정도와 비슷한 정도의 폭력을 겪었다는 의미이다.**



수렵채집인의 종교와 사회구조가 다양했던 만큼 이들의 폭력과 전쟁도 매우 다양하게 분포했을 것이다. 특정 시기, 특정 지역의 사람들은 풍요와 평화를 즐긴 반면, 다른 무리들은 격렬한 폭력으로 고통받았을지도 모른다.

### 3-4. 침묵의 커튼

뼈나 석기의 유물이 말하는 내용은 많지 않다. 인체의 해부학적 구조, 기술, 먹는 음식, 사회 구조 정도만 추측할 수 있을 뿐 무리의 동맹, 망자의 전령, 주술사 등의 내용은 알 수 없다. 이런 **침묵의 커튼은 수만 년의 역사를 감추고 있다.**



답을 얻을 수 없는 질문을 하는 것은 매우 중요하다. 그러지 않으면 우리는 인류 역사의 6만~7만 년을 '그 시기에 살았던 인류는 중요한 일이라고는 전혀 하지 않았다'는 핑계로 일축하고 싶어 질 수 있다. **하지만 그들은 중요한 일들을 많이 행했다. 특히 그들은 대부분의 사람들이 인식하는 것보다 주변 세계를 크게 바꾸어 놓았다.**
