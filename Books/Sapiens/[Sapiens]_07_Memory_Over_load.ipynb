#### 2부. 농업혁명

# 7. 메모리 과부화

세계 모든 곳의 강아지들이 벌이는 **개싸움의 규칙은 그들의 유전자에 각인되어 있다. 하지만 인간의 십 대에게는 축구 유전자가 없다.** 그럼에도 그들은 완전히 낯선 사람들과 게임을 할 수 있는데, 왜냐하면 이들이 축구에 대해 배운 **일련의 개념들이 완전히 같기 때문이다.** 이보다 규모가 더 큰 왕국, 교회, 무역망에도 똑같은 원리가 적용되지만, **수백만 명이 연관되는 대규모 협력시스템의 막대한 정보를 관리하고 저장해야 한다.**



농업혁명에 이어 유달리 복잡한 사회가 등장하면서 완전히 새로운 정보가 중요해졌다. 바로 숫자다. 수렵채집인은 나무에 달린 과일의 개수 따위를 따로 외울 필요가 전혀 없었다. 따라서 인간의 뇌는 숫자를 저장하고 처리하는데 적응하지 않았지만 커다란 왕국을 유지하려면 수학적 데이터가 핵심이었다. **이런 문제를 처음으로 극복한 것은 메소포타미아 남부에 살던 고대 수메르인이었다.** 기원전 3500~3000년 어느 시기, 익명의 수메르 천재들이 **'쓰기'라는 시스템**을 발명하여 뇌의 제약에서 벗어나 도시, 왕국, 제국의 출현에 이르는 길을 열었다.

----

### 7_1. 쿠심이 서명했다.

점토판에 눌러쓴 두 종류의 기호를 사용했으며 한 유형은 숫자를 나타냈다. 각각 1, 10, 60, 600, 3600, 36000을 나타내었다. 다른 유형은 사람, 동물, 사유품, 토지, 날짜 등을 나타내었다. 초기 쓰기는 사실과 숫자에 한정되었다. 많일 위대한 수매르 소설이 존재했더라도 점토판에 쓰이지는 않았다. **조상이 남긴 가장 초기의 메시지는 '보리 29,086자루 37개월, 쿠심'인데, 이 문장은 '37개월에 거쳐 보리 29,086자루를 받았다. 서명자 쿠심'일 것이다.**

### 7_2. 관료주의의 불가사의

마침내 메소포타미아 사람들은 단조로운 수학 데이터 이외의 것을 쓰고 싶어 졌다. 기원전 3000~2500년 사이에 문자 체계에 점점 더 많은 기호가 추가되어 **오늘날 쐐기문자라고 불리는 완전한 문자 체계로 점차 바뀌었다.** 또 다른 완전한 문자 체계는 기원전 1200년경 중국에서, 기원전 1000~500년경 중미에서 발달했다. 사람들이 점점 더 많은 것을 기록하게 됨에 따라, 특히 행정부의 기록보관소는 엄청난 덩치로 커지면서 새로운 문제가 등장했다. 



시간이 지나면서 점토판은 엄청난 양으로 불어났고 특정 점토판을 찾는 일이 너무나 힘든 일이다. 설령 찾는다 해도 가장 최신의 문서라고 확인하기 어렵고, 분실되거나 빗물에 깨끗이 씻겨 나갔다면 다시 복구할 수도 없다. 이를 위해 **목록 같이 조직화 방법, 복사기 같은 복사 수단, 컴퓨터처럼 빠른 검색법, 이를 사용할 수 있는 사서가 필요하다.** 하지만 이런 방법들을 발명하는 것은 문자를 발명하는 것보다 더 힘든 일이다. **수메르가 이집트, 고대 중국 잉카 제국이 달랐던 점은 문자 기록을 보관하고 목록을 만들고 검색하는데 뛰어난 기술을 개발했다는 점이다.**



관료제에서는 모든 것이 분리되어 있어야 한다. 그래서 서랍을 추가하고 지우고 재배열하는 일을 영원히 계속한다. 그런 서랍 시스템을 제대로 운영하기 위해서는 그는 사람으로서 생각하기를 중단하고 회계사로서 사고체계를 다시 정착해야 한다. 문자 체계가 인간 역사에 가한 가장 중요한 충격은 정확히 이것, **즉 인간이 세계를 생각하는 방식과 세계를 보는 방식이 점차 바뀌었다는 점이다.**

### 6_2. 진정한 신자들

자연의 질서는 안정된 질서다. 설령 사람들이 중력을 믿지 않는다 해도 내일부터 중력이 작용하지 않을 가능성은 없다. 이와 반대로 **상상의 질서는 언제나 붕괴의 위험을 안고 있다.** 이런 질서는 신화에 기반하며 신화는 사람들이 믿지 않으면 사라지기 때문이다. 이런 상상의 질서를 보호하기 위해 일부는 폭력과 강요를 사용한다.

군대, 경찰, 법원, 감옥은 상상의 질서에 맞추어 행동하도록 쉴 새 없이 돌아간다. 고대 바빌론 때는 '눈에는 눈'을 강요했고, 1860년 미국인들은 아프리카 노예의 자유권 때문에 전쟁을 치러야 했다.



하지만 상상의 질서는 폭력으로만 유지할 수 없으며 **진정으로 믿는 사람들이 필요하다. 주창자 대 다수가 인, 예, 효를 신봉하지 않았다면, 유교는 2천 년 넘게 이어질 수 없었을 것이다.** 역대 대통령과 의원 대다수가 인권을 신봉하지 않았다면 미국 민주주의는 250년간 지속되지 않았을 것이다. 만일 투자자와 은행가 대다수가 자본주의를 신봉하지 않았다면, 현대 경제 시스템은 단 하루도 유지하지 못했을 것이다.

### 7_3. 숫자라는 언어

정부나 기구, 회사의 결정에 영향을 미치고 싶은 사람은 숫자로 말하는 법을 배워야 한다. **전문가들은 심지어 '빈곤', '행복', '정직' 같은 개념도 숫자로 번역하려고 최선을 다하고 있다.**('빈곤선', '주관적 웰빙 수준', '신용등급'). 쓰기는 인간의 의식을 돕는 하인으로 탄생하였지만, 점점 더 우리의 주인이 되어가고 있다.
