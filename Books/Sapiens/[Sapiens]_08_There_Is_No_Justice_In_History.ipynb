#### 2부. 농업혁명

# 8. 역사에 정의란 없다.

우리는 상상의 질서와 문자체계 2가지의 발명을 통해 생물학적 틈을 매웠다. 하지만 **이런 협력망들의 출현은 의심스럽고 불안한 축복이었다. 그 그물을 지탱하는 상상의 질서는 중립적이지도 공정하지도 않았다.** 그 망은 사람들을 가상의 집단으로 나누었다. 상류층이 특권과 권력을 향유하는 동안 하류층은 차별과 압제로 고통받았다.


독립선언서에 서명한 많은 이들이 노예소유주였다. 이들은 서명과 동시에 노예를 해방하지도 않았으며 스스로를 위선자로 여기지도 않았다. 이들의 견해에 따르면 인간의 권리는 깜둥이와는 아무런 관련이 없기 때문이다.

**상상의 질서는 스스로가 허구에 근원을 두고 있다는 점을 부인하고 자연적이고 필연적이라고 주장하는 것이 역사의 철칙이다.** 함무라비는 노예가제가 신이 정해놓은 것이라고 보았다. 아리스토텔레스는 노예에게는 '맹종하는 본성'이 있고, 자유시민에게는 '자유로운 본성'이 있다고 주장했다. 이들의 사회적 지휘는 단순이 이들의 본성을 반영한 것일 뿐이라는 생각이었다.

하지만 우리가 아는 한 이런 위계질서는 모두 상상의 산물이다 정말로 브라만과 수드라가 윈시적 존재의 각기 다른 신체부위로 만들어진 것은 아니다. 두 계급의 차별은 약 3천년전 인도 북부에서 인간이 발명한 법과 규범에 의해 창조된 것이다. 현대 서구인들은 인종간의 위계질서가 있다는 생각은 비웃으라고 교육받는다. 

**불행이도 복잡한 인간사회에서 위계질서와 불공정한 차별이 필요한 것으로 보인다. 학자들이 알기로 대규모 사회치고 차별을 완전히 없앤 경우는 이제까지 없었다.** 사회의 구성들은 가상의 범주에 따라 분류하여 사회에 질서를 창조하는 일을 되풀이했다. 범주는 예컨데 귀족, 평민, 노예, 백인, 흑인 등으로 수백만명의 사람들의 관계를 조율했다. 

**위계질서는 한 가지 중요한 기능을 한다. 서로 모르는 사람들끼리 아는 사이가 되느날 시간과 에너지를 낭비하지 않고 서로 어떻게 대해야 하는지  알 수 있게 해준다는 것이다.** 물론 사회적 차별이 형성되는 데는 타고난 능력의 차이도 한 몫하지만 능력과 성격의 다양성은 보통 상상의 질서의 영향을 받기 마련이다. 첫째이자 중요한 점은 대부분의 재능에는 육성과 개발이 필요하다는 점이다.둘째는 다른 계층의 사람들이 정확히 같은 능력을 개발하여도 이들이 똑같이 성공할 가능성이 적다. 게임에 적용되는 규칙이 다른기 때문이다.

### 8_1 악순환

위계질서의 차이는 어디서 올까? 왜 인도의 전통 사회는 카스트에 따라, 오트만 사회는 종교에 따라, 미국은 인종에 따라 사람들의 등급을 나눌까? 힌두교의 카스트 제도가 형성된 것은 약 3천년던 인동라이아 사람들이 인도 아대륙을 침략해 현지인들을 복속시켰을 때이다. 이들 침략자들은 계층화된 사회를 건설하여 자신들은 윗자리를 차지하고 현지인들을 하인과 노예로 만들어버렸다.수가 적은 침략자는 사람들은 카스트로 구분하였고, 각 카스트별로 특정한 직업을 가지도록 했다. 이런 구별은 종교적인 산화와 관습의 일부가 되었다. 

힌두교의 계급제도와 여기서 수반되는 청결의 법칙은 인도 문화에 깊이 새겨져 인도아리아인의 침략이 잊힌 지 오랜 뒤에도 지속되었으며 카스트를 뒤석는 오염을 혐오했다. 새로운 직업이나 새로운 집단이 나타나면 반드시 어떤 종류의 카스트로써 인식되어야 힌두교 사회에서 정당한 지위를 얻을 수 있었고 그렇지 않으면 버림받았다.

인도의 민주정부가 그런 차별을 철폐하려고 온갖 노력을 다하면 카스트를 섞어도 오염되는 것은 아무것도 없다고 힌두교를 설득해왔지만 모두 허사였다.

### 8_2 미국에서 청결 문제

16세기에서 18세기에 유럽 정복자들은 아프리카에서 수백만 명의 노예를 미국으로 수입해 광산과 대규모 농장에서 일하게 했다. 미국에서 급성장하는 새로운 사회에서 유럽게 백인의 지배 카스트와 아프리카 흑인 종속 카스트로 나뉠 운명이 되었다. 하지만 사람들은 자신이 특정 인종이나 출신을 노예로 쓰는 이유가 오로지 경제적으로 유리하하기 때문이라고 말하고 싶어하지 않는다. 대신 경제적으로 성공하고 신앙심 깊고, 정의로우며, 객곽적인 사람이으로 보여지고 싶어 했다. 그래서 지배와 종속의 관계를 정당화하기 위해 종교적, 과학적 신화가 동원되었다. 신학자들은 아프리카인들니 노아의 아들인 햄의 자손이라고 주장했고, 햄은 그 아버지로부터 '네 자손들은 노예가 되리라'는 저주를 받았다.

산업혁명과 이민으 물결은 미국을 극도로 유동적인 사회, 가냉뱅이가 부자가 될 수 있는 사회로 만들었다. 만일 문제가 되는 것은 돈 뿐이었다면 인종 간의 첨예한 분리는 빠르게 흐려졌을 것이고, 거기에는 인종간 결혼이 적잖은 기여를 했을 것이다. 하지만 그런 일은 일어나지 않았다. 1865년이 되자 백인 뿐만 아니라 흑인들도 흑인에 대한 편견을 사실이라고 믿기 시작했다. 백인에 비해 지능이 낮고 게으르고 폭력적이고 문란하고 더럽다고 말이다.

편견은 시간이 흐르면서 점점 더 굳어졌다. 좋은 직업은 모조리 백인이 차지하고 있기 때문에 상대적으로 흑인이 실제로 열등하고 믿기 쉬었다. 악순환은 여기서 멈추지 않았다. 1876~1965년 시행되었던 짐 크로 법으로 흑인은 백인과 같이 공부할 수도, 먹을 수도, 잠을 잘 수도 없었다. 흑인한테 오염될까 봐 두려워서였다. 이런 악순환은 수세기 수천 년 지속되면서 역사적으로 우연히 발생한 질서에 불과한 상상의 위계질서를 지속시킬 수 있었다. 부당한 차볗은 시간이 흐르면서 개선되는 것이 아니라 더 심해질 수 있다. 돈은 돈 있는자한테 가난은 가난뱅이를 방문하는 법이다. 

대부분의 사회정치적 차별에는 논리적, 생물학적인 근거가 없으며, 우연한 사건이 신화에 뒷받침을 받아 영속화 한 것에 불과하다. 우리가 역사를 공부해야 하는 휼륭한 이유 중 하나가 이것이다.

### 8_3 그와 그녀

많은 사회에서 여성은 남성의 소유물에 불과했다 주인은 아버지인 경우가 가장 많았고, 남편이나 남자 형제일 때도 있었다. 많은 법 체계를 강간을 재산권 침해로 다루었는데, 달리 말해 강간의 피래자는 강간당한 피해자가 아니라 그 여성을 소유한 남성이라는 뜻이다. 어느 남자에게도 속하지 않은 여성을 강간하는 것은 전혀 범죄가 되지 않았다. 복잡한 거리에서 누군가 잃어버린동전 하나를 줍는 것은 도둑질로 취급되지 않는 것처럼 말이다. 고대 히브리인들은 이것을 타당하다고 보았다.

이런 사고방식은 고대 중동에서만 통용되는 것이 아니었다. 2006년 기준 53개국에서 아내는 남편을 강간죄로 기소할 수 없었다. 심지어 독일서도 1997년에서야 강간법이 개정되어 부부간에 강간이라는 법적 범주가 만들어졌다.

많은 사회가 일련의 속성을 남성성과 여성성에 결부시키지만 대체로 생물학적으로 분명한 근거는 없다. 기원전 아테네에서는 자궁을 가진 개인은 법적 지위를 가질 수 없었고, 평의회 의원이나 판사가 될 수 없었다. 자궁이 있으면 생물학적으로 이런 직업에 맞지 않는 것일까? 고대 아테네 사람들은 그렇게 생각했다. 

생물학적으로 결정되어 있는 것과 단지 사람들이 생물학적 신화를 통해 정당화 하려고 노력하는 것을 어떻게 구분할 수 있을까? "자연은 가능하고 문화는 금지하는" 기준으로 구분할 수 있다. 생물학은 여성에게 아이를 낳을 수있는 능력을 주었고 문화는 이를 의무로 지웠다. 생물학은 남자끼리 성관계를 즐길 수 있게 했고, 일부 문화는 그런 가능성을 실현하는 것을 금지했다.

생물학 과점에서 보면 사실 부자연스러운 것이란 것은 없다. 가능한 것은 처음부터 자연스러운 것이며 정말로 부자연스럽고 자연법칙에 위배되는 행동은 에초에 존재 자체가 불가능하다. 수고스럽게 남자에게 광합성을 금지하거나, 여성에게 빛 보다 빨리 달리지 못하게 하거나, 음전하를 띈 전자가 서로에게 끌리지 못하도록 금지하는 문화는 하나도 없다. 진실을 말하자면 '자연스럽다'는 단어는 기독교에서 나온 것이며 '자연을 창조한 신의 뜻에 맞는다'는 뜻이다. 우리가 사지와 장기를 신의 마음에 그렸던 목적에 맞게 사용하면 자연스러운 활동이라는 것이다. 하지만 진화에는 목적이 없다. 장기는 어떤 목적을 가지고 진화한 것이 아니며 그것의 사용방식은 끊임없이 진화한다. 인체의 장기 중에서 원형 상태로 수억 년 전 처음 등장했을 때 했던 일만 하고 있는 것은 하나도 없다.

### 8_4. 생물학적 성과 사회적 성

남자는 단지 XY염색체와 고환 같은 생물학적 속성을 지닌 사피엔스를 일컫는 것만이 아니다. 그보다는 그가 속한 사회사 상상하는 인간의 질서 상에서 특정한 자리에 딱 맞는 존재를 일컫는다. 예를 들어 정치참여, 투표권, 군복무를 부과한다. 여자도 마찬가로 XX염색체, 하나의 자궁, 많은 에스테로켄 호로몬을 지닌 사피엔스를 말하는 것이아닌 그 사회에서 여성다운 역할(육아), 폭력으로부터 보호, 남편에게 복종하는 의무를 지는 것을 의미한다. 생물학적 특징이 아닌 신화가 남녀의 역할, 권리, 의무를 규정하기 때문에 "남성성"과 "여성성"은 사회에 따라 크게 달랐다. 

남성의 일원이 되는 것은 간단한다. X,Y 염색체를 가지고 태어나면 된다. 하지만 남자나 여자가 되는 것은 매우 복잡하고 요구사항이 많은 프로젝트이다. 남성적이나 여성적인 특징이 생물학적인 것이 아닌 문화적인 것이기 때문이다. 게다가 이런 자격은 한 번 얻었다고 계속 안주할 수 있는 월계관도 아니다. 자신의 젠더를 평생 끊임없는 의례와 퍼포먼스로 증명해야 한다. 

그게 꼭 성공한다는 보장은 없다. 특히 남성은 자신의 남성성을 잃을까 봐 끊이 없이 두려워하며 살아간다. 역사를통틀어 남자는 "그는 진짜 남자야"라는 말을 듣기 위해 기꺼이 생명의 위험을 무릅쓰거나 심지어 목숨을 바쳐왔다.

### 8_5 남자가 뭐가 그리 좋을까?

농경사회에서는 남자를 여자보다 더 높게 평가하는 부계사회였다. 부계사회에서는 남자에게 남성적으로 생각하고, 행동하라고 가르친다. 물론 알파의 지위까지 올라란 여성이 한 줌 있기는 했다. 이집트의 클레오파트라, 중국의 추천무후, 영국의 엘리자베스 1세, 하지만 이들은 규측의 존재를 증명하는 예외에 해당한다. 엘리자베스 여왕이 치세였던 45년 내내 모든 의원은 남자였고, 육군과 해군 장교는 모두 남자였다.

가부장제는 너무나 보편적이었기 때문에 우연한 사건에 의해 촉발된 모종의 악순환의 결과일 수 없다. 심지어 1492년 콜럼버스의 미 대륙 상륙 이전에도 미 대륙과 아프로아시아의 대부분이 가부장 사회였다. 이전 수처년간 두 대륙이 전혀 접촉하지 않았는데도 말이다.

### 8_6 근력

가장 흔한 이론은 남자가 근력으로 여성을 강제로 굴복시켰다고 말한다. 남자는 힘히 세기 때문에 밭갈기나 추수처럼 힘든 노동이 필요한 업무를 독점할 수 있다고 말한다. 더분에 식략 생산량을 통제할 수 있었고, 이것이 정치적 영향력으로 나타났다는 것이다.

하지만 2가지 문제가 있다. 힘이 강하다는 것은 평균적으로만, 그리고 특정 종류의 힘에 대해서만 옳다. 일반적으로 여성은 남자보다 굶주림, 질병, 피로에 대한 저항력이 크고 남자보다 빨리 달리거나 더 강한 여성도 많다. 그리고 사제, 법률가, 정치인 처럼 근력이 필요없는 직업에서조차 배재되어 수공예, 가사노동처럼 힘든 노동에 종사했다는 점도 문제점이다. 

더 중요한 사실은 육체적인 힘과 사회적 권력 사이에는 직접적인 관련이 없다는 것이다. 이십대의 청년들이 더 힘이 세지만 육십대의 사람들이 이십대의 청년들에게 권력을 행사는 것이 보통이다. 수렵체집 사회에서 정치적 지배력을 지는 사람은 근육 조직이 아니라 사회성이 가장 뛰어난 사람이었다.

사실 역사를 보면 신체적 기량과 사회적 권력 사이에 반비례 관계가 성립하는 경우가 흔히 있다. 어쩌면 이것은 먹이사스에서 호모 사피엔스가 차지하는 지위를 반영하는 현상일지도 모른다. 먹이사슬의 중간쯤이던 우리가 최상위 자리를 차지할 수 있었던 것은 정신적, 사회적 기량 덕분이다.

### 8_7 사회의 쓰레기

또 다른 이론은 남성의 지배가 힘이 아니라 공격성의 결과라는 설명이다. 전시에 군대를 통솔하는 것은 남자들이었으며, 이런 통제권을 이용해 더 많은 전쟁을 벌였고, 그 횟수가 늘어날수록 통제력도 커졌다. 실제로 호로몬 연구 결과 남성이 여성보다 병사로 복무하기에 더 적합하다는 가정을 뒷받침 한다.

하지만 병사가 모두 남자라고 해서 전쟁을 관리하고 결실을 차지하는 사람도 남자라야 한다는 법이 있을까? 목화를 재배하는 노예가 흑인라고 해서 농장주도 흑인이라고 가정하는 것고 마찬가지다. 하지만 역사를 보면 귀족, 부유층, 고육받은 사람이 자동으로 장교로 임명되었으며 병사로 복무한 일 없이 고위 장교로 승진했다.

나폴레옹의 숙적 웰링턴 공장은 18세에 영국군에 들어갔을 때 즉각 장교로 임명되었고 "우리는 지상의 쓰레기들을 징집해 병사로 쓰고있다"고 편지에 쓰기도 하였다. 중국에서는 칼을 한 번도 휘둘러보지 않는 관리가 전쟁을 지휘하는 경우가 흔했다. 

보통 전쟁의 승리의 열쇠는 본국에서 평화를 유지하고, 해외나 동맹국을 구하고, 다른 사람들의 마음을 읽는 능력이 일반적이다, 따라서 공격적인 야수는 전쟁 지휘관으로서는 최악일 때가 많다. 군사적으로 무능했던 아우구스투스는 안정적인 제국 체계를 건설하는데 성공하여 자신보다 훨씬 더 뛰어난 장군이었던 율리우스 카이사르나 알렉산드로스 대왕이 이루지 못한 것을 성취했다. 현대 역사가들은 이런 업적을 그의 온화함과 관용이라는 미덕 덕분이라고 해석한다. 이런 진실을 알았더라면 여자들은 뛰어난 정치가나 제국 건설자가 되었어야 한다.

### 8_8 가부장적 유전자

세번째 설명으로는 진화과정을 통해 남녀가 서로 다른 생존 및 번식 전략을 발전시켰다는 주장이다. 남자들이 가임기의 여성을 놓고 서로 경쟁하면서 가장 야심차고 공격적이고 경쟁적인 남자가 후대에 유전자를 물려주었다는 것이다. 반면 여자는 남자를 찾는데 전혀 어려움이 없었지만 임신 기간중에 도움을 줄 수 있는 남자가 필요했으며 이를 위해 자신과 자녀의 생존을 보장하려면 남자가 내세운 조건을 받아들일 수 밖에 없었다. 세월이 흐르면서 순종적이고 집안일을 잘 돌보는 여자의 유전자가 후대에 전해지게 되었다. 

하지만 이런 접근법은 거짓임이 들어났다. 동물의 세계에서 의존적인 암컷들과 경쟁적인 수컷들 간의 역학관계의 결과로 모권 중심의 사회가 나타난 종들이 많다.(코끼리, 보노보) 동물이 이럴 수 있다면 호모 사피엔스가 못할 이유가 무엇일까? 사피엔스는 상대적으로 힘이 약한 동물이고 그 장점은 대규모로 협력하는 능력에 있다. 만일 그렇가면 여자들이 비록 남자한테 의존한다고 해도, 우월한 사회적 기술을 이용해 남자들의 허를 찌르고 조종하리라고 예상 할 수 있다. 

지난 세기를 거치면서 젠더의 역할의 역할은 커다란 혁명을 겪었다. 오늘날 사회는 남녀에서 동등한 법적 지위난 정치적 권리, 경제적 기회를 부여하고 있다. 그 뿐만 아니라 젠더와 성에 대한 가장 기본적은 개념들을 완전히 다시 생각하고 있다. 그 격차는 아직도 상당하지만 변화는 굉장한 속도로 진행되어 왔다. 이런 극적인 변화들 때문에 젠더의 역사가 그토록 혼란스러운 것이다. 가부장제가 생물학적 사실보다 근거 없는 신화에 기반을 둔 것이라면 이 제도가 보편적이고 안정된 이유가 대체 무엇일까?
