####  3부. 인류의 통합

# 12. 종교의 법칙

종교는 흔히 의견 충돌과 분열의 근원으로 여기지지만 **종교는 돈과 제국만큼 강력하게 인류를 통일시키는 매개체이다.** 모든 사회의 위계질서는 상상 속의 산물이기 때문에 모두 취약하기 마련이다. 종교는 이런 취약한 구조에 초월적인 정당성을 부여하는 역할을 한다. 종교는 법이 절대적인 최고 권의자가 정해놓은 것이라고 단언한다. 따라서 종교는 **'초인적 질서에 대한 믿음을 기반으로 하는 인간의 규범과 가치체계'**라고 정의할 수 있다.

- 종교의 기준
    1. **인간의 변덕이나 계약의 산물이 아닌 초인적 질서가 있다고 여긴다.** 프로축구는 종교가 아니다. 수맣은 규칙과 의식이 있지만 결국 인간이 발명한 것이다. 국게축구연맹은 언제라도 골문의 크기를 늘리거나 오프사이드 규칙을 폐기할 수 있다.
    1. **이런 초인적 질서를 기반으로, 스스로 구속력이 있다고 여기는 규범과 가치를 설정한다.** 많은 서구인들이 유령과 요정, 환생 등을 믿지만, 이런 믿음은 도덕과 행동의 기준의 원천이 아니므로 종교라고 할 수 없다.
    
종교를 통해 서로 다른 인간 집단을 합치기 위해서는 추가적인 속성이 필요하다. 언제 어디서나 보편적인 진리를 설파하고, 모든 사람한테 강력히 전파해야 한다. 즉, 종교는 **보편적인면서 선교적이어야 한다.** 실상 고대 종교는 지역적이고 배타적이었던 것에 반해 이슬람교나 불교는 보편적이며 선교적이다.

### 12_1. 양들을 침묵시키기

**초기 애니미즘적 종교는 그 세계관이 매우 국지적이고, 특정 장소나 기후현상의 특징을 강조하는 경향이 있었다.** 대부분의 수렵채집인은 자긴의 지역에서 평생을 보냈고, 먼 곳의 다른 계곡의 사람들에게 똑같은 규칙을 따르라고 설득하는 것은 무의미한 일이었다. 인더스강 유역에 사는 사람들이 갠지스강 유역에 선교단을 보내 흰꼬리여우를 사냥하지 말라고 설득하는 수고를 하지 않았을 것이다.

농업혁명은 종교혁명을 동반한 것으로 보인다. 농부들은 동식물을 소유하고 조작했다. **농업혁명이 미친 최초의 종교적 효과는 동식물을 영혼의 원탁에 앉은 동등한 존재에서 소유물로 끌어 내린 것이다.**

사람들이 평생을 수백 제곱킬로미터 안에서 보내는 한 그 지역의 정령만으로도 자신들의 필요를 충분히 만족시킬 수 있었다. 하지만 일단 왕국과 교역망이 확대되자, 왕국 전체나 교역 지대 전체를 아우르는 권력과 권위를 지는 존재가 필요해졌다. 이런 수요에 부응하고자 **다신교가 출현했다.** 

위대한 신의 등장으로 가장 크게 영향을 받은 것은 양이나 악마가 아니라 호모 사피엔스의 지위였다. 애니미즘은 인간을 세상에 살고 있는 많은 생명들중에 하나로 보는 반면, 다신교는 세상의 신들과 인간의 관계를 반영한다는 시각을 점점 키워나가기 시작했다. **우리의 기도와 희생과 죄업과 선행이 생태계 전체의 운명을 결정했다.**

옛 애니미증 체계에 속하던 다른 불운한 존재들은 그 지위를 잃어리고, 인간과 신들의 위대한 드라마 속에 엑스트라로 전락했다.



----

### 12_2. 우상숭배의 이점

**사실 대부분의 다신교는 다른 신들의 배후에 있는 최고권력을 인정했다.** 고전 그리스 다신교에서는 제우스, 헤라, 아폴론과 그 동료들은 모든 것을 다스리는 전능한 힘, 즉 운명의 여신(모이라, 아낭케)에게 복종한다. 북유럽 신들 역시 라그라오크의 격변으로 사라질 운명에 처해있다. 

일신교와 구분되는 다신교의 근본적인 통찰에 따르면 세상을 지배하는 최고권력은 관심이나 편견을 가지고 있지 않다. 모든 것을 아우르는 위치에서 보면 특정 왕국의 승리나 패배, 특정 도시의 번역과 쇠퇴, 특정인의 회복과 사망은 아무런 차이가 없는 일이다 

인간은 모든 것을 아루는 최고권력을 가신 신의 부분적인 힘만을 가지는 신들과 거래를 할 수 밖에 없었으며, 전쟁에서 이기고 질병을 낫기 위해 그들의 도움을 의지할 수 있다. 

**다신교도는 심지어 거대한 제국을 정복했을 때도 피정복민을 개종시켜는 노력을 하지 않았다.** 이집트인, 로마인, 아즈텍인은 오시리스, 유피테르, 우이칠로포치틀리에 대한 신앙을 전파하려 선교사를 파견하지 않았고, 이를 목적으로 군대를 파견하지도 않았다. 

로마인들이 오랫동안 관용을 거부했던 유일한 신은 린신교적이고 개종을 요구하는 기독교의 신이었다. 로마인인 제국의 수호신과 황제의 신성에 대한 경의를 표할 것을 기대했고, 기독교인은 이를 열렬히 거부하고 화해를 위한 모든 시도를 거절하는 데까지 나아가자 로마인들은 정치적 전복을 꾀하는 세력이라고 보아 박해로 대응했다.

신학논쟁은 무척이나 격렬해져 마침내 16~1세기에는 카톨릭교와 개신교도는 수십만명이나 서로 살해했다. 1572년 8월 24일, 선행을 강조하는 프랑스 가톨릭교도들은 하느님의 인간 사랑을 강조하는 프랑스 개신교를 공격했다. 성 바르톹로메오 축일의 대학살이라고 불리는 이 공격에서 5천~1만 명의 개신교도들이 살해되는 데는 채 하루가 걸리지 않았다. 로마 교황은 프랑스에서 전해진 소식을 듣자 몹시 기뻐하며, 이 사건을 기념하기 위한 축하 기도회를 조직하고 조르조 바사리에게 명해 **바티칸의 방 하나를 대학살에 대한 프레스로로 장식하게 했다.**

### 12_3. 신은 하나다.

최초의 일신교는 기원전 1350년경 이집트에서 나타났다. 파라오 아케나텐은 그저그런 신의 아텐신이 사실 우주를 지배하는 최고의 신이라고 주장했다. 하지만 그의 종교혁명은 성공하지 못했다. 그의 사후에는 아텐 숭배는 사라지고 옛 만신전이 돌아왔다. **다신교는 여기저기서 다양한 일신교를 잉태했으나, 이런 종교들은 주변부에 남아 있었다. 이런 단계를 우리는 '지역적 일신론'이라고 부를 수 있을 것이다.**

비약적인 돌파구는 기독교와 함께 왔다. 예수에 대한 좋은 말씀(복음)을 전세계로 전파할 필요가 있다고 추론했고 이런 바울의 주장은 비옥한 땅에 씨를 뿌렸다. 기독교인은 모든 인류를 겨냥하여 선교활동을 조직하기 시작했다. **이 비약적인 유대교 분파가 로마 제국을 접수한 것은 역사상 가장 이상한 사태 전개로 꼽힌다.**

일신론자는 다신론자에 비해 훨씬 더 광신적이고, 전도에 헌신하는 경향이 있다. 일신론자들은 자신들이 한 분 밖에 없는 절대신의 모든 메세지를 가지고 있다고 믿었기 때문에 다른 종교를 도저히 인정할 수 없었다. 기원후 1세기 초반, 세상에는 일신론자가 거의 없다싶이 했지만 오늘날에는 동아시아를 제외한 다른 지역 사라믇르은 대부분 이런저런 유일신을 충실히 믿고 있으며, 세계 정치질서 또한 유일신적 기초 위에 세워져 있다.

**하지만 다신교내에 애니미즘이 살아있듯이 일신교 안에서도 다신교 역시 살아남았다.** 사실 일신론 신핫은 최고신 이외의 모든 신의 존재를 부정하며, 감히 그런 잡신을 믿는 자한테는 지옥불과 유황을 퍼붓는 경향이 있다. 하지만 신학 이론과 실재 사이에는 언제나 틈이 있기 마련이다. 대부분의 사람들은 일신교 사상을 완전히 이해하기 어려웠다. 세상을 '우리'와 '그들'로 나누었고, 우주의 최고 권력이 자신들의 세속적 욕구에 비해 너무 멀고, 낯설다는 시각을 견지했다. **기독교는 성자들로 구성된 나름의 만신전을 발달시켰는데, 이것은 다신교의 만신전과 거의 다를 바가 없다.**

성 엘모는 굴뚝 청소부들을 보호했고, 성 마태오는 괴로워하는 세금 징수관들에게 도움을 주었고, 두통이 있다면 성 아가티우스에게 기도해야 하지만, 치통을 앓고 있다면 성 아폴로니아가 훨씬 더 잘 맞는 기도 대상이었다.

### 12_4. 선과 악의 싸움

**다신교는 일신교만 낳은 것이 아니라 이신교도 낳았다. 이신교는 서로 반대되는 두 힘 즉, 선과 악을 믿는다.** 인신교는 악의 문제에 대해 간명한 답을 주기 때문에 매력적이다. 

"세상에 왜 악이 존재하며, 왜 고통이 존재하는가?" 이런 의문에 답을 하기 위해서는 일신론자는 곡예를 부려야 한다. 전지전능한 하느님이 세상에 이토록 많은 고통을 허락하는 이유는 무엇이란 말인가? 널리 알려진 설명에 따르면 인간에게 자유의지를 허락하는 신의 방식이라고 한다. 하지만 이 또한 많은 의문이 꼬리를 문다. 반면 이신론자는 악을 설명하기 쉽다. 세상에는 독립된 악의 힘이 돌아다니고, 악의 힘은 나쁜일을 저지른다.

하지만 이신론자는 악의 문제를 풀어주지만 세상의 질서 문제 앞에서는 당황하게 된다. 세상에 선과 악이 있다면 이 둘을 관장하는 법칙을 정한 존재는 누구인가? 요약하자면 **일신론자는 질서를 설명하지만 악 앞에서는 쩔쩔맨다. 이신론자는 악을 설명하지만 질서 앞에서는 당황한다.**

이신교는 1천년 이상 번성했다. 기원전 1500년에서 기원전 1000년 사이의 어느 시기에 조로아스터란 이름의 예언자가 중앙아시아의 어느 지역에서 활동했다. 그의 교리는 시간이 지나 마침내 조로아스터교가 되었다. 그 신봉자들은 세상을 선신인 아후라 마즈다와 악신인 앙라마이뉴 사이에서의 우주적 싸움으로 보았다. 이 종교는 고대 페르시아 아카메네스 제국에서 중요한 종교였고, 나중에는 사산 제국의 공식 종교가 되었다. 이후 중동과 중앙아시아에서 발흥한 거의 모든 종교가 되었다. 

하지만 마니교도들은 로마에 영혼을 빼았겼고, 조로아스터교를 신봉한 사산 제국은 일신교를 믿는 무슬림들에게 무너졌다 이렇게 해서 이신교의 파도는 잦아들었지만 **일신교의 물결이 정말로 이신교를 싹 쓸언낸 것은 아니다. 유대교, 기독교, 이슬람교는 이신교에서 수많은 신앙과 관례를 흡수하였고, 오늘날 일신교라고 부르는 것의 가장 기본적인 사상 일부는 사실 그 기원이나 정신이 이신교적이다.** 기독교인이 악마라고부르는 것이 그런 존재이다. 이 존재는 선한 신에게 대항해 독자적으로 싸울 수 있고, 신의 허락 없이 파괴를 불러 올 수 있다.

인간에게는 모순을 믿을 수 있는 놀라운 능력이 있다. 그러므로 수백만의 경견한 기독교인, 무슬림, 유대교인들이 전능한 신과 독립적인 악마를 둘 다 믿는다고 해서 놀라 울 것이 없다. 

일신론자들은 악의 문제를 다루는데 도움이 된다는 이유로 이신론자들의 이분법에 매혹될 수밖에 없었다. 이 대립은 결국 기독교와 무슬림 사상의 초석이 되었다. 천국과 지옥에 대한 믿음 역시 그 기원은 이신론에 있었다. 구약에서는 이런 믿음의 흔적조차 없다. **사실 일신론은 역사에서 나타났듯이 이신론과 이신론, 다신론, 애니미즘 유산이 하나의 신성한 우산 밑에서 뒤섞여 있는 만화경이다.**

종교학자들은 이처럼 서로 다르고 심지어 상충하기까지 하는 사상을 동시에 인정하는 행위와 각기 다른 원천에서 가져온 의례와 관례를 혼합하는 행위에 대한 명칭으로 **제설혼합주의**를 썼다. 실제로 제설혼합주의야말로 단 하나의 위대한 세계종교일지도 모른다.

### 12_5. 자연의 법칙

지금까지 알아본 종교는 모두 초자연적인 실체에 대한 믿음에 초점을 맞추고 있다. 반면 기원전 1000년부터 완전히 새로운 종류의 종교가 아프로아시아로 퍼져나가기 시작했다. **인도의 자이나교와 불교, 중국의 도교와 유교, 지중해 분지의 스토아철학, 견유철학, 에피쿠로스주의와 같은 신생 종교들의 특징은 신을 섬기지 않는다는 점이다.**

불교의 중심은 신이 아닌 인간, 바로 고타마 싯다르타다. 불교 전통에 의하면 기원전 500년경 히말라야에 있던 작은 왕국의 후계자였다. 고타마는 29세에 가족과 재산을 뒤로하고 한밤중을 빠져나왔다. 그는 길을 찾으며 집 없는 방랑자로 인도 북부를 구석구석 떠돌았다. 그는 6년에 걸쳐 인간의 번뇌의 핵심과 원인과 치유법에 대해 명상했고, **마침내 그 번뇌의 원인은 불운이나 사회적 불공정, 신의 변덕이 아니었음을 깨달았다. 번뇌는 사람의 마음이 행동하는 패턴에서 일어나는 것이었다.**

가장 위대한 왕이라도 슬픔과 번민으로 부터 끊임없이 달아나며 더 영원히 큰 즐거움을 뛰쫓는 번뇌 속에 살 운명이다. 고타마는 이런 악순환에서 벗어나는 방법을 깨달았다. **즐거운 일이든 슬픈일 일이든 마음이 사물을 있는 그대로 이해할 수 있다면 거기에는 고통이 없다.** 불이 완전히 꺼지면 집차근 완벽한 만족과 평온의 상태와 자리를 바꾸게 되는데 이것이 바로 **열반**이다. 열반은 문자 그대로 '불 끄기'라는 뜻이다. 열반에 이른 사람은 모든 고통에서 해방된다. 집착이 없는 사람은 고통받지 않는다.

불교는 신들의 존재를 부정하지 않는다. 신들을 비와 승리를 가져다줄 수 있는 강력한 존재로 묘사한다. 하지만 이 신들도 집착에서 고통이 일어난다는 법칙에 아무 영향을 주지 못한다. 어떤 사람의 마음에 집착이 일어난다면 우주의 그 어떤 신도 그를 번뇌에서 구해주지 못한다.

시간이 흐르면서 여러 불교의 분파들이 부처들과 보살들로 구성된 만신전을 발전시켰다. 신을 숭배하는 대신에 많은 불교도들은 이런 깨달은 자들을 숭배하기 시작했다. 이들에게 열반에 이르도록 도와달라고 요청할 뿐만 아니라 세속의 문제를 해결해달라고 빌었다.

### 12_6. 인간숭배

근대는 강력한 종교적 열정의 시대, 전대미문의 표고 노력과 역사상 가장 피비린내라는 종교전쟁의 시대였다. **수많은 자연법칙 종교가 새로 등장했다. 자유주의, 공산주의, 민족주의, 국가주의가 그런 예이다.** 불교도들은 자연법칙을 카를 마르크스, 프리드리히 앵겔스, 블라드미르 일리치 레닌에 의해 발견되었다고 믿었다. 그리고 다른 종교와 마찬가지로 공산주의에는 **경전과 예언서**가 있다. 프롤레타리아의 궁극적 승리와 함께 역사는 곧 종말을 맞을 것이라고 예언한 **마르크스의 [자본론] 같은 책이다.** 더구나 나름의 기념일과 축제가 있는데 5월 1일 노동절과 10월 혁명 기념일이 그런 예이다.

이들은 일신교나 대중불교 못지않게 혼합적이다. 일신교도들이 악마를 믿을 수 있었던 것처럼, **오늘날 전형적인 미국인은 동시에 국수수의자면서 자유시장 자본주의자면서, 자유주의적 인본주의자일 수 있다.** 

유신론적 종교는 신에 대한 숭배를 초점을 맞춘다. 인본주의적 종교는 인간, 좀 더 정확히는 호모 사피엔스를 숭배한다. 인본주의는 호모 사피엔스에게는 특유의 성질이 있고, 이 성질은 다른 모든 동물들보다 다른 모든 현상의 성질과 근본적으로 다르다는 믿음이다.

모든 인본주의자는 인간성을 숭배하지만 그에 대한 정의는 각기 다르다. '인간성 Humanity'의 정확한 정의를 두고 크게 3가지의 분파로 나뉘었다. 오늘날 가장 중요한 분파인 **'자유주의적 인본주의'로 인간성은 개별 인간의 속성이며, 개인의 자유는 더할 나위 없이 신성하다고 믿는다.** 자유주의적 인본주의의 주된 계명들은 이런 인간성이 지닌 자유를 침입이나 손상으로부터 보호하기 위한 것이다. 이런 계명들을 통칭하여 **'인권'**이라고 부른다.

다른 분파는 **사회주의적 인본주의**이다. 사회주의자들은 **인간성을 개인주의적인 것이 아닌 집단의 것이라고 믿는다.** 자유주의자들은 개개인의 최대한의 자유를 추구하는 반면, 사회주의적 인본주의는 모든 인간의 평등을 추구한다. 불평등은 인간의 존엄성에 대한 최악의 모독이다. 

전통적 일신론의 속박에서 벗어난 유일한 인본주의는 **진화론적 인본주의**로 가장 유명한 예는 **국가 사회주의 즉, 나치**이다. 나치는 인류를 보편적이고 영원한 무언가가 아닌 진화하거나 퇴화할 수 있는 변하기 쉬운 종으로 보았다. 인간은 초인으로 진화할 수 있고, 인간 이하로 퇴화할 수 있었다. 나치에 따르면 아리아인은 최상의 자질을 가지고 있었고 인류를 초인류로 바꿀 수 있는 잠재력이 있었다.

나치는 인간을 혐오하지 않았다. 나치가 자유주의, 공산주의와 싸운 것은 그들이 오히려 인간을 찬양하며 인류의 위대한 잠재력을 믿었기 때문이다. 하지만 가장 우수한 자들만 생존하고 번식하게 해야 한다고 주장했다. 

히틀러와의 전쟁이 끝난 후 60년간, 인본주의를 진화와 연관시키는 것은 금기였다. 생물학적인 방법에 의한 호모 사피엔스의 '업그레이드'를 옹호하는 것도 마찬가지였다. 하지만 요즘은 이런 프로젝트가 다시 진행되고 있다. 하급인종이나 열등한 집단을 멸절시키자고 말하는 사람은 없지만, **많은 사람이 인간 생물학에 대한 우리의 해박한 지식을 이용해 초인간을 만드는 문제를 심사숙소 하고 있다.** 이와 함께 자유주의적 인본주의 신조와 생명과학의 최근 발견 사이에 엄청난 간극이 벌어지고 있다.

인간이라는 유기체의 작동방식을 연구하는 과학자들은 거기서 아무런 영혼도 발견하지 못했다. 인간의 행동은 자유의지가 아니라 호르몬, 유전자, 시냅스에 의해 결정된다는 주장을 펴는 과학자가 점점 늘고 있다. 우리는 생물학을 법학과 정치학으로부터 구분하는 벽을 과연 언제까지 유지할 수 있을까?
