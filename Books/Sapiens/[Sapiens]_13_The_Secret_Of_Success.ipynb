####  3부. 인류의 통합

# 13. 성공의 비결

다수의 작은 문화에서 몇 개의 큰 문화로, 마지막에서는 전 지구적 사회로 이행되는 것은 아마도 인간사 역학에 따른 필연적 결과일 것이다. 하지만 지구촌 사회가 필연적이라고 해서 최종 결과가 지금 우리가 사는 특정한 종류의 지구촌 사회처럼 되어야 한다는 법은 없다. 

왜 영어가 이렇게 널리 쓰이게 되었으까? 세계에 기독교 신자는 약 20억명, 무슬림은 12억 5천만명, 조로아스터교인은 15만명인반해, 마니교 신자는 0명인 이유는 무엇일까? 역사가 다시 진행되어도 매번 일신교가 등장하고 이신교가 쇠퇴할까? 이런 실험은 불가능하지만 역사의 핵심적인 특징을 살펴보면 단서를 찾아볼 수 있다.

### 13_1. 사후 깨달음의 오류

4세기 시작 무렵 로마 제국 앞에는 다양한 종교적 선택의 가능성이 펼쳐져 있으며 다채로운 다신교를 고수할 수 있었다. 아니면 당대에 있었던 수많은 종교들, 이를 테면 마니교, 미트라교, 이시스교, 키벨레교, 조로아스터교, 유대교 심지어 불교도 포함되었다. 하지만 콘스탄티누스는 왜 예수를 선택했을까? 역사학자들은 여러 추측을 내놓았지만 확정적인 답을 내놓지 못했다. 

**'어떻게'를 서술하는 것과 '왜'를 설명하는 것은 뭐가 다를까?** '왜'를 설명하는 것은 다른 가능성이 있음에도 불구하고 이 사건이 일어났는지 **'인과관계'**를 찾는 것을 의미한다. 

학문 분과로서의 역사학의 가장 큰 특징은 특정한 역사 시대를 알면 알수록 왜 파일 일이 그런 식으로 전개되었으며 다른 식으로 전개되지 않았는지 설명하기 점점 더 어려워진다는 점이다. **특정 시대에 대해 피상적인 지식만 있는 사람들은 실제로 실현될 가능성에만 초점을 맞추는 경향이 있다.** 이들은 사후 깨달음을 근거로, 어째서 그런 결과가 필연적으로 일어날 수 밖에 없었는지를 증명도 반증도 불가능한 이론으로 설명한다. 반면 해당 시대에 대해 더 깊이 알고 있는 사람들은 실제로 진행되지 않는 경과를 훨씬 많이 인식하고 있다.

특히 동시대 사람들에게 아주 희박해 보였던 가능성이 종종 실현되곤 한다는 사실을 강조할 필요가 있다. 306년 콘트탄티누스 대제가 제위에 올랐을 때, 기독교는 비밀스러운 동방의 분파에 지나지 않았다. 당시에 기독교가 로마의 국교가 될 것이라고 했다면 사람들은 비웃었을 것이다. 마치 오늘날 2050년이 되면 힌두교의 하레 크리슈나 교단이 미국의 국교가 될 것이라고 말하는 것과 마찬가지다.

이는 모든 일이 가능하다는 말이 아니다. **지리적, 생물학적, 경제적 힘은 제약을 만들어 낸다.** 하지만 그 제약 속에서도 어떤 결정론적 법칙에도 메이지 않는 것처럼 보이는 놀라운 일이 전개될 여지는 매우 많다.

역사는 이름바 **2단계 카오스계**이다. 카오스계에는 두 종류가 있다. 1단계 카오스는 자신에 대한 예언에 반응하지 않는 카오스다. 예를 들면 날씨 같은 것이다. 2단계 카오스는 **스스로의 예측에 반응하는 카오스다.** 사장이 그런 예이다. 만일 내일의 석유가격을 정확히 예측하는 프로그램이 있다면 시장의 변화가 생기고 그에 따라 가격이 필연적으로 변동하게 된다.

정치 또한 2단계 카오스이다. 소련 연구잗ㄹ은 1989년 혁명을 예측하지 못했고, 중동 전문가들은 2011년 '아랍의 봄' 혁명을 예측하지 못했다. 이를 두고 비난하거나 혹평하는 사람이 많지만, 이런 비난은 공정하지 못하다. 혁명은 그 정의상 예측이 불가능하다. 예상 가능한 혁명은 발생하지 않는다.

**역사를 연구하는 것은 미래를 알기 위해서가 아니라 우리의 지평을 넓히기 위해서이다.** 우리의 현재 상황이 자연스러운 것도 필연적인 것도 아니라는 사실을 이해하기 위해서이다. 그 결과 우리 앞에는 우리가 상상하는 것보다 많은 가능성이 있다는 것을 이해하기 위해서이다.

### 13_2. 역사의 여신은 장님

역사의 선택을 설명할 순 없지만 그 선탣에 대해 매우 중요한 발견을 할 수 있다. 역사의 선택은 인류를 위해 이루어지는 것이 아니라는 사실이다. 기독교가 마니교보다 더 나은 선택이었는지, 아랍 제국이 페르시아의 사산 왕조보다 더 도움이 되었다는 등의 증거는 없다. 우리에게는 그런 **이익을 측정할 척도가 없기 때문이다.** 어쩌면 기독교나 이슬람교가 사라지고 패배했더라면 우리는 더욱 잘 살았을지도 모른다.

문화를 일종의 정신적 감염이나 기생충으로 보는 견해가 잇다. 우연히 출현해서 자신이 감염시킨 모든 사람을 이용하는 정신의 기생충에 더 가깝다. 이런 접근법은 때로 문화 구성요소학, 혹은 **밈 연구**로 불린다. 유기체 진화가 '유전자'라 불리는 정보 단위의 복제에 기반을 둔 것과 마찬가지로, 문화적 진화는 '밈'이라 불리는 문화적 정보 단위의 복제에 기반을 두고 있다. 

대부분은 인문학자는 이런 밈 연구를 멸시하지만 밈 연구의 쌍둥이 자매 격인 **포스트모더니즘**을 고수한다. 이들은 담론을 들먹이지만 이들역시 문화는 인간의 이익과 무관하게 스스로 퍼져나가는 존재라고 생각한다.이들은 민족주의를 19,20세기에 퍼져서 전쟁, 압제, 증오, 인종청소를 일이킨 치명적인 전염병으로 묘사한다.

사회과학에서도 **게임이론** 비호 아래 비슷한 주장이 흔히 이야기 된다. 게임이론이란 다수가 참여하는 게임에서 어떻게 모두에게 해가 되는 시각과 행동 패턴이 뿌리는 내리고 퍼져나가는지를 설명해준다. 유명한 예가 군비경쟁이다. 군비경쟁은 하나의 행동 패턴으로 이 나라에서 저 나라로 바이러스 처럼 퍼져나가 모두에게 해를 끼친다. 하지만 스스로에세는 이롭다. 생존과 번식이라는 진화적 기준에서 보면 그렇다.

1500년경 역사는 가장 중대한 선택을 했다. 인류의 운명뿐만 아니라 지구상의 모든 생명체의 운명을 바꿀 수 있는 선택이었다. 우리는 이것을 **과학혁명**이라고 부른다. 그 혁명은 서유럽에서, 아프로아시아의 서쪽 끝에 있는 커다란 반도에서 시작되었다. 그때까지 역사에서 중요한 역할을 전혀 하지 못하던 지역에서 말이다.
