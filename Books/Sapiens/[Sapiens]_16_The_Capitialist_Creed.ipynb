####  4부. 과학혁명

# 16. 자본주의의 교리

인류 역사 대부분의 기간 동안 경제는 대체로 같은 규모를 유지해왔다 물론 지구총샌산은 증가했지만 이것은 대체로 인구 팽창과 정착지의 확대 덕분이다. 1인당 생상은 정체 상태였다. 하지만 근대에 와서 바뀌었다. 1500년 재화와 용역의 지구총샌산은 약 2,500억 달러였는데, 오늘날 이 수치는 60조 달러까지 증가했다. 더욱 중요한 것은 1500년간 연간 1인당 총생산은 550달러였지만 오늘날 모든 남녀와 어린이가 1인당 연평균 8,800달러를 생산한다는 점이다.

**인간의 상상력이 지닌 놀라운 능력에게 바치는 헌사다.** 은행과 경제 전체를 살아남게 하고 꽃피게 하는 것은 바로 우리 미래에 대한 신뢰다. 오로지 이 신뢰가 세계의 돈 대부분을 뒷받침한다고 볼 수 있다. 모든 기업은 이런 상상된 미래에 대한 신뢰 위에 세워져 있다.

돈은 무엇인든 대표할 수있고 무엇이든 다른 거의 모든 것으로 바꿀 수 있기 때문에 대단한 존재이다. 하지만 근대이전에는 이 능력이 제한적이었다. 대부분의 경우 돈이 대표할 수 있는 것은 현실에 존재하는 것뿐이었다.

**사람들은 '신용'이라 불리는 특별한 종류의 돈이 상상속의 재화를 대표하는데 동의했다. 신용은 미래를 비용으로 삼아 현재를 건설할 수 있게 해준다.** 만일 신용이 그토록 놀라운 것이라면, 어째서 아무도 좀 더 일찍 그런 생각을 하지 못했을까? 이런 저런 종류의 신용이 과거 모든 문화권에 존재했으며 수메르까지 거슬러 올라간다. 즉 신용을 생각하지 못한 것은 아니지만 그 신용을 확장하려는 경우가 드물었다.

당시 사람들은 부의 총량은 늘어나지 않고 한정되어 있다고 믿었다. 더 많은 부를 창출하는 것은 위험한 것으로 제로섬 게임으로 생각했다. 즉 특정 빵집의 이익이 증가한다는 것은 다른 빵집의 희생을 통해서만 가능하다는 논리이다. **파이를 자르는 방법은 많았지만 파이를 더 키울 수 없없다. 많은 문화권에서 돈을 많이 버는 것은 죄악이라고 한 이유가 여기에 있다.**


신용은 오늘의 파이와 내일의 파이의 차이이다. 만일 파이의 크기가 늘 똑같다면 왜 외상을 주겠는가? 근대 이전에는 대출 받기 힘들었고, 기업가는 새 빵집을 열기 어려웠고, 왕들은 세금과 관세를 걷어 왕궁을 짖고 전쟁을 일으켰다. 이것은 모두에게 부정적이었다. 신용이 제한되어 신규 사업의 자금 조달이 어려웠고 경제는 성장하지 못했다. 사람들은 지금까지 성장이 없었으니 앞으로도 없을 것이라고 판단했다.

### 16_1. 커지는 파이

그때 과학혁명과 진보라는 개념이 도래했다. 기존의 빵집을 망치지 않고 초코 케이크와 크루아상을 전문으로 하는 제과점을 열 수 있으며, 당신을 가난하게 만들지 않고 나는 부자가 될 수 있다. 지구상의 파이는 더 커질 수 있다라는 아이디어는 사람들로 하여금 미래를 신뢰하게 만들었다. **신뢰는 신용을 창조했고, 신용은 현실 경제를 성장시켰으며, 성장은 미래에 대한 신뢰를 강화하고 더 많은 신용을 향한 길을 열었다.** 지구상의 파이가 점점 커지고 있다는 믿음은 결국 혁명이 되었다.

1776낸 애덤 스미스는 경제학 역사에서 중요한 [국부론]을 썼다. 제1권 제8장에서 민간기업의 수입 증대는 공동체의 부와 번영을 늘리는 기초가 된다는 결론이 나온다. 이런 주장은 인류 역사상 가장 혁명적인 아이디어에 속한다. 경제적 관점에서뿐 아니라 도덕적, 정치적 관점에서 더더욱 혁명적이다. **스미스는 탐욕이 선한것이며 내가 부자가 되는 것인 나만이 아닌 모두에게 이득이 된다고 했다. 이기주의가 곧 이타주의라고.**

스미스의 이론에서 사람들은 이웃의 것을 빼앗아 부자가 되는 것이 아니라 전체 파이를 늘림으로써 부자가 된다. 파이가 커지면 모두에게 이득이 되기 때문에 부자는 사회헤서 가장 쓸모있고 인정 많은 사람이다. 모두에게 이익이 되도록 성장의 바퀴를 돌리는 사람이기 때문이다.

하지만 이 모든 것에는 전제가 있다. 부자가 자신의 수익을 비생산적인 활동이 아닌 공장을 짖고, 사람들을 고용하는데 사용해야 한다는 것이다. 그래서 '수익이 늘면 직원을 고용할 것'이라고 했지 '수익이 늘면 스쿠루지는 돈을 숨겨서 매번 혼자서 세어볼 것이다'라고 하지 않았다. **새로운 자본주의 교리에서 가장 신성한 제1규율은 "생산에 따른 이윤은 생산 증대를 위해 재투자 되어야 한다"이다.**

자본주의가 자본주의로 불리는 이유가 여기에 있다. '자본'을 단순한 '부'와 구별하는데 **자본이란 생산에 투자되는 돈과 재화와 자원을 의미한다.** 반면에 부는 땅에 묻혀 있거나 비생산적인 활동에 낭비된다. 비생산적인 피라미드에 자원을 쏟아 붇는 파아로는 자본주의자가 아니다.

근대 이전의 사람들은 생산이 어느정도 일정하다고 생각했다. 무슨일을 하든 생산이 늘지 않을 것이기 때문에 귀족들은 과시적인 소비를 하였다. 근대에 이르러 귀족은 자본주의 신조를 믿는 새로운 앨리트인 회장, 주식거래인, 기업 경영자 등에게 추월당했다. 그들은 회의에서 회의로 뛰어다니며 자본을 어디로 투자해야 좋을지 고민하고, 자신이 보유한 유가증권의 등락을 계속 추적한다. 비싼 양복을 입고 고속 비행기를 타고 다니지만 생산을 늘리기 위해 투자하는 액수에 비하면 아무것도 아니다.



**자본주의는 경제가 어떻게 기능하는지에 대한 이론으로서 시작되었다.** 하지만 자본주의는 경제적 교리를 넘어서 하나의 윤리가 추가되었는데 바로 **'경제성장이 최고의 선'**이라는 것이다. 정치와 자유, 심지어 행복도 경제성장에 좌우되기 때문이다. 이 새로운 종교는 현대 과학의 발전에도 영향을 미쳤다. 특정 과학 프로젝트에 투자할 때 보통 '이 프로젝트는 우리의 생산량과 수익을 얼마나 늘려줄 것인가?' 라고 질문하기 때문이다.

반대로 자본주의의 역사는 과학을 고려하지 않으면 이해할 수 없다. 근현대 기간 계속된 급성장은 과학자들이 몇 년마다 새로운 발견이나 장치를 발명했기 때문이다. 아메리카 대륙, 내연기관, 유전자 복제 양 같은 것들이다. 은행과 정부는 돈을 찍어내지만 궁극적으로 그 비용을 부담하는 것은 과학자 들이다. **만일 거품이 터지기 전에 연구실들이 그 기대를 충족시키 못한다면, 우리 미래는 무척 힘들어 질 것이다.**

###  16_2. 콜롬버스가 투자자를 찾는다.

신용은 근대 유럽에서 발명된 것이 아니다. 초기 유럽 자본주의는 아시아의 경제발전과 관련되어 있었다. 18세기 후반까지 세계 최대의 경제적 실세는 아시아였다. 하지만 중국, 인도, 무슬림 사회에서 신용의 역할은 크게 작용하지 않았다. 반면 유럽에서는 왕과 귀족들의 상인의 사고방식을 따르기 시작했고 상인과 은행가가 지배 엘리트가 되었다. 자본자가 일을 지휘하는 경우가 많아졌고 이들의 주된 야망은 투자를 통해 최대의 수익을 올리는 것이었다.

1484년 크리스토퍼스 콜롬버스는 포르투갈 왕을 찾아가 동아시아를 향한 새로운 무역로 개척을 제안한다. 하지만 왕은 그에 따른 막대한 비용을 고려하여 거절한다. 콜롬버스는 이어 이탈리아, 프랑스, 영국을 거쳐 다시 포르투갈로 돌아오며 똑같은 제안을 했지만 전부 거절당했다. 그는 이번에 새로 통일된 스페인의 통치자 페르디난도와 이사벨라에게 운을 시험해보았다. 결국 이사벨라 여왕을 설득하는데 성공하였고 곧 여왕은 대박을 터뜨리게 된다. 

그로부터 1백년 뒤, 왕과 은행가들은 콜롬버스의 후계자들한테 보다 더 많은 신용대출을 하기 원한다. 탐사에 대한 잠재력을 더 신뢰하게 되었으며 자기도 참여하고 싶어한 것이다. **이것이 제국주의적 자본주의의 마법의 순환이었다. 신용대출은 새로운 발견을 할 자금을 공급했고, 발견은 식민지로 이어졌고, 식민지는 수익을 제공했고, 수익은 신뢰를 만들어냈고, 신뢰는 더 많은 신용대출로 바뀌었다.**

하지만 원정은 위험도가 큰 사업이었다. 영국인들은 북극을 통해 아시아로 가는 항로를 개척하려 했지만 실패하였으며, 돌아오지도 못한 경우도 많았다. 그래서 유럽인들은 합자회사에 투자하기 시작한다. 한 대의 배에 투자하는 것이 아닌 많은 투자자들한테 돈을 모아 위험 부담을 줄이려고 하였다. 한 척에 배에 조금만 투자해도 성공하다면 백만장자가 될 수 있었다.

16세기 스페인은 유럽에서 가장 강력한 국가로서, 광대한 세계 제국을 가지고 있었다 유럽의 많은 부분과, 남북 아메리카의 땅덩이, 필리핀 제도, 아프리카와 연안의 수많은 기지를 지배했다. 아프리카와 아시아의 보물을 실은 선단이 해마다 세비야와 카다스 항구로 들어왔다.

1563년 개신교도인 네덜란드인은 카톨릭교인 스페인 군주로 싸움을 벌인다. 80년후 네덜란드는 스페인으로 부터 독립을 쟁취하며 네덜란드를 세계적인 제국으로 건설, **유럽에서 가장 부유한 나라로 발전하는데 그 비결은 바로 신용이었다.** 네덜란드 소시민은 용병을 고용하여 싸우게 하였다. 그동안 자신들은 바다로 나가 더 큰 선단을 꾸렸다. 

하지만 그렇기 위해서는 훌륭한 대포, 용맹한 군인들이 필요한데 이를 위해선 많은 돈이 필요했다. **하지만 네덜란드은 스페인보다 자금을 쉽게 조달할 수 있었다. 왜냐하면 급성장하는 유럽의 신용을 얻었기 때문이다.** 당시 스페인 왕은 금융의 신뢰를 갈아먹고 있었다.

네달란드는 어떻게 신뢰를 얻었을까? **첫째로 이들은 기일에 맞추어 전액을 반드시 갚았다. 둘째, 사법제도가 독립되어 있으며 사유재산권을 보호했다.** 네덜란드 제국을 세운 것은 네덜란드라는 국가가 아니라 상인들이었다.  스페인왕은 계속해서 정복사업을 벌이며 많은 자금 조달을 위해 세금을 높여 원성을 샀다. 네달란스 상인은 돈을 빌려 정복사업의 자금을 댓지만, 한편으로 자기 회사의 주식을 팔아서 자금을 조달하는 경우가 늘어났다. 

네덜란드의 가장 유명한 주식회사는 네덜란드 동인도회사로 1602년 설립되었다. 당시 네덜란드는 스페인의 지배를 떨쳐버리기 위해 투쟁중었다. 네덜란드 동인도회사는 주식을 팔아서 마련한 돈으로 선반을 건조하고, 그 배를 아시아로 보내 중국, 인도, 인도네시아의 재화를 실어왔다.



일부에서는 21세기 기업들은 너무 많은 권력을 가지고 있다고 경고한다. **근대 초기 역사를 보면, 기업이 이익을 무한히 추구하게 두면 어떤 일이 벌어지는 알 수 있다.** 네덜란드 동인도 회사는 인도양에서 맹활약한 반면, 네덜란드 서인도외사는 대서양에서 부지런히 움직였다.  허드슨 강에서 이루어지느 무역은 중요했기에, 회사는 이를 장악하기 위해 강 입구에 뉴암스테르감이란 정착지를 건설한다. 이 정착지는 1644년 영국의 수중에 들어가고 섬의 이름을 뉴욕으로 바꿨다. 

네달란드 서인도회사가 식민지에서 원주민과 영국인으로 부터 지키기 위해 세웠던 성벽의 잔해 위에 깐 도로는 세계에서 가장 유명한 거리, 월 스트리트가 되었다.
