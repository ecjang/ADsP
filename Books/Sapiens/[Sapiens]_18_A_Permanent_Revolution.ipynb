####  4부. 과학혁명

# 18. 끝없는 혁명

세상이 호모 사피엔스의 필요에 맞게 변형되고, 생태계가 파괴되고 종들은 멸종의 길을 걸었다. 과거 녹색과 푸른색이었던 우리의 행성은 이제 콘크리트와 플라스틱으로 만든 쇼핑센터가 되어가고 있는 중이다. 지구상에서는 70억명이 넘은 사피엔스들이 살고 있다. 모든 사람들을 거대한 저울에 올려놓는다면 그 무게는 약 3억톤이 될 것이다. 그리고 우리가 가축화한 농장의 동물(암소, 돼지, 양, 닭 등)을 더 거대한 저울에 세운다면 약 7억톤에 달할 것이다.

이와 대조적으로 현재 살아있는 야샹동물(심지어 고래까지 포함해도)의 무게를 모두 합혀도 1억톤에 못 미친다. **인류는 정말로 지구를 접수했다.** 생태계의 파괴에 대한 두려움은 너무나 확실하다. 미래의 사피엔스는 온갖 원자재와 에너지원의 보고를 손에 넣되 겨우 남아 있는 서식지를 파괴하고 대부분의 종을 멸종시킬지도 모른다.

인류가 자신의 힘으로 자연의 힘에 대항하고 생택를 자신의 필요와 변덕에 종속시킨다면, 미처 예상하지 모한 위험한 부작용을 더 많이 초래할지 모른다. 이를 통제하는 방법은 생태계를 극적으로 조작하는 것인데, 이것은 더더욱 큰 혼란을 초래할 것이다. **많은 사람들은 이것을 자연파괴라고 부른다.** 하지만 이것은 사실 파괴가 아니라 변형이다.**자연은 파괴되지 않는다.** 6,500년 전, 소행성이 공룡을 쓸어버렸지만, 그럼으로써 포유류가 번성할 수 있는 길이 열렸다.

산업혁명 아래 세계 인구는 사상 유래없이 급증했다. 1700년 세계 인구는 7억명이었다. 1800년에는 9억 5천만명이었다. 1900년이 되자 거의 2배로 늘어 16억 명이 되었고, 2000년에는 4배로 늘어 60억명이 되었다. 오늘날에는 70억명이 약간 넘는다.

### 18_1. 현대의 시간

산업혁명의 결과 사회공학적 실험이 계속해서 이어졌다. 많은 변화 중 하나는 전통농업의 리듬이 산업의 획일적이고 정밀한 스케쥴로 대체된 것이다. 전통 농업사회에서는 시계와 시간표가 없어도 잘 굴러갔으며 오직 태양의 움직임과 식물의 성장주기의 변화만 있었다. 사람들은 언제 우기가 시작되는지, 수확기는 언제쯤인지 정도만 관심이 있었고 지금이 몇 년도인지는 중요하지 않았다. 

이에 비해 현대의 산업은 태양이나 계절의 주기 보다는 정밀성과 획일성을 중요하다. 중세 구두가게에서는 한 명이 늦는다고 해도 다른 사람의 일이 늦어지지 않았지만 현대 공장에서는 한 사람 때문에 다음 공정으로 넘어가지 못할 수 있다. 이런 참사를 막기 위해 정확한 시간표가 존재하며 동일한 점심시간을 가지고, 근무시간 끝나면 모두가 같이 집으로 돌아간다.

산업혁명은 시간표와 조립라인이 거의 모든 인간들을 가두었다. 공장을 시작으로 병원, 정부기관, 식품점 등으로 퍼졌다. 이런 확산은 대중교통이 결정적인 역활을 했는데 출근시간에 맞추어 버스가 도착해야 했다.

1784년 영국에서 출발시간을 명시한 마차가 등장했다. 당시 각 지역간 시차가 있었지만 사람들은 크게 신경쓰지 않았다. 그 뒤 최초의 상업용 기차가 리버풀에서 멘체스터 사이를 운행했고, 10년이 지난 1830년 최초의 기차 시간표가 나왔다. 기차는 마차보다 훨씬 빨랐으므로 사람들은 지역간의 시차를 불편하게 생각하기 시작했다. 1847년 영국의 철도 회사들은 현지 시간이 아닌 그리니치 천문대 표준시에 맞추기로 합의한다.

**이는 전례가 없는 일이다. 역사상 처음으로 국가시간을 채택하고 해가 뜨고 지는 것이 아닌 시계가 시간의 기준이 된 것이다.** 이는 몇 시를 넘어 몇 십분의 일 초까지 정밀하게 맞추는 세계적 시간표 네트워크를 낳았다.

이런 시간표 네트워크를 위해 값싸고 정교한 휴대용 시계가 보급되었다. 현대의 사람들은 하루에도 몇 번이나 시계를 본다. 아침 7시에 일어나 아침밥을 60초 동안 가열하고, 7시 반에 집에서 나와 9시까지 출근한다. 심지어 정신과 의사들도 정확히 50분의 제한된 시간에만 상담해준다.

산업혁명은 여러가지 커다란 변화를 가져왔다. 산업적 시간에 적응하는 것은 그 중 하나에 불과하다. 다른 예로는 도시과, 농민의 소멸, 프롤레타리아의 등장, 보통 사람한테 주어진 힘, 민주화, 청년 문화, 가부장제 해체 등등이 있다. 하지만 가장 큰 변화는 **가족과 지역 공동체 -가 붕괴하고 국과와 시장이 대신 들어선 것이다.** 인류는 약 1백만연 년 전부터 대부분 친밀한 가족단위 공동체로 생활했지만 불과 2세기 남짓만에 이 단위는 산산히 부셔졌다. 가족과 공동체가 수행하던 기능은 국가와 시장으로 넘어갔다.
