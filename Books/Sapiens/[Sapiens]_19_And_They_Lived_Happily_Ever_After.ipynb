####  4부. 과학혁명

# 19. 그리고 그들은 행복하게 살았다.

지난 5백여년 시간동안 과학과 산업혁명으로 초인적인 힘과 실질적으로 무한한 에너지를 가지게 되었다. 사회질서는 완전히 바뀌었으며 정치, 일상생활, 인간의 심리도 그렇게 되었다. **하지만 우리는 더 행복해졌는가? 그 시간동안 쌓아온 부는 인류를 더 풍족하게 하였는가? 발전한 과학기술은 인류에세 만족을 주었는가? 무한한 에너지는 행복의 창고를 열어주었는가?** 달 표면에 발자국을 남긴 닐 암스트롱은 3만 년전 쇼베 동굴에서 손자국을 남겼던 어떤 수렵채집인 보다 더 행복했을까? 이제는 과학과 산업을 발전시키는 것이 무슨 의미가 있는지 생각해볼 때이다.

이것은 역사에 대한 가장 중요한 질문 중 하나이다. 현대의 이데올로기와 정치는 무엇이 진정 인류를 행복하게 해줄 수 있는지 모른다. 민족주의자는 자기결정권이, 공산주의자는 프롤레타리아의 독재가, 자본주의자는 자유시장이 최대행복을 보장한다고 한다. 하지만 이런 것이 틀렸다고 하면 어떨까? 만일 대제국의 신민이 독립국의 신민보다 더 행복하다면? 영국의 식민지배가 내부의 독재자의 지배보다 행복하다면? 만일 그렇가면 우리는 탈식민지화, 민족자결의 가치를 어디에 두어야 할까?

지금까지 역사학자들은 이런 질문 자체를 애초부터 기피했다. 모든 분야의 역사에 대해 연구했지만 그 모든 것이 인류의 행복에 어떤 영향을 미치는지 생각하는 일은 드물었다. **보통의 학자들은 역사가 진행되는 동안 인류의 능력도 같이 커졌고, 인류는 그 능력을 이용해 자신의 불행을 줄이고 소망을 충족하였으므로, 당연히 오늘날 인류는 중세시대 선조들보다 반드시 행복할 것이라는 선입견을 가지고 있다.** 하지만 익히 아는 데로 새로운 재능, 형태, 기술이 더 나은 내일을 약속하는 것이 아니다. 농부들은 이전의 수렵채집인보다 더 오래 노동을 하고 더 적게 먹었고 더 많은 질병을 겪었다. 유럽 제국의 확대로 상업과 기술이 발전했지만 수백만 명의 아프리카, 호주 사람들에게는 좋은 소식이 아니었다.

역사를 통해 인간이 권력을 남용한다는 것이 증명되었다. **더 많은 영향력을 누르면 더 행복해질 것 처럼 보인다. 하지만 반대파에서는 인간의 능력와 행복은 역관계가 존재한다고 한다. 권력은 부패하고, 인류가 점점 더 많은 힘을 가질 수록 우리의 욕구와는 동떨어진 차가운 기계적인 세상이 될 것이라고 한다.** 우리의 마음과 신체는 수렵채집인에 맞추어 주조되었지만 뒤이은 농업과 과학혁명은 타고난 성향과 본능을 제한하였다는 것이다. 매머드 사냥에 성공한 수렵채집인의 흥분과 기쁨은 도시의 중산층은 근접할 수 없을 것이다. 새로운 발명이 있을 때마다 우리는 에덴 동산에서 조금씩 멀어져갔다.

**하지만 이런 발명의 어두운 면만 고집해서는 안된다. 우리 내면은 수렵채집인과 끊겼을지라도 그게 꼭 나쁘기만 한 것은 아니다.** 의료기술의 발전으로 유아사망률이 엄청나게 낮아졌다. 죽을 운명의 어린이가 살 수 있다는 것이 행복하지 않다라고 할 수 있는 사람이 있을까? 

그리고 이런 빛과 그림자에서 미묘한 중도를 취하는 입장도 있다. **지난 긴 사간동안 인류는 자신의 능력을 현명하게 사용하는 법을 배웠다. 의학을 발전, 극감한 폭력, 사라진 전쟁과 기근 등이 그런 예이다.** 하지만 문제가 있다. 첫째로, **낙관적인 평가 기간이 너무 짧다.** 눈부신 현대 의학은 발전은 1850년 이후이며, 20세기에 들어서야 유아사망율이 낮아졌다. 대규모 기근은 20세기까지 그 영향을 미쳤고, 핵무기가 등장하기 전까지 국제전은 1945년에도 있었다. 둘째로, **최근 몇십년이 전대미문의 황금기가 미래의 파국의 씨앗이 될지도 모른다는 것이다.** 인류는 지구의 생태적 균형을 수없이 많은 방법으로 교란시켰고, 이제는 그 끔찍한 결과를 확인하고 있는 중이다. 

**결론적으로 인류는 다른 모든 동물의 운명을 깡그리 무시할 때만 전례없는 성취를 자축할 수 있다.** 가난과 기근으로 보호해주는 물질적 풍요는 실험실의 원숭이, 컨테이너 벨트의 병아리, 농장의 돼지와 소들의 희생덕부니다. 동물권리 운동가들의 주장을 10%만 받아들여도 현대의 기업농은 역사상 가장 큰 범죄를 저지르고 있다고 할 수 있다. 지구 전체의 행복은 상류층이나 유럽인이나 남자만을 대상으로 하지 않느다. 인류만의 행복을 고려하는 것도 마찬가지의 오류를 범하는 것이다.

### 19_1. 행복 계산하기

과거 철학자, 사제, 시인들은 수천 년간 행복에 대해 곰곰히 생각해본 결과, 행복은 물질적인 부분만큼이나 사회적, 정신석, 윤리적인 조건이 중요하다고 결론지었다. 그래서 우리들의 선조들은 공동체, 종교에서 큰 만족감을 느끼고, 현대인들은 풍요속에서 소외와 고통을 받고 있는 것은 아닐까. 최근 심리학자와 생물학자는 행복의 본질을 연구하기 시작했다.

우선 첫번째로 **무엇을 측정해야 할지 정해야 한다. 행복에 대한 보편적 정의는 '주관적 안녕'이다.** 내 삶에서 내가 느끼는 즉각적인 기쁨이나 장기적인 만족감으로, 이것을 측정하기 위해 설문조사를 실시하였다. 각각 연봉이 10만, 5만 달러인 1천만 명의 사람들한테 얼마나 행복한지 측정하는 방식이다. 조사 결과, **돈은 실제로 행복을 가져다 주지만 어느 정도까지만 작용한다.** 자신의 생활 수준이 올라갈 수록 더 많은 행복감을 느끼겠지만 곧 익숙해지며 장기적으로 지속되지 않았다.

또 다른 흥미로운 결과는 **질병과 행복의 관계**이다. 질병이 행복감을 낮추는 것은 너무나 당연한 사실이다. 하지만 질병 또한 돈과 마찬가지로 어느정도의 범위 이상만 그 악효과가 나타난다. 좀 더 구체적으로 말하면 점점 상태가 악화되거나 오랫동안 지속되는 고통과 질병이라고 할 수 있다. 당뇨병 같은 만성질환을 겪는 사람은 당연 우울해지지만 더 악화되지만 않는다면 상황에 바로 적응할 수 있다.

**가족과 공동체는 앞의 돈과 질병보다 행복에 더 큰 영향을 줄 가능성이 크다.** 강한 유대감을 가진 가족은 그렇지 않은 가족보다 더 행복해보인다. 그런 면에서 결혼은 무척이나 중요하다. 최근의 연구결과에서는 좋은 결혼은 행복과, 나쁜 결혼은 불행과 밀접한 관련이 있다고 말한다. 무일푼이라도 사랑하는 배우자와 헌신적인 가족이 있다면 억만장자보다 행복할 수 있다.

그렇다면 지난 세기동안 진행된 가족의 붕괴로 많은 현대인들은 1800년도 사람들보다도 행복하지 않을 가능성이 높다. 개인의 삶을 결정하는데 전례없이 큰 힘을 누리게 되었지만 그만큼 동시에 남에게 헌신하기 힘들어지고, 가족과 공동체가 해체되고, 점점 더 외로운 세상에 가까워지고 있다.

이런 사실들을 알기 위해 엄청난 설문자료가 필요하지 않다. 이미 수천년 전부터 예언가, 철학자, 시인들은 가진것에 만족하라는 교훈을 말해왔고, 현대의 연구결과과 도표들 또한 비슷한 결론에 도달한다.**인간의 기대가 결정적으로 중요하다는 점**은 행복의 역사에 지대한 영향을 끼친다. 현대인들은 안정제와 진통제를 얼마든지 사용할 수 있게 되었지만, **안락함과 즐거움은 크게 기대하면서 불편함과 불쾌함은 점점 더 참지 못하게 되었다. 그 결과 선조들보다 더 많은 고통을 겪는다고 해도 무리가 아니다.**

정말 큰 문제는 우리의 정신 속에 깊이 박혀있는 추론의 오류에 있다. 다른 사람이 얼마나 행복한지 상상할 때, 우리 자신의 상황을 대입하거나, 혹은 자신의 기대를 타인의 물질적인 조건에 끼워넣기 때문이다. 예컨데, 현대인들은 매일 씻고 옷을 갈아입지만 중세 농부들은 몇 달동안이나 씻지 않고 옷을 갈아입지 않는 것이 당연하였다. 10대 아이는 과거라면 무척이나 높은 자존감을 받을 수 있을 테지만, 안타깝게도 오늘날에는 TV, 페이스북, 광고판에 등장하는 슈퍼모델, 영화배우 등과 자신을 비교하며 스스로를 부적격자로 느낄 수 있을 것이다. 이런 것이 바로 **기대의 문제**다. 

**심지어 오늘날에서는 '영원한 생명'까지도 불만이 될 수 있다.** 이 새로운 기적의 요법을 받을 수 없는 대부분의 가난한 사람들은 격렬한 분노에 휩싸일 것이다. 역사를 통들어 가난하고 고통받는 사람들의 가장 큰 위안 거리를 누구한테나 공평하게 적용되는 죽음이었다. 그런데 자신은 죽을 수 밖에 없는데 돈이 많은 사람은 영원히 젊고 아름답게 살 수 있다고 하면 마음이 편치 않을 것이다. 

하지만 그 선택받은 극소수의 사람들이라고 행복하지만은 않을 것이다. 걱정해야 할 일들이 많아지기 때문이다. 이 새로운 요법은 영원한 삶을 보장할 수 있지만 시체를 살릴 수는 없다. 트럭에 치이거나 폭탄으로 산산조각나면 그걸로 끝이다. 영원한 삶을 얻어도 조그마한 위험도 기피하게 되고, 배우자나 자녀, 친한 친구를 잃는데 따르는 고통을 견딜 수 없어 질 것이다.

### 19_2. 화학적 행복

##### 행복 프로그램

우리 신채에는 행복 수준을 일정하게 유지하도록 설정되어 있는 듯 하다. **진화에서 행복과 불행은 생존과 번식에 깊은 관련이 있다.** 그래서 진화의 결과로 보면 우리들은 행복하게 만들어지지 않다는 것을 알 수 있는데, 쾌락은 일시적으로만 작용하고 영원히 지속되지 않기 때문이다. 예를 들어, 남자는 가임기의 여성과 성관계를 해서 유전자를 퍼뜨리면 쾌락이라는 보상이 주어지지만, 이 쾌락은 오래 지속되지 않는다. 만일 영원히 오르가즘을 느끼는 행복한 남자는 그 어떤 것에도 흥미를 잃고 여자를 찾지 않을 것이다.

많은 사람들이 지금의 고통에서 벗어나면, 복권에 당첨되면, 근사한 선물을 받으면 엄청나게 행복할 거라고 생각하지만 실제로 그런 것을 얻어도 그렇게 행복해지지 않는다. **우리의 행복 생화학 시스템을 변화시킬 수 있지만 곧 개인마다의 원래의 설정값으로 돌아간다.** 행복한 사람은 그 수치가 높을 것이고, 그렇지 않은 사람은 보다 낮을 것이다.

##### 행복에 대한 생물학적 접근

**우리가 행복에 대한 생물학적 접근을 받아들인다면, 이것은 곧 역사는 별로 중요치 않다는 의미가 된다.** 대부분의 역사적인 사건들은 우리의 생화학에 영향을 미치지 않기 때문이다. 역사는 세로토닌 분비를 유발하는 외부자극을 변화시킬 수 있지만, 그 결과로 세로토닌 수준을 바꾸지 않는다. 따라서 사람들을 더 행복하게 만들 수 없다. 

프랑스 혁명 당시 왕을 처형하고, 땅을 분배하고, 인권 선언이 이루어지는 모든 과정중 어느 것도 프랑스 인들의 생화학 시스템을 바꾸지 못했다. 그 결과 혁명의 엄청난 이데올로기에도 불구하고 프랑스 인의 행복에 미친 영향은 크지 않다. 그럼 사람들이 조금도 행복해지지 않았다면 그 모든 혼란, 공포, 전쟁은 무슨 의미가 있을까? **실질적인 중요성을 지닌 역사적 진전은 하나, 오늘날 우니는 진정한 행복의 열쇠가 생화학 시스템에 달렸다는 것을 인식하게 되었다는 점이다.** 즉, 개혁이나 반란에 시간을 낭비하는 대신 우리를 정말로 행복하게 해주는 일에 집중할 수 있게 되었다.

##### 세로토닌, 도파민, 옥시토신

뇌의 생화학 시스템을 이해한다면 엄청난 투자가 없어도 충분히 행복할 수 있다. 프로작(항우울제)는 세로토닌의 수치를 높여줌으로 우울증에서 빠져나오도록 도와준다. 사회적 지워, 성형수술, 아름다운 집 등은 전혀 행복을 가져다 주지 못하지만 세로토닌, 도파민, 옥시토신은 지속적인 행복을 전해준다.

미국 대공황 절정의 시기 1932년에 출간된 올더스 헉슬리의 디스토피아 소설 [멋진 신세계]속에서 행복은 인류 최고의 가치이고, 향정신 약물이 정치의 기반이 되어 모든 사람이 '소마'라는 약물을 복용한다.생산성과 효율성을 해지지 않고 사람들을 행복하게 만들어주는 합성 마약이다. 모슨 사람이 현재 사정이 어떻든 대단히 행복하지만 우리의 마음은 불편해하지만 이걸 설명하기가 힘들다. 모든 사람이 행복하다는데 거기에 무슨 문제가 있을 수 있을까?

### 19-4. 삶의 의미

#### 행복의 정의

헉슬리의 가상의 행복한 세계는 쾌락이 행복과 동일하다는 기반에서 시작한다. 즉, 높은 행복도는 사람들의 생화학 시스템을 조작하여 쾌락 상태를 유지하는 것이다. **노벨경제학상 수상자인 대니얼 카이먼**은 사람들에게 하루의 활동을 평가하는 실험을 하였는데, 이 실험에서 **스스로의 삶을 보는 시각에 대한 역설적인 부분을 발견했다.** 아이의 귀저기를 갈아주고, 설거지를 하고, 육아를 하는 일은 불쾌한일로 평가하지만 대부분의 부모는 아이가 행복의 원천이라고 말한다.**자신이 무엇에 행복을 느끼는지 정확히 모르는 것일까?**

다른 시각에서는 **행복이란 불쾌한 순간을 상쇄하고도 남은 여분의 즐거움의 총합이 아니라, 중요한 인지적, 윤리적 요소가 존재한다고 본다.** 위의 부모들은 '아기 독재자의 비참한 노예'로 볼 수도, 아니면 '사랑을 다해 생명을 키우고 있는 사람'으로 볼 수도 있다. 이 차이를 설명하는 것은 우리의 가치다.

#### 행복의 의미

중세의 사람들은 현대에 비해 너무나 열약함에는 틀림이 없다. 하지만 그들이 사후에는 영원한 행복이 온다고 믿었다면, 그들이 현대의 사람들보다 더 행복하게 살았다고 평가할 수 도 있다. 순수과학의 입장에서 볼 때 **인간의 삶은 절대 아무런 의미가 없다. 인류의 목적이나 의도같은 것은 없이 진행되는 눈 먼 진화과정의 산물이다. 우리의 행동은 뭔가 신성한 우주적 계획의 일부가 아니다.** 그러므로 자신의 삶에 어떤 의미를 부여하던 망상에 지나지 않는다. 

중세 사람들이 숭배했던 의미는 현대의 인본주의와 민족주의적 의미보다 더 심한 망상이 아니었다. 고향을 지키는 병사, 회사를 키우는 기업가가 가지고 있는 의미가 중세시대 십자군에 참여한 군인과 성당을 짖는 성직자가 숭배했던 의미보다 더 환상적이라고 할 수 없다. **행복에 대한 관건은 의미에 대한 개인의 환상을 폭넓게 퍼진 집단적 환상에 맞추는데 있는지도 모른다.** 개인과 집단의 내러티브가 일치할 때 내 삶의 의미를 확신할 수 있고, 그 확신을 통해 행복할 수 있다. **행복은 정말로 자기자신한테 달려있는게 아닐까?**

### 19-5. 너 자신을 알라

#### 주관적인 생각에 대해

오늘날 사회는 자유주의이다. 개인의 주관적 기분을 최고로 치고, 선과 악은 각각 개인이 어떻게 느끼느냐에 따라 달라진다. 이 시대의 **인문학은 이처럼 스스로 생각할 것을 강조한다. 장 자크 루소의 "자신에게 충실하라", "자신에게 귀를 귀울여라", "내면의 소리를 따르라"라는 견해 가장 좋은 고전적인 슬로건이다.** 하지만 대부분의 종교나 이데롤로기는 선과 악의 객관적인 척도를 규정한다. 기독교 신학자는 인간은 죄 많은 존재라 쉽게 사탄의 유혹에 빠진다고 할 것이다. 

주관적인 느낌을 신뢰할 수 없다고 말하는 것은 기독교만이 아니다. 찰스 다윈, 러처드 도킨스도 성 바오로, 성 아우구스티누스와 같은 입장에 설 것이다. **이기적 유전자 이론에 따르면 자연선택은 사람들로 하여금 유전자 복제에 좋은 행동을 하게 만든다. 설령 그것이 개체로써 자신에게 나쁜 영향을 주더라도 말이다.** 많은 사람들이 가정의 평화를 위해 노동과 경쟁을 하는 것은 자신의 DNA 때문이라는 것이다. 사탄과 마찬가지로 DNA도 덧없는 기쁨을 이용해 사람들을 유혹하고 사진의 손아귀에 넣는다.

#### 불교에서의 행복

불교는 2,500여년 동안 행복에 대해 고민하고 연구했다. 불교의 접근법과 생물학적인 접근법은 기본적은 통찰의 측면에서 비슷한 부분이 있는데 **행복은 외부 세계가 아니라 신체 내부에서 일어나는 과정의 결과라고 생각한다.** 하지만 그 결과는 매우 달랐다. 대부분의 사람들은 행복을 즐거운 감정, 고통을 불쾌한 감정과 동일시 하여 고통을 피하고 더 많은 즐거움을 추구한다고 본다. **문제는 우리의 감정이 매 순간 파도처럼 변화한다는 것이다.** 지금 즐겁다가도 5분 뒤에 바로 우울해질 수도 있고, 끊임없이 즐거움을 추구하려 노력해도 그 보상은 없는데, 그토록 덧없는 보상을 받는 것이 뭐가 그리 중요한가? 나타마자 바로 사라질 수 있는 행복을 위해 그토록 힘들게 노력해야 할까? **불교는 번뇌의 근원을 이처럼 순간적인 감정을 무의미하게 끊임없이 추구하는데 있다고 본다.** 이렇기 때문에 우리 마음은 결코 만족하지 못하고 즐거움을 느껴도 만족스럽지 못한 것이다.
