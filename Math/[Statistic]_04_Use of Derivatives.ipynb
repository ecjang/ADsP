### 04_1. 접선의 방정식



##### 곡선 위의 점에서의 접선의 방정식

함수 $f(x)$가 미분가능할 때, 곡선 $y=f(x)$ 위의 점 $P(a, f(a))$에서의 접선의 기울기는 $x=a$에서의 미분계수 $f'(a)$와 같다.
- **점 $P(a,f(a))$에서의 접선의 방정식 : $y-f(a)=f'(a)(x-a)$** <br><br>
- **점 $P(a,f(a))$과 수직인 방정식 : $y-f(a)=-\frac{1}{f'(a)}(x-a) \text{ (단, f'(a)≠0)} $**

#####  예 : 곡선 $y=x^3+x^2$ 위의 점 $P(1,2)$ 에서의 접선의 방정식
$y'=3x^2+2x$이므로 기울기는 $y'_{x=1}=3+2=5$이므로 접선의 방정식은 아래와 같다.
- 접선의 방정식: $ y-2=5(x-1) \,=\, y=5x-3 $


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-2,3, 0.01)
plt.plot(x,x**3 + x**2, label="$f(x)$")
plt.plot(x, 5*x-3, label="$y=5x-3$")
plt.plot(1,2,'o')
plt.grid()
plt.legend()
plt.annotate('P(1,2)',xy=(1,2), xytext=(0.2,5), fontsize=20 )
plt.show()
```

##### 기울기가 주어진 접선의 방정식

함수 $f(x)$가 미분가능할 때, 기울기가 $m$이고 곡선 $y=f(x)$에 접하는 **직선의 방정식**은 다음과 같은 순서로 구한다.
1. 접점을 $(t, f(t))$로 놓는다.
1. $t$에 대한 방정식 $f'(t)=m$을 만족시키는 실수 $t$의 값을 구한다.
1. 접선의 방정식 $y=f(t)=f'(t)(x-t)$를 구한다.

##### 예: 곡선 $y=x^3$에 접하고 기울기가 3인 직선의 방정식
- 접점을 $(t,t^3)$으로 놓으면 $y'=3x^2$이므로 $3t^2=3$ 이므로 $t=±1$이다. 
- 점 $P(-1,-1)$에서의 접선의 방정식은 $y+1=3(x+1)$에서 $y=3x+2$ 이다.
- 점 $Q(1,1)$에서의 접선의 방정식은 $y-1=3(x-1)$에서 $y=3-2$이다.

```python
plt.plot(x,x**3, label="$y=x^3$")
plt.plot(x,3*x+2, label="$y=3x+2$")
plt.plot(x,3*x-2, label="$y=3x-2$")
plt.plot(-1,-1,'o'); plt.plot(1,1,'o')
plt.annotate('P(-1,-1)', xy=(-1,-1), xytext=(-1.5,2), fontsize=20)
plt.annotate('Q(1,1)', xy=(1,1), xytext=(1,-2), fontsize=20)
plt.grid();plt.legend();plt.show()
```

##### 곡선 밖의 한점에서  곡선에 그은 접선의 방정식

함수 $f(x)$가 미분가능할 때, 곡선 $y=f(x)$ 밖의 한 점 $(a,b)$에서 곡선 $y=f(x)$에 그은 접선의 방정식은 다음과 같은 순서로 구한다.

1. 접점을 $(t,f(t))$로 놓는다.
1. 점 $(t,f(t))$에서 접선의 방정식 $y-f(t)=f'(t)(x-t) ...㉠ $을 구한다.
1. 점 $(a,b)$는 접선 위의 점이므로 $x=a, y=b$를 ㉠에 대입한다.
1. $t$에 대한 방정식 $b-f(t)=f'(t)(a-t)$에서 실수값 $t$를 구한다.
1. 접선의 방정식 $y-f(t)=f'(t)(x-t)$를 구한다.

##### 예: 원점에서 곡선 $y=x^2+1$에 그은 접선의 방정식
- 접점을 $(t,t^2+1)$로 놓으면 $y'=2x$이므로 접선의 방정식은 $y-(t^2+1)=2t(x-t)$이므로 접선의 방정식은 $y-(t^2+1)=2t(x-t)$이다.
- 접선은 원점을 지나므로 $-(t^2+1)=2t(-t)$에서 $t^2=1$ 이므로 $t=±1$ 이다.
- $y-2=-2(x-1)$ 에서 $y=-2x$
- $y-2=2(x-1)$ 에서 $y=2x$

```python
x = np.arange(-2,2, 0.01)
plt.plot(x,x**2+1, label="$y=x^2+1$")
plt.plot(x, -2*x, label="$y=-2x$")
plt.plot(x, 2*x, label="$y=2x$")
# plt.plot(1,2,'o')
# 

# plt.annotate('P(1,2)',xy=(1,2), xytext=(0.2,5), fontsize=20 )
plt.grid(); plt.legend(); plt.show()
```

##### 두 곡선에 접하는 접선의 방정식 (접점이 같을 경우)

두 곡선 $y=f(x), y=g(x)$가 한 점 $P$에서 만나므로 점 $P$의 x좌표를 $t$라고 하면 $x=t$에서의 함숫값은 서로 같다.<br>
또한, **점 $P$에서의 접선은 같은 직선이므로 $x=t$에서의 미분계수는 값다.** 즉 다음 등식이 성립한다.

$$ f(t)=g(t), \quad f'(t)=g'(t) $$


```python
x = np.arange(-10,10, 0.01)
plt.plot(x,(x-2)**2 + 2, label="$f(x)$")
plt.plot(x,-(x-2)**2 + 2, label="$g(x)$")
plt.plot(x, 1*x, label="$y=x$")
plt.plot(2, 2 ,'o')
plt.annotate('P(2 ,2)',xy=(1,2), xytext=(-1,20), fontsize=20 )
plt.grid(); plt.legend(); plt.show()
```

##### 두 곡선에 접하는 접선의 방정식 (접점이 다를 경우)

두 곡선 $y=f(x), y=g(x)$의 접점을 각각 $P(\alpha, f(\alpha)), Q(\beta, g(\beta))$로 놓으면 두 점 $P,Q$에서의 접선은 같은 직선이다.<br>
점 $P$에서의 접선은 $y=f'(\alpha)(x-\alpha)+f(\alpha)$이고, 점 $Q$에서의 접선은 $y=g'(\beta)(x-\beta)+g(\beta)$이므로 다음과 같은 등식이 성립한다.

$$ f'(\alpha)=g'(\beta), \quad -\alpha f'(\alpha)+f(\alpha) = -\beta g'(\beta)+g(\beta) $$
