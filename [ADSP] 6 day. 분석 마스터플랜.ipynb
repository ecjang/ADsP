<h1><center>1절. 마스터플랜 수립 프레임워크</center></h1>

## 1. 분석 마스터플랜 수립 프레임워크 
분석 과제를 대상으로 전략적 중요도, 비즈니스 성과 및 ROI, 분석 과제의 실행 용의성 을 고려한다.

* [ISP Information Strategy Planning 정보전략계획] : 정보시스템의 전략적 활용을 위해 문제점을 도출, 사용자 요구사항을 분석하여 중장지 마스터플랜을 수립하는 절차

* [AMP Analysis Master Planning 분석 마스터플랜] : 일반적인 ISP를 활용하여 과제의 우선순위를 단, 중, 장기로 나누어 결정한 계획

## 2. 우선순위 평가 방법 및 절차
ISP와 같은 과제 우선순위 평가를 위해 전략적 중요도, 실행 용의성에 따라 우선순위 기준을 수립한다.

* [Strategic Importance 전략적 중요도] : 전략적 필요성(직접적인 연관, 미해결시 위험 정도), 시급성(경쟁우위 확보를 위한 중요도)

* [Ease of Execution 실행 용이성] : 투자 용이성(기간 및 인력, 투자예산 확보 가능성), 기술 용이성(적용 기술의 안정성, 유지보수 정도, 기술 성숙도)



### 2_1. ROI 관점의 특징

[ROI Return of Investment 투자회수] 관점에 따라 [Volume 크기], [Variety 다양성], [Velocity 속도] 3V가 있다.

* 비즈니스 [Return 효과] 요소 : [Value 가치] (데이터 분석의 결과 및 목표 가치)

* 4V : 크기, 다양성, 속도, 가치 

 

### 2_2. 우선순위 평가 기준

데이터 분석 과제를 추진할 때 어떤 것을 우선적으로 다룰지 평가하는 기준이 있다.

* [Ugency 시급성] : 가장 중요한 판단 기준으로 현재 관점, 중장기적 관점에 집중할지 고려하여 판단

* [Difficulty 난이도] : 현시점에서 과제 추진을 적용 비용과 범위 측면에서 판단



### 2_3. 포트폴리오 사분면

우선순위 선정 기준으로 토대로 4가지 유형으로 구분할 수 있다.

* 4가지 유형 : 1 사분면(급하고 어려움 ), 2 사분면(여유로운데 어려움), 3 사분면(급한데 쉬움), 4 사분면(여유롭고 쉬움) 

* 3 사분면이 가장 우선적으로 분석이 필요, 2 사분면이 가장 우선순위가 낮음

* 시급성에 중점 : 3 -> 4 -> 2 사분면, 쉬운 일 중 당장 급한 거부터 처리

* 난이도에 중점 : 3 -> 1 -> 2 사분면, 당장 급한 일중 쉬운 거부터 처리


## 3. 이행계획 수립
### 3_1. 로드맵 수립 

포트폴리오 사분면을 통해 1차적 우선순위를 결정한 뒤 단계적 구현 로드맵을 수립한다. 추진 과제별 선후행 관계를 고려하여 단계별 추진 내용을 정렬한다.

<h1><center>2절. 분석 거버넌스 체계 수립</center></h1>

## 1. 거버넌스 체계
단순히 데이터를 수집하는 것보다 어떤 목적으로 어떤 데이터를 어떻게 분석하느냐가 더 중요하기 때문에 데이터 분석을 위한 체계적인 관리가 중요하다.

* 거버넌스 체계 구성 : 조직, 프로세스, 시스템, 데이터, 육성 체계


## 2. 데이터 분석 수준 진단
데이터 분석의 도입을 위해 현재 분석 수준을 점검할 필요가 있다. 진단을 통해 무엇을 보충할지 알 수 있다.

* 분석 [준비도 Readiness]와 분석 [성숙도 Maturity] 2가지 프레임워크를 사용해 진단



### 2_1. [Analysis Readiness 분석 준비도]

기업의 데이터 분석 도입 수준을 파악하기 위한 진단 방법.

* 6가지 구성 : UOTDCI, 분석 업무 [Understanding 파악], 인력 및 [Organization 조직], 분석 [Technique 기법], 분석 [Data 데이터], 분석 [Culture 문화], IT [Infra 인프라]



### 2_2 [Analysis Maturity 분석 성숙도]

조직의 성숙도 평가 도구 [CMMI Capability Maturity Model Integration 능력 성숙도 모델 통합] 모델.

* 4가지 수준 : IUDO, [Introduction 도입], [Use 활용], [Diffusion 확산], [Optimization 최적화] 단계

* 시스템을 구축(도입 단계)하고, 분석 결과를 실무에 적용(활용 단계)하고, 전사 차원에서 공유(확산 단계)하고, 혁신 및 향상에 기여(최적화 단계) 한다



### 2_3 분석 관점에서 사분면 분석

분석 수준 진단의 결과를 준비도와 성숙도로 구분하여 목표 방향을 결정하거나 유형별로 개선방안을 수립한다. 

* [Diffusion 확산] : 높은 준비도, 높은 성숙도, 지속적인 확산이 필요

* [Introdution 도입] : 높은 준비도, 낮은 성숙도, 분석 업무는 부족하지만 바로 도입할 수 있음

* [FIxed 정착] : 낮은 준비도, 높은 성숙도, 내부에서 제한적으로 사용하여 1차적으로 정착이 필요

* [Ready 준비] : 낮은 준비도, 낮은 성숙도, 사전 준비가 필요.

## 4. 데이터 거버넌스 체계
[Data Governance 데이터 거버넌스]는 데이터의 표준화 관리 체계를 수립하고 운영을 위한 [Framework 프레임워크] 및 [Repository 저장소]를 구축하는 것을 의미한다. 주로 마스터 데이터, 메타 데이터, 데이터 사전을 관리하고 데이터의 가용성, 유용성, 통합성, 보안성, 안정성을 확보한다.

* 구성요소 : [Principle 원칙], [Organization 조직], [Process 프로세스]



### 4_1. 데이터 거버넌스 체계

* [Data Standardization 데이터 표준화] : 표준 용어, 명명 규칙, 데이터 사전, 메타 데이터 등을 구축

* [Data Management System 데이터 관리 체계] : 메타데이터, 데이터 사전 관리 원칙 수립, 데이터 생명주기 관리방안을 수립

* [Data Reportsitory System 데이터 저장고 관리] : 전사 차원의 저장소를 구성, 데이터 구조 변경 사전 영향도 평가

* [Standardization Activaties 표준화 활동] : 표준 준수 여부를 주기적으로 점검, 변화 관리 및 교육 진행

## 5. 데이터 분석 조직
비즈니스 질문에 부합하는 가치를 찾고 최적화하기 위해 인사이트를 찾고 행동하는 조직과 인력이 필요하다.



### 5_1. 3가지 구조

* [Concentration 집중] : 전사 분석 업무를 한 부서에서 담당, 이중화 및 이원화 가능성 큼 

* [Functional 기능] : 각 부서에서 기존의 분석업무를 진행. 핵심 분석 어려움

* [Distribution 분산] : 분석 인력을 각 부서에 배치, 분석 결과의 신속한 행동이 가능



### 5_2. 분석 조직 인력 구성

* 비즈니스 인력 : 비즈니스를 이해하고 분석 요소를 탐색

* IT기술 인력 : 분석에 필요한 IT 기술 동향 파악, 기술 아키텍처를 수립

* 분석 전문 인력 : 고급 통계 기법을 이해, 모델링 설계 및 검증 가능

* 변화관리 인력 : 경영층을 대상으로 분석 문화 확안, 변화관리를 담당 (겸직 가능)  

* 교육담당 인력 : 분석 기법에 대한 심도 있는 교육 담당 (겸직 가능)



## 6. 관리 프로세스
분석 과제를 체계적으로 관리하기 위한 프로세스를 수립해야 하며 크게 [Discovery Task 과제 발굴]과 [Perform Task 과제 수행]으로 구분한다.

* 과제 발굴 : 분석 아이디어 발굴 -> 분석 과제 후보 제안 -> 분석 과제 확정

* 과제 수행 : 팀 구성 -> 분석 과제 수행 -> 진행 관리 -> 결과 공유 및 개선

## 7. 분석 교육 및 변화 관리
최근 구성원들에게 데이터 분석을 교육하는 경우가 늘고 있다. 하지만 새로운 체제 도입은 기존의 성향으로 돌아가려는 관성을 야기할 수 있으므로 적극적인 변화관리가 필요하다.

* 분석 내재화 단계 : 준비기 -> 도입기 -> 안정 추진기

* 보통 도입기에 많이 포기하므로 성공 시 강한 탄성으로 변화를 가속화
